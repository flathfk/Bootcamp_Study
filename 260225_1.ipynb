```python
%%writefile ex1.js

// 1. const: 변하지 않는 값
const PI = 3.14159;
const myName = "Gemini";
// myName = "AI"; // 에러 발생: Assignment to constant variable.

// 2. let: 변하는 값
let score = 0;
score = 10; // 재할당 가능
console.log("현재 점수:", score);

// 3. const와 객체 (참조 타입의 특이점)
const user = { id: 1, name: "철수" };
user.name = "영희"; // 객체 내부의 속성은 변경 가능 (메모리 주소는 그대로이기 때문)
console.log("사용자 정보:", user);
```

const는 한 번 값을 넣으면 바꿀 수 없음

let은 나중에 값을 바꿀 수 있음

const가 막는 건 *변수가 가리키는 대상을 바꾸는 것* : 객체는 메모리에 저장되고, user는 그 위치 (주소)를 가지고 있음

즉, user.name = "영희" 는 주소는 그대로 두고 *그 안의 내용만 수정*하는 거라 const가 막지 않음

```python
%%writefile ex2.js

// 1. if...else 문
const age = 21;
if (age >= 20) {
    console.log("성인입니다.");
} else {
    console.log("미성년자입니다.");
}

// 2. switch 문 (요일 확인)
const day = "월";
switch (day) {
    case "월":
        console.log("월요병 극복 중...");
        break;
    case "금":
        console.log("불타는 금요일!");
        break;
    default:
        console.log("평범한 하루입니다.");
}

// 3. 삼항 연산자와 엄격한 일치(===)
const input = "10";
const message = (input === 10) ? "숫자 10입니다." : "숫자 10이 아니거나 타입이 다릅니다.";
console.log(message); // "10"은 문자열이므로 거짓 출력
```

=== (*엄격한 일치*)는 값과 타입을 둘 다 비교함

"10"은 문자열, 10은 숫자

값은 같아 보여도 타입이 다르기 때문에 === 비교는 거짓

참고로 == (느슨한 일치)는 타입을 무시하고 비교해서 "10" == 10을 참으로 봄 (JS에선 이것 때문에 버그가 생기기 쉬워서 ===를 쓰는 것이 기본 원칙)

```python
!node ex2.js
```

```python
%%writefile ex3.js

// 1. 기본 for문: 1부터 5까지 출력
console.log("--- for문 시작 ---");
for (let i = 1; i <= 5; i++) {
  // let i = 1 -> 시작값
  // i <= 5 -> 이 조건이 참인 동안 반복
  // i++ -> 한 번 돌 때마다 i에 1 더하기
    console.log(`현재 번호는 ${i}입니다.`);
}

// 2. while문: 조건이 충족될 때까지 반복
let energy = 3;
console.log("--- while문 시작 ---");
while (energy > 0) {
    console.log(`에너지 소모 중... 남은 에너지: ${energy}`);
    energy--; // energy--;는 energy = energy -1 의 단축 표현
}
console.log("에너지가 방전되었습니다.");

// 3. for...of문: 배열 데이터 순회
const fruits = ["사과", "바나나", "포도"];
console.log("--- for...of문 시작 ---");
for (const fruit of fruits) {
    console.log(`내가 좋아하는 과일: ${fruit}`);
}

// 4. for...of문: 배열 데이터 순회 예제
const citys = ["서울", "대전", "대구", "광주", "울산", "부산", "세종"];
console.log("--- for...of문 시작 ---");
for (const city of citys) {
    console.log(`대한민국의 도시들: ${city}`);
}

```

*for* -> 몇 번 반복할지 횟수가 명확할 때

*while* -> 횟수는 모르고 조건만 있을 때

*for...of* -> 배열 요소를 하나씩 꺼낼 때

```python
!node ex3.js
```

```python
%%writefile ex4.js

// 1. 함수 표현식: 두 수의 합을 구함
const addNumbers = function(a, b) {
    return a + b;
};
const sum = addNumbers(10, 20);
console.log("합계:", sum);

// 2. 화살표 함수: 인사 메시지 생성
// 로직이 한 줄일 때는 { }와 return을 생략할 수 있습니다.
const getGreeting = (userName) => `안녕하세요, ${userName}님! 오늘 하루도 화이팅!`;
console.log(getGreeting("소라"));

// 3. 함수를 인자로 넘기는 콜백 체험 (일급 객체)
const runner = (action) => {
    console.log("달리기를 시작하기 전에...");
    action(); // 인자로 받은 함수를 실행
};

runner(() => console.log("열심히 달립니다! =3"));
// runner를 호출하면서 () => console.log("열심히 달립니다! =3")라는 함수를 인자로 넘김
// 이렇게 다른 함수의 인자로 넘겨지는 함수를 '콜백 함수'라고 함
```

const addNumbers는 변수를 만드는 것. 근데 그 변수에 숫자나 문자열 대신 함수 자체를 담음. *함수 표현식*이라고 함

*화살표 함수*는 함수를 더 짧게 쓰는 문법 0> function 대신 => (화살표)를 씀

: 로직이 한 줄이면 {}와 return을 생략 가능

action이라는 매개변수엔 값이 아니라 함수가 들어옴. action()처럼 ()를 붙이면 그 함수를 실행

JS에서 함수는 변수에 담거나 인자로 넘길 수 있음 -> *일급 객체*라고 표현

```python
%%writefile /content/crp-demo/server.js

// server.js
const express = require("express");
const app = express();

// Blocking JS: intentionally slow (3 seconds)
app.get("/block.js", async (req, res) => {
  res.setHeader("Content-Type", "application/javascript; charset=utf-8");
  setTimeout(() => {
    res.send(`
      // Simulate heavy/slow script
      alert("Blocking Script Loaded (simulated delay)");
      console.log("block.js executed");
    `);
  }, 3000);
});

// HTML page with blocking script
app.get("/blocking", (req, res) => {
  res.setHeader("Content-Type", "text/html; charset=utf-8");
  res.send(`<!DOCTYPE html>
<html>
<head>
  <meta charset="UTF-8" />
  <title>CRP Demo - Blocking Script</title>
</head>
<body>
  <h2>Blocking Script Demo</h2>

  <p id="before">This text is before the script tag.</p>

  <script src="/block.js"></script>

  <p id="after">This text is after the script tag (DOM continues after script executes).</p>
</body>
</html>`);
});

app.listen(3000, () => console.log("Server running on port 3000"));
```

```python
%%writefile /content/crp-demo/server.js

// server.js
const express = require("express");
const app = express();

// Blocking/Deferred JS: intentionally slow (3 seconds)
app.get("/slow.js", async (req, res) => {
  res.setHeader("Content-Type", "application/javascript; charset=utf-8");
  setTimeout(() => {
    res.send(`
      console.log("slow.js executed");
      alert("slow.js executed after DOM parsing (defer test)");
    `);
  }, 3000);
});

// Blocking page: script without defer
app.get("/blocking", (req, res) => {
  res.setHeader("Content-Type", "text/html; charset=utf-8");
  res.send(`<!DOCTYPE html>
<html>
<head>
  <meta charset="UTF-8" />
  <title>CRP Demo - Blocking Script</title>
</head>
<body>
  <h2>Blocking Script Demo</h2>
  <p id="before">This text is before the script tag.</p>

  <script src="/slow.js"></script>

  <p id="after">This text is after the script tag (may be delayed).</p>
</body>
</html>`);
});

// Defer page: script with defer (non-blocking parsing)
app.get("/defer", (req, res) => {
  res.setHeader("Content-Type", "text/html; charset=utf-8");
  res.send(`<!DOCTYPE html>
<html>
<head>
  <meta charset="UTF-8" />
  <title>CRP Demo - Defer Script</title>
  <script src="/slow.js" defer></script>
</head>
<body>
  <h2>Defer Script Demo</h2>
  <p id="before">This text is before content.</p>
  <p id="after">This text should appear even while slow.js is downloading.</p>

  <script>
    // Observe DOM readiness timing
    document.addEventListener("DOMContentLoaded", () => {
      const el = document.createElement("p");
      el.textContent = "DOMContentLoaded fired (DOM ready)";
      document.body.appendChild(el);
    });
  </script>
</body>
</html>`);
});

app.listen(3000, () => console.log("Server running on port 3000"));
```
