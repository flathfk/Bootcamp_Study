## 6. 함수표현식, 화살표 함수, 콜백함수

### 1). 함수 표현식 (Function Expression)

변수에 담아 사용하는 방식으로, 호출 전 반드시 선언되어야 하는 안정적인 구조입니다.

```python
%%writefile fx1.js

// 1. 기본 인사 출력
const sayHi = function() {
  console.log("1. 안녕하세요!");
};
sayHi();
// 함수를 변수에 담은 것
// sayHi라는 변수가 함수를 가리키고 ()로 실행


// 2. 이름 환영 인사
const greet = function(name) {
  return name + "님, 환영합니다!";
};
console.log("2.", greet("지수"));
// name은 매개변수
// 호출할 때 "지수"를 넣으면 함수 안에서 name = "지수"가 됨


// 3. 숫자 두 배 만들기
const double = function(n) {
  return n * 2;
};
console.log("3. 결과:", double(10));
// n에 10이 들어가서 10 * 2 = 20 반환


// 4. 성인 확인 (20세 기준)
const isAdult = function(age) {
  return age >= 20;
};
console.log("4. 성인인가요?", isAdult(25));
// >= 비교 결과는 true 또는 false
// 25 >= 20이니까 true 반환


// 5. 로그인 상태 메시지
const getLoginMsg = function(id) {
  return id + " 사용자가 로그인했습니다.";
};
console.log("5.", getLoginMsg("gemini123"));
// 문자열 + 문자열 연결
// "gemini123" + " 사용자가 로그인했습니다." 이어붙임


// 6. 두 수의 차이 구하기
const subtract = function(a, b) {
  return a - b;
};
console.log("6. 차이:", subtract(100, 40));
// 매개변수 두 개
// a=100, b=40 순서대로 들어가서 100 - 40 = 60


// 1-6 공통 패턴: 전부 함수 표현식 (const 변수명 = function(){}) 형태
// 함수를 값처럼 변수에 담는 방식
```

```python
!node fx1.js
```

### 2). 화살표 함수 (Arrow Function)

`function` 키워드 없이 `=>`를 사용하며, 한 줄일 때 `return`과 `{}`를 생략한 간결한 형태입니다.

```python
%%writefile fx2.js

// 1. 로그 기록하기
const log = () => console.log("1. 로그를 기록합니다.");
log();
// 매개변수 없을 때 () 반드시 써야 함 → {} 없이 바로 실행


// 2. 간단한 덧셈
const add = (a, b) => a + b;
console.log("2. 합계:", add(10, 20));
// 매개변수 2개 이상 → () 필수
// 한 줄이라 return 생략


// 3. 배열의 첫 번째 요소 가져오기
const getFirst = (arr) => arr[0];
console.log("3. 첫 번째 과일:", getFirst(["사과", "바나나", "포도"]));
// arr[0] → 배열의 0번 인덱스(첫 번째 요소)
// 매개변수 1개 → (arr) 또는 arr 둘 다 가능
// () 생략은 선택사항


// 4. 원의 넓이 계산하기 (반지름 * 반지름 * 3.14)
const getArea = (r) => 3.14 * r * r;
console.log("4. 원의 넓이:", getArea(5));
// 매개변수 1개라 (r) => 대신 r => 로 써도 됨


// 5. 메일 주소 생성기
const makeEmail = (id) => id + "@gmail.com";
console.log("5. 이메일:", makeEmail("js_park"));
// 매개변수 1개라 () 생략 가능


// 6. 홀수 확인하기
const isOdd = (n) => n % 2 !== 0;
console.log("6. 7은 홀수인가요?", isOdd(7));
// % 는 나머지 연산자
// 비교 결과 true/false 가 return 없이 바로 반환됨
```

```python
!node fx2.js
```

### 3). 일급 객체의 핵심: 콜백 함수 (Callback Function)

콜백 함수는 특정 함수에게 "이 작업이 끝나면 내가 준 이 함수를 대신 실행해줘!"라고 부탁하는 방식입니다.

```python
%%writefile fx3.js

// 콜백 함수 = 함수를 인자로 넘겨서, 받은 쪽에서 실행하는 함수
const fn = (callback) => {
  callback(); // 넘겨받은 함수를 여기서 실행
}
fn(() => console.log("@@나는 콜백@@"));


// 1. 기본 실행형 콜백
// 인자로 받은 함수를 내부에서 단순히 실행만 함
const repeat = (callback) => {
  console.log("1. 작업을 시작합니다...");
  callback();
};
repeat(() => console.log("   -> 콜백 함수 실행 완료!"));
// 가장 기본형
// 함수를 받아서 그냥 실행만 함


// 2. 데이터를 전달받는 콜백
// 내부에서 계산된 값을 콜백 함수의 인자로 넣어줍니다.
const getDouble = (num, callback) => {
  const result = num * 2;
  callback(result);
};
getDouble(5, (result) => console.log("2. 계산된 결과값:", result));
// 안에서 계산한 값을 콜백한테 넘겨줌 -> 콜백이 그 값을 받아서 씀


// 3. 배열 필터링 콜백 (실무 다수 사용)
// 조건(함수)을 넘겨주면 그 조건에 맞는 것만 남깁니다.
const filterArr = (arr, condition) => {
  const result = [];
  for (const item of arr) {
    if (condition(item)) result.push(item);
  }
  return result;
};
const numbers = [1, 2, 3, 4, 5];
const odds = filterArr(numbers, (n) => n % 2 !== 0);
console.log("3. 홀수만 필터링:", odds);
// 조건 자체를 함수로 넘기는 패턴
// condition(item) → 넘긴 함수를 item마다 실행해서 true/false 판단


// 4. 시간 지연 콜백 (비동기 입문)
// 정해진 시간이 지난 뒤에 함수를 실행합니다.
const waitAndRun = (seconds, callback) => {
  console.log(`4. ${seconds}초 뒤에 함수가 실행됩니다.`);
  setTimeout(callback, seconds * 1000);
};
waitAndRun(2, () => console.log("   -> 2초가 지났습니다!"));
//setTimeout은 시간이 지난 뒤에 함수를 실행하는 내장 함수
// 지금 당장 실행 안 하고 나중에 실행할 함수를 넘기는 것 → 콜백의 핵심 용도


// 5. 조건에 따른 선택적 콜백
// 성공했을 때와 실패했을 때 서로 다른 함수를 실행하게 합니다.
const checkScore = (score, success, fail) => {
  if (score >= 80) success();
  else fail();
};
checkScore(85,
  () => console.log("5. 합격입니다! 축하드려요."),
  () => console.log("5. 불합격입니다. 다음 기회에!")
);
// 콜백을 2개 넘김 (조건에 따라 다른 함수 실행)
// 성공/실패 처리를 호출하는 쪽에서 직접 정의할 수 있음


// 6. 버튼 클릭 시뮬레이션 콜백
// 어떤 이벤트(클릭 등)가 발생했을 때 실행할 로직을 넘깁니다.
const onClick = (element, action) => {
  console.log(`6. [${element}] 요소가 클릭되었습니다.`);
  action();
};
onClick("로그인 버튼", () => console.log("   -> 메인 페이지로 이동합니다."));
// 이벤트 리스너의 원리
// 실제 브라우저에서 버튼.addEventListener("click", () => {}) <-이게 이 구조
// 어떤 일이 생겼을 때 실행할 함수를 미리 넘겨두는 것
```
