##1. 메모리 구조 (스택 / 힙)
코드가 실행되면 데이터는 두 곳에 저장됨

스택: 단순한 값 저장. 빠름. 크기 고정

힙 : 복잡한 값 저장. 느림. 크기 유동적

````python
%%writefile 1.js

let num = 5           // 스택에 5 저장
let obj = { a: 1 }   // 힙에 저장, 스택엔 주소만
````



---



##2. 원시값 vs 참조값

원시값 (숫자, 문자열, 불린, `null`, `undefined`) → 스택에 값 자체 저장. 복사하면 완전히 독립

참조값 (객체, 배열, 함수) → 힙에 저장, 스택엔 주소만. 복사하면 주소 공유

````python
%%writefile 2.js

let a = 5;  let b = a;  b = 10;
console.log(a) // 5 (독립)

let o1 = { name: "소라" };  let o2 = o1;  o2.name = "철수";
console.log(o1.name) // "철수" (공유)
````



---



##3. 전개 연산자 ...

배열/객체를 펼쳐서 복사하거나 합칠 때 씀

````python
%%writefile 3.js

const a = [1, 2];
const b = [...a, 3]; // [1, 2, 3]

const obj = { name: "소라" };
const obj2 = { ...obj, age: 20 }; // { name: "소라", age: 20 }
````



---



##4. 얕은 복사 vs 깊은 복사

... 전개 연산자는 1단계만 새로 복사

중첩 객체는 주소 공유

````python
%%writefile 4_1.js

const user = { name: "소라", hobby: { main: "코딩" } };
const copy = { ...user };
copy.hobby.main = "게임";
console.log(user.hobby.main) // "게임" (같이 바뀜)
````

중첩까지 완전히 분리하려면 `structuredClone()`

````python
%%writefile 4_2.js

const copy = structuredClone(user);
copy.hobby.main = "게임";
console.log(user.hobby.main) // "코딩" (안 바뀜)
````

##7. 콜백 함수

```
## *9. 크리티컬 렌더링 패스 (Critical Rendering Path, CRP)*

브라우저가 서버로부터 HTML을 받아 화면에 픽셀로 그리기까지의 일련의 단계

단계: DOM → CSSOM → Render Tree → Layout → Paint

```
HTML 파싱 → DOM 생성
CSS 파싱  → CSSOM 생성
             ↓
         둘 합쳐서 Render Tree 생성
             ↓
         Layout (위치/크기 계산)
             ↓
         Paint (화면에 그림)

핵심: `CSS`와 `JS`는 이 과정을 막을 수 있음 (렌더링 블로킹)

해결: `CSS`는 `<head>`에, `JS`는 `<body>` 맨 아래 또는 `defer` 속성으로 씀
