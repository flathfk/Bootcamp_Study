### 콜백 함수를 배우는 이유

**유연성**: 함수 하나(`filterArr`)를 만들어두고, 콜백 함수(`condition`)만 바꾸면 홀수 찾기, 짝수 찾기, 10보다 큰 수 찾기 등 무한한 변신이 가능합니다.

````python
%%writefile ex.js

/**
 * 콜백 함수의 유연성 실습
 * 목적: 메인 로직(순회)은 유지하되, 조건(판단)만 교체하여 함수를 무한히 재활용함
 */

// 1. 메인 엔진 (필터 틀)
// arr: 데이터 배열, condition: 참/거짓을 판단할 콜백 함수
const filterArr = (arr, condition) => {
  // 여기서 condition은 그냥 빈 자리 (아직 어떤 함수인지 모름)
    const result = [];
    for (const item of arr) {
        // 콜백 함수(condition)에 현재 요소를 전달하여 결과가 true일 때만 담음
        if (condition(item)) {
            result.push(item);
        }
    }
    return result;
};

// 2. 테스트 데이터
const numbers = [1, 5, 8, 10, 12, 15, 20, 27];

// 3. 콜백 함수만 바꿔가며 호출하기
console.log("--- 콜백 함수의 유연한 변신 ---");


// (1) 홀수 찾기: n을 2로 나눈 나머지가 있으면 홀수
const odds = filterArr(numbers, n => n % 2 !== 0);
console.log("1. 홀수 리스트:", odds);
// condition 자리에 n => n % 2 !== 0 들어감
// item을 하나씩 꺼내서 condition(item) 호출 → 나머지가 0이 아니면 true → 담음


// (2) 짝수 찾기: n을 2로 나눈 나머지가 없으면 짝수
const evens = filterArr(numbers, n => n % 2 === 0);
console.log("2. 짝수 리스트:", evens);
// (1)이랑 구조 동일. 나머지가 0이면 true → 담음


// (3) 10보다 큰 수 찾기
const biggerThanTen = filterArr(numbers, n => n > 10);
console.log("3. 10보다 큰 수:", biggerThanTen);
// condition(item) 호출 → item이 10보다 크면 true → 담음
// 10은 > 조건에서 제외됨. (>= 아님!!!!)


// (4) 5의 배수 찾기
const multiplesOfFive = filterArr(numbers, n => n % 5 === 0);
console.log("4. 5의 배수:", multiplesOfFive);
// item을 5로 나눈 나머지가 0이면 true → 담음
// 5, 10, 15, 20 통과


// (5) 특정 범위 (10 이상 20 이하) 찾기
const rangeValues = filterArr(numbers, n => n >= 10 && n <= 20);
console.log("5. 10~20 사이 값:", rangeValues);
// 콜백 안에 조건이 두 개. &&라 둘 다 true여야 통과
// condition(item) 호출 → 10 이상이면서 동시에 20 이하인 것만 담음


// (6) 응용: 제곱했을 때 100이 넘는 수 찾기
const powerLarge = filterArr(numbers, n => n * n > 100);
console.log("6. 제곱이 100 초과:", powerLarge);
// condition(item) 호출 시 item을 넘기고, 콜백 안에서 n * n 계산까지 함
// 10 → 10*10=100 → 100 > 100은 false → 버림
// 12 → 12*12=144 → 144 > 100은 true → 담음
// 콜백 안에서 단순 비교뿐 아니라 연산도 가능하다는 것!!
````

비동기 처리: 서버에서 데이터를 받아오는 것처럼 오래 걸리는 작업을 기다리는 동안, 다른 일을 하다가 작업이 끝났을 때만 특정 로직을 실행하기 위해 필수적입니다.

````python
%%writefile ex2.js

/**
 * 자바스크립트 비동기 핵심: "멈추지 않는 엔진"
 * 1. 서버 응답을 기다리는 동안(setTimeout)
 * 2. 메인 스레드는 다음 코드를 즉시 실행하고
 * 3. 작업이 완료되면 '태스크 큐'에 담겨있던 콜백이 실행됨
 */

// [함수 정의] 서버에서 데이터를 가져오는 함수
const getDataFromServer = (id, callback) => {
    console.log(`\n[STEP 1] 서버에 ${id}번 데이터 요청을 보냈습니다.`);
    // getDataFromServer는 서버 요청을 보내고 기다리지 않음!!
    // 3초 후에 할 일(콜백)만 등록해두고 바로 다음 코드로 넘어감

    // 서버 통신 시뮬레이션 (3초 대기)
    setTimeout(() => {
      // 3초 후 여기 실행
        const dummyData = { // → renderPage(dummyData) 호출
            userId: id,
            title: "자바스크립트 비동기 마스터",
            content: "콜백 함수는 비동기 처리의 기초입니다."
        };
        console.log(`\n[STEP 3] 서버 응답이 도착했습니다. (3초 경과)`);

        // 데이터가 준비되었을 때만 콜백 실행
        callback(dummyData);
    }, 3000);
    // 3초가 지난 후에야 callback(dummyData) 호출 = renderPage가 그때 실행됨
    // 데이터(dummyData)가 준비된 시점에 콜백을 실행하는 것
};

// [함수 정의] 데이터를 받아서 화면에 그리는 함수 (콜백용)
const renderPage = (data) => {
    console.log(`[STEP 4] 페이지 렌더링: ${data.title}`);
    console.log(`         내용: ${data.content}`);
    console.log("========================================\n");
};

// --- 실제 실행 흐름 ---

console.log(">>> 프로그램을 시작합니다.");

// *콜백이 파라미터로 넘어가는 부분*
// 비동기 함수 호출
// renderPage 함수 자체를 callback 자리에 넘김. 지금 실행하는 게 아님
getDataFromServer(5, renderPage);

// 중요: 아래 코드는 STEP 1 직후에 바로 실행됩니다. (기다리지 않음)
console.log(">>> [STEP 2] 다른 로직 실행 중...");
console.log("    - 배너 광고 로딩 중...");
console.log("    - 실시간 채팅 연결 중...");

let seconds = 1;
// *setInterval 부분*
const timer = setInterval(() => {
    if (seconds >= 3) {
        clearInterval(timer);
    } else {
        console.log(`    ... 서버 기다리는 중 (${seconds++}초)`);
    }
}, 1000);
// 1초마다 콜백 실행. seconds가 3 이상이면 clearInterval로 타이머 멈춤
// seconds++ : 출력 먼저 하고 그 다음에 1 증가 (1초, 2초만 출력되고 3초에서 종료)

/* JS는 싱글 스레드라 한 번에 한 가지만 처리함
그런데 setTimeout/서버요청 같은 건 브라우저/Node가 대신 처리해줌
그 결과가 준비되면 콜백을 태스크 큐에 넣고, 메인 스레드가 한가해지면 그때 실행함
그래서 기다리는 동안 다른 코드가 먼저 실행되는 것 */
````

````python
!node ex2.js
````

### 콜백이 필요한 상황 2: “리스트를 검사해서 합격자만 골라내기”

### 시나리오

학생 점수 목록에서 “60점 이상”만 합격자로 추린다.

### 문제 모델링

- 입력: 점수 배열
- 처리: 조건(60 이상)을 만족하는지 검사
- 출력: 합격자 배열

````python
%%writefile ex3.js

function filterScores(scores, predicate) {
    // 예외처리: 숫자가 아닌 값이 섞여 있으면 바로 종료
    const hasInvalid = scores.some(s => typeof s !== "number");
    if (hasInvalid) {
        console.log("숫자가 아닌 값이 포함되어 있습니다.");
        return [];
    }

    const passed = [];
    for (let i = 0; i < scores.length; i++) {
        if (predicate(scores[i])) passed.push(scores[i]);
    }
    return passed;
}

const scores = [45, 72, "88", 59, 60];

const pass = filterScores(scores, (score) => score >= 60);
console.log(pass);
````

### 시나리오

장바구니 상품 가격 배열이 있다.

“조건에 맞는 상품만 골라서(콜백), 할인 적용(콜백) 후 최종 합계”를 계산하라.

### 모델링

- 입력: 가격 배열

- 처리1: 필터(조건 콜백)

- 처리2: 변환(할인 콜백)

- 출력: 합계

````python
%%writefile ex4.js

// 전체 구조:
// 함수가 순서대로 데이터를 가공하는 파이프라인
// prices → filter → filtered → map → discounted → sum → total
function filter(arr, predicate) {
  const out = [];
  for (let i= 0; i < arr.length; i++) {
    if (predicate(arr[i])) out.push(arr[i]);
  }
  return out;
}

function map(arr, transform) {
  const out = [];
  for (let i = 0; i < arr.length; i++) {
    out.push(transform(arr[i]));
  }
  return out;
}

function sum(arr) {
  let total = 0;
  for (let i = 0; i < arr.length; i++) total += arr[i];
  return total;
}

// 데이터
const prices = [12000, 8000, 30000, 15000, 4000];

// 1) 10000원 이상만 구매(조건 콜백)
const filtered = filter(prices, (p) => p >= 10000);
// filter : 조건 통과한 것만 남김

// 2) 10% 할인 적용(변환 콜백)
const discounted = map(filtered, (p) => Math.floor(p * 0.9));
// map : 배열 요소를 전부 변환해서 새 배열 반환
// filter랑 구조 동일한데 transform(arr[i]) 결과를 무조건 다 담는지 차이

// 3) 합계
const total = sum(discounted);
// sum : 배열 전부 더하기
// total은 재할당이 계속 일어나서 let (const 쓰면 에러)

console.log("원가:", prices);
console.log("필터 후:", filtered);
console.log("할인 후:", discounted);
console.log("최종 합계:", total);

// 콜백 관점
/* filter는 true/false 반환하는 함수를 받고,
map은 값을 변환해서 반환하는 함수를 받음
둘 다 "뭘 할지"를 콜백으로 위임하는 구조는 동일 (목적만 다름) */
````

````python
!node ex4.js
````

## 초급 실습

### 시나리오

피자를 주문하면 → 조리 완료 후 배달 시작 → 배달 완료 후 알림 출력

````python
%%writefile ex5.js

// 1) 조리: 끝나면 onCookComplete 실행
function cook(foodName, onCookComplete) {
  console.log(foodName + " 조리 시작...");

  setTimeout(() => {
    console.log(foodName + " 조리 완료!");
    onCookComplete();
  }, 1000);
}

// 2) 배달: 끝나면 onDeliveryComplete 실행
function deliver(onDeliveryComplete) {
  console.log("배달 출발...");

  setTimeout(() => {
    console.log("배달 완료!");
    onDeliveryComplete();
  }, 1000);
}

// 3) 실행(콜백으로 순서 연결)
cook("피자", () => {
  deliver(() => {
    console.log("고객에게 전달 완료!");
  });
});
````

## 중급 실습

 콜백 + 화살표 + “규칙(조건) 분리” + 단계 파이프라인

### 시나리오

주문 생성 → 결제 → 조리 → 배달 → 알림

(성공/실패 분기 포함)

````python
%%writefile ex6.js

// [데이터] 메뉴판
const MENU = {
  pizza: 15000,
  chicken: 18000,
  burger: 9000,
};

// [유틸] 지연 함수
function delay(ms, onDone) {
  setTimeout(onDone, ms);
}

// 1) 주문 생성
function createOrder(itemName, quantity, onSuccess, onFail) {
  delay(300, () => {
    const unitPrice = MENU[itemName];

    if (!unitPrice) return onFail("메뉴 없음: " + itemName);
    if (quantity <= 0) return onFail("수량 오류: " + quantity);

    const order = {
      itemName,
      quantity,
      unitPrice,
      totalPrice: unitPrice * quantity,
      createdAt: new Date().toLocaleTimeString(),
    };

    onSuccess(order);
  });
}

// 2) 결제
function pay(order, onSuccess, onFail) {
  delay(400, () => {
    const paymentFailed = Math.random() < 0.2; // 20% 실패(실습용)

    if (order.totalPrice <= 0) return onFail("결제 금액 오류");
    if (paymentFailed) return onFail("결제 실패: 잔액 부족(가정)");

    const paidOrder = {
      ...order,
      paid: true,
      paidAt: new Date().toLocaleTimeString(),
    };

    onSuccess(paidOrder);
  });
}

// 3) 조리
function cook(order, onDone) {
  delay(500, () => {
    const cookedOrder = {
      ...order,
      cooked: true,
      cookedAt: new Date().toLocaleTimeString(),
    };
    onDone(cookedOrder);
  });
}

// 4) 배달
function deliver(order, onDone) {
  delay(500, () => {
    const deliveredOrder = {
      ...order,
      delivered: true,
      deliveredAt: new Date().toLocaleTimeString(),
    };
    onDone(deliveredOrder);
  });
}

// 5) 알림(알림 방식은 콜백으로 교체 가능)
function notify(order, notifier) {
  notifier(order);
}

// [실행] 파이프라인
console.log("=== 배달 시뮬레이터 시작 ===");

createOrder(
  "pizza",
  2,

  // 주문 성공
  (order) => {
    console.log("[주문 완료]", order);

    pay(
      order,

      // 결제 성공
      (paidOrder) => {
        console.log("[결제 완료]", paidOrder.totalPrice + "원");

        cook(paidOrder, (cookedOrder) => {
          console.log("[조리 완료]", cookedOrder.itemName);

          deliver(cookedOrder, (deliveredOrder) => {
            console.log("[배달 완료]", deliveredOrder.deliveredAt);

            notify(deliveredOrder, (completedOrder) => {
              console.log(
                "알림: " +
                  completedOrder.itemName +
                  " " +
                  completedOrder.quantity +
                  "개 배달 완료! (총 " +
                  completedOrder.totalPrice +
                  "원)"
              );
            });
          });
        });
      },

      // 결제 실패
      (reason) => {
        console.log("[종료] 결제 실패:", reason);
      }
    );
  },

  // 주문 실패
  (reason) => {
    console.log("[종료] 주문 실패:", reason);
  }
);
````

````python
!node ex6.js
````

# 시험 이론 요약 정리

## 1세트 · 자료구조 / OS

**Stack vs Queue**  
Stack은 LIFO(마지막에 넣은 게 먼저 나옴), Queue는 FIFO(먼저 넣은 게 먼저 나옴).  
Stack → 괄호 검사·함수 호출 / Queue → CPU 스케줄링

**단일 연결 리스트**  
노드 = 데이터 + 다음 노드 포인터. 메모리 불연속, 임의 접근 불가 O(n), 중간 삽입은 배열보다 빠름.

**배열 vs 연결 리스트**  
- 배열: 메모리 연속 / 임의 접근 O(1) / 삽입·삭제 느림  
- 연결 리스트: 메모리 불연속 / 임의 접근 O(n) / 삽입·삭제 빠름

**이진 탐색**  
정렬된 배열에서 절반씩 줄여가며 탐색. O(log n). 선형·순차 탐색은 O(n).

**트리 vs 그래프**  
트리 → 루트 1개·계층구조·사이클 없음  
그래프 → 사이클 가능·방향 무방향 모두 가능. 트리는 그래프의 특수한 형태.

**CPU 구성**  
ALU(연산) + 제어장치(명령 해석) + 레지스터(임시 저장). 가상메모리는 CPU 구성요소 아님.

**스케줄러**  
여러 프로세스 중 어떤 걸 CPU에 올릴지 순서·우선순위 결정.

## 2세트 · Linux / Shell

**Ubuntu / LTS**  
Ubuntu = Debian 기반 오픈소스 리눅스. LTS = Long Term Support (장기 지원 버전).

**디렉토리 역할**  
- `/etc` : 시스템 설정 파일  
- `/var/log` : 시스템 로그  
- `/bin`, `/usr/bin` : 실행 파일  
- `~` : 사용자 홈

**Shell**  
사용자 명령을 해석해서 커널에 전달하는 인터페이스. 기본 표준 쉘은 bash.  
핵심 기능 3가지: 명령 해석·실행 / 환경변수 관리 / 스크립트 자동화

**REPL 4단계**  
Read(입력) → Eval(해석·실행) → Print(출력) → Loop(반복)

**자주 쓰는 명령어**  
- `ls -la` : 숨김파일 포함 상세 출력  
- `cd ..` : 상위 디렉토리 이동  
- `rm -r folder` : 디렉토리 재귀 삭제  
- `chmod +x file` : 실행 권한 추가  
- `chmod 755` : 소유자 rwx / 그룹 r-x / 기타 r-x  
- `ps aux` : 전체 프로세스 확인  
- `df -h` : 디스크 사용량 (사람이 읽기 쉬운 단위)  
- `grep -i` : 대소문자 무시 검색

**파이프(|)**  
앞 명령의 출력을 뒤 명령의 입력으로 전달.  
`ps aux | grep python` → 실행 중인 프로세스 중 python 관련만 필터링.

**리다이렉션**  
`>` : 파일에 덮어쓰기 / `>>` : 파일에 이어쓰기

**sudo**  
일반 사용자가 관리자(root) 권한으로 명령 실행.

## 3세트 · HTTP / 브라우저 / 서버

**브라우저 URL 입력 후 흐름**  
브라우저 → DNS 조회 → TCP 연결 → HTTP 요청 생성 → Cloudflare(Reverse Proxy) → Node 서버 → 응답 반환 → 브라우저 렌더링

**HTTP 요청 메시지 구조**  
- Request Line: Method + URI + HTTP 버전  
- Headers: 메타정보 (Content-Type, Cookie 등)  
- Body: POST 요청 시 클라이언트 데이터 담김

**HTTP 응답 구조**  
Status Line → Headers → Blank Line → Body  
200 = 정상 / 404 = 자원 없음 / 500 = 서버 오류

**Express 처리 흐름 (POST)**  
요청 도착 → 미들웨어(urlencoded로 body 파싱) → 라우팅(Method+URI 매칭) → 응답 생성

**Web Server vs WAS**  
- Web Server: 정적 파일 제공 (HTML, CSS, 이미지)  
- WAS: 비즈니스 로직 처리, 동적 응답 생성 (Node/Express)

**Cloudflare**  
Reverse Proxy + TLS(HTTPS) 처리. 공인 IP 없이 HTTPS 제공(Tunnel). 서버 IP를 직접 노출하지 않아 보안상 이점.

**HTTP Stateless**  
요청 간 상태를 서버가 기억하지 않음. 매 요청이 독립적. → 로그인 유지 등을 위해 Cookie·Session·JWT 같은 별도 상태 관리 필요.

**nohup 명령어에서 &**  
백그라운드 실행. 터미널 종료해도 서버 계속 동작.

## 4세트 · 모던 바닐라 JS

**바닐라 JS / 모던 JS**  
바닐라 = 프레임워크·라이브러리 없이 순수 JS만 사용.  
모던 = ES6+ 이후 문법(let/const/화살표함수/구조분해 등) 사용.

**프레임워크보다 바닐라 JS를 먼저 배우는 이유**  
- JS 동작 원리를 정확히 이해해야 프레임워크 내부 구조를 파악할 수 있음  
- 프레임워크는 추상화된 것이라 기초 없이 쓰면 에러 원인 파악 불가

**var 문제점**  
- 재선언 허용 → 같은 이름 변수를 실수로 덮어써도 에러 안 남  
- 함수 레벨 스코프 → if/for 블록 밖에서도 접근 가능 → 격리 안 됨  
- let/const는 블록 레벨 스코프 → 블록 밖 접근 시 에러

**호이스팅 / TDZ**  
- var: 선언+undefined 초기화가 함께 끌어올려짐 → 선언 전 접근해도 undefined 출력  
- let/const: 선언만 끌어올려지고 초기화는 코드 도달 시점에 됨 → 그 사이 구간이 TDZ → 접근 시 ReferenceError  
- TDZ 존재 이유: 선언 전 접근이라는 실수를 에러로 잡아주기 위함

**const 객체**  
const는 변수가 가리키는 메모리 주소(참조)를 고정. 객체 내부 값 변경은 주소가 안 바뀌므로 가능. 변수 자체에 새 객체 재할당은 주소가 바뀌므로 불가.

**let 사용 기준**  
재할당이 필요한 경우에만 사용. 대표 사례: 반복문 카운터(`i`), 누적 합산 변수(`total`)

**=== vs ==**  
`==`는 타입 자동 변환 후 비교 → `"10" == 10`이 true.  
`===`는 타입+값 모두 일치해야 true → `"10" === 10`은 false. 예상치 못한 형변환 버그 방지.

**삼항 연산자**  
`조건 ? 참일 때 값 : 거짓일 때 값`  
단순한 if-else를 한 줄로 표현할 때 편리. 변수 할당 시 주로 씀.

## 5세트 · 브라우저 렌더링 / HTML / JS 메모리

**Critical Rendering Path**  
HTML 파싱 → DOM 생성 / CSS 파싱 → CSSOM 생성 → Render Tree(DOM+CSSOM 결합, display:none 제외) → Layout(위치·크기 계산) → Paint(색상·텍스트 그리기) → Composite(레이어 합성)

**script 태그**  
일반 script: HTML 파싱 중단 → JS 다운로드·실행 → 파싱 재개  
defer: 파싱과 병렬 다운로드 → 파싱 완료 후 실행. DOM 완성 보장.

**CSS Render Blocking**  
CSSOM 완성 전에 화면 출력하면 스타일 없는 화면이 잠깐 보이는 FOUC 발생 → 브라우저가 CSS 완성될 때까지 렌더링 차단.

**시맨틱 마크업**  
태그에 의미를 담아 구조를 전달. 접근성·SEO·유지보수에 유리.  
- `<main>` : 문서당 하나, 핵심 콘텐츠  
- `<article>` : 독립 배포 가능한 콘텐츠 (뉴스 기사 등)  
- `<section>` : 주제별 묶음, 독립성 없음

**웹 접근성**  
스크린 리더는 시맨틱 태그 기반으로 해석. 이미지엔 alt 속성, 제목은 h1→h2→h3 순서 지켜야 함.

**원시값 vs 참조값**  
- 원시값(Number, String, Boolean, null, undefined): Stack에 값 직접 저장, 복사 시 값 자체 복사 → 독립적  
- 참조값(Object, Array, Function): Heap에 데이터 저장, Stack엔 주소만 저장 → 복사 시 주소 공유 → 원본 영향 받음

**Stack vs Heap**  
Stack: 함수 호출·원시값 저장, 빠름, 자동 해제  
Heap: 객체·배열 저장, 크기 유동적, GC가 해제

**얕은 복사**  
1단계 값만 복사. 중첩 객체는 주소를 공유 → 내부 객체 수정 시 원본도 바뀜.  
전개 연산자(`...`)로 복사해도 얕은 복사.

**함수 선언문 vs 표현식**  
선언문: 호이스팅 완전히 됨 → 선언 전 호출 가능  
표현식: 변수에 담긴 함수 → 호이스팅 안 됨 → 선언 전 호출 시 에러

**화살표 함수**  
자체 this 없음. 선언된 위치의 this를 그대로 사용(렉시컬 this).

// 16. multiply 함수 표현식
const multiply = (a, b) => a * b;

// 17. 짝수 필터
const getEvens = (arr) => {
  const result = [];
  for (let i = 0; i < arr.length; i++) {
    if (arr[i] % 2 === 0) result.push(arr[i]);
  }
  return result;
};

// 18. 합산
const sumAll = (arr) => {
  let total = 0;
  for (let i = 0; i < arr.length; i++) total += arr[i];
  return total;
};

// 19. 얕은 복사 - 중첩 객체 원본 변경
const original = { name: "소라", address: { city: "서울" } };
const copy = { ...original };
copy.address.city = "부산"; // 원본도 바뀜
console.log(original.address.city); // "부산"

// 20. 원시값 vs 객체 복사 비교
let a = 10;
let b = a;
b = 99;
console.log(a); // 10 (독립적)

const obj1 = { name: "소라" };
const obj2 = obj1;
obj2.name = "영희";
console.log(obj1.name); // "영희" (공유됨)
