### 4. 실전! 메서드 체이닝 (Method Chaining)

비변경 메서드들은 새로운 배열을 반환하기 때문에, 여러 메서드를 사슬처럼 연결하여 가독성 높은 코드를 작성할 수 있습니다.

> **예제**: 가격이 1000원 이상인 상품의 이름을 대문자로 바꾸기
>

```python
%%writefile ex.js

// 체이닝(Chaining):
// 배열 메서드를 .으로 연결해서 순서대로 처리하는 패턴
const items = [
  { name: 'apple', price: 1200 },
  { name: 'pear', price: 800 },
  { name: 'orange', price: 1500 }
];

const result = items
  .filter(item => item.price >= 1000)
  // 1. 필터링:
  // 배열을 순회하면서 조건이 true인 것만 남긴 새 배열 반환
  .map(item => item.name.toUpperCase());
  // 2. 이름 변환:
  // 배열의 각 요소를 변환한 새 배열 반환 (여기선 name 속성만 꺼내서 대문자로)

  // filter와 map 둘 다 원본 배열을 건드리지 않고 새 배열을 반환하기 때문에 체이닝이 가능
  // (앞 메서드가 반환한 배열에 바로 다음 메서드를 호출하는 구조)

console.log(result); // ['APPLE', 'ORANGE']

/* 아래 두 코드는 완전히 동일
const r1 = items.filter(...).map(...);

const filtered = items.filter(...);
const r2 = filtered.map(...); */
```

```python
!node ex.js
```

**실습 1. `forEach` - 출석부 명단 출력하기**

- **조건**: 새로운 배열을 만들지 말고 순회만 하세요.

```python
%%writefile ex2.js

// forEach: 배열의 요소를 하나씩 꺼내서 반복 실행하는 메서드 (map이나 filter와 달리 반환값이 없음(undefined))

/* forEach로 출석부 명단 출력하기 : 기능(동작) 설명

1. students 배열에는 민수, 영희, 지혁, 혜원 네 명의 이름이 담겨 있음
2. .forEach를 사용해 배열을 처음부터 끝까지 순서대로 순회
   -> 매 순회마다 꺼낸 이름 뒤에 "님 환영합니다!"를 붙여 콘솔에 출력
   -> 새로운 배열을 만들지 않고 출력만 수행
3. 최종적으로 콘솔에 "민수님 환영합니다!", "영희님 환영합니다!", "지혁님 환영합니다!", "혜원님 환영합니다!" 가 순서대로 출력 */

const students = ['민수', '영희', '지혁', '혜원'];

students.forEach(name => {
  console.log(`${name}님 환영합니다!`);
});
```

**실습 2. `map` - 상품 가격 10% 인상하기**

물가가 올라 모든 상품의 가격을 10%씩 인상한 **새로운 객체 배열**을 만드세요.

- **조건**: 원본 `products` 배열은 수정되지 않아야 합니다.

```python
%%writefile ex3.js

/* map으로 상품 가격 10% 인상하기 : 기능(동작) 설명

1. products 배열에는 커피, 라떼, 케이크 세 상품의 이름과 가격이 객체 형태로 담겨 있음
2. .map을 사용해 배열을 순회하며 각 객체를 새로운 객체로 변환
   -> 이때 ...item으로 기존 객체의 모든 속성을 그대로 복사한 뒤 price만 item.price * 1.1로 덮어씀
   -> 원본 객체를 건드리지 않고 새 객체를 만드는 방식이라 products 원본은 변하지 않음
3. 최종적으로 raisedProducts에는 가격이 10% 오른 새 객체 배열이 담기고 products[0].price는 여전히 4000원임 */

const products = [
  { name: '커피', price: 4000},
  { name: '라떼', price: 4500 },
  { name: '케이크', price: 6000 }
];

const raisedProducts = products.map(item => ({
  ...item,
  price: item.price * 1.1 // 가격 10% 인상
}));

console.log(raisedProducts);
console.log(products[0].price);
```

```python
!node ex3.js
```

**실습 3. `filter` - 재고가 있는 상품만 골라내기**

재고(`inStock`)가 `true`인 상품만 필터링하여 따로 관리하세요.

- **조건**: `filter` 메서드를 사용하세요.

```python
%%writefile ex4.js

/* filter로 재고 있는 상품만 골라내기 : 기능(동작) 설명

1. inventory 배열에는 아이폰, 맥북, 아이패드, 에어팟 네 상품의 id, 이름, 재고 여부가 객체 형태로 담겨 있음
2. .filter를 사용해 배열을 순회하며 item.inStock이 true인 것만 남긴 새 배열을 반환
   -> false인 맥북과 에어팟은 탈락
3. 최종적으로 availableItems에는 아이폰과 아이패드 객체만 담긴 배열이 저장 */

const inventory = [
  { id: 1, name: '아이폰', inStock: true },
  { id: 2, name: '맥북', inStock: false },
  { id: 3, name: '아이패드', inStock: true },
  { id: 4, name: '에어팟', inStock: false }
];

const availableItems = inventory.filter(item => item.inStock); // 재고가 있는 상품만 추출

console.log(availableItems);
```

**실습 4. `reduce` - 장바구니 총 결제 금액 계산**

장바구니에 담긴 모든 상품의 가격을 합산하여 최종 금액을 구하세요.

- **조건**: `reduce`의 초기값(Initial Value)을 `0`으로 설정하세요.

```python
%%writefile ex5.js

const cart = [
  { name: '셔츠', price: 35000 },
  { name: '바지', price: 48000 },
  { name: '양말', price: 5000 }
];

// TODO: reduce를 사용하여 총 합계를 구하세요.
const totalPrice = cart.reduce((acc, cur) => acc + cur.price, 0);


/* reduce로 장바구니 총 결제 금액 계산하기 : 기능(동작) 설명

1. cart 배열에는 셔츠, 바지, 양말 세 상품의 이름과 가격이 객체 형태로 담겨 있음
2. .reduce를 사용해 배열을 순회하며 모든 상품의 가격을 하나의 값으로 축약함
   -> acc는 이전 단계까지의 누적 합계가 저장되는 바구니, cur은 현재 순회 중인 상품 객체
   -> 매 순회마다 acc + cur.price를 계산해 acc에 누적
   -> 초기값 0으로 시작하므로 첫 순회에서 acc는 0
3. 최종적으로 totalPrice에 88000이 저장되고 "총 결제 금액: 88000원"이 출력 */


/* reduce는 값의 축약을 의미
const totalPrice = cart.reduce((acc, cur) => acc + cur.price, 0); */
//                             ───  ───      ─────────────    ─
//                              ▲    ▲             ▲          ▲
//                              │    │             │          │
// [1] acc (Accumulator) ───────┘    │             │          │
//     : '누적 합계'가 저장되는 바구니 (이전 단계의 결과값)   │          │
//                                   │             │          │
// [2] cur (Current Value) ──────────┘             │          │
//     : '현재 요소', 즉 배열에서 지금 막 꺼낸 상품 객체     │          │
//                                                 │          │
// [3] 실행부 (acc + cur.price)   ───────────────────┘          │
//     : 지금까지의 합계(acc)에 현재 상품 가격(price)을 더함            │
//                                                            │
// [4] 0 (Initial Value)   ───────────────────────────────────┘
//     : '시작 숫자', 계산기를 0으로 리셋하고 시작한다는 의미


console.log(`총 결제 금액: ${totalPrice}원`);
```

**실습 5. 종합(Chaining) - 고득점자 명단 가공하기**

점수가 80점 이상인 학생들만 뽑아서, 이름만 담긴 배열을 만들고, 가나다순으로 정렬하세요.

- **조건**: `filter`, `map`, `sort`를 체이닝으로 연결하세요.

```python
%%writefile ex6.js

/* filter+map+sort 체이닝으로 고득점자 명단 가공하기 : 기능(동작) 설명

1. scores 배열에는 5명의 학생 이름과 점수가 객체 형태로 담겨 있음
2. 세 단계의 체이닝으로 처리함
   - 1단계: .filter로 s.score >= 80인 학생만 남긴 새 배열을 반환 (70점 김철수와 60점 최유리는 탈락)
   - 2단계: .map으로 앞서 걸러진 객체 배열에서 s.name만 꺼내 문자열 배열로 변환
   - 3단계: .sort()로 문자열 배열을 가나다순으로 정렬
3. 최종적으로 honorsList에 ['박지성', '이영희', '정우성']이 저장됨 */

const scores = [
  { name: '김철수', score: 70 },
  { name: '이영희', score: 95 },
  { name: '박지성', score: 88 },
  { name: '최유리', score: 60 },
  { name: '정우성', score: 92 }
];

const honorsList = scores
  .filter(s => s.score >= 80) // 1. 80점 이상인 학생만 남긴 새 배열 반환
  .map(s => s.name) // 2. 이름을 문자열 배열로 반환
  .sort(); // 3. 가나다순으로 저장

  console.log(honorsList);
```

```python
!node ex6.js
```
