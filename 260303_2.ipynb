# CSS3 기초 II: Flex, Container, Wrap

### 1. Flex Layout (플렉스 레이아웃)의 정의

Flexbox(**Flexible Box**)는 1차원(행 또는 열)을 기준으로 요소를 배치하는 레이아웃 모델입니다. 복잡한 계산 없이 요소의 크기와 순서를 컨테이너 내에서 유연하게 **관리**하기 위해 설계되었습니다.

Flex는 화면의 크기가 변경될 때 화면에 따라 배치되도록 하기 위해 사용하며, **부모**(**Container**)에서 flex를 정의하고 **자식**(**Item**)에서 크기와 기준을 정의하여 사용합니다.

### 2. 주요 구성 요소: Container와 Item

- **Flex Container (플렉스 컨테이너):** `display: flex;`가 선언된 부모 요소입니다. 모든 플렉스 아이템을 감싸는 **최상위 박스**로, 내부 아이템들의 정렬 방향(Direction), 줄 바꿈(Wrap), 간격(Gap)을 총괄 제어하는 **지휘소** 역할을 합니다.
- **Flex Item (플렉스 아이템):** 컨테이너 직계 자식들로, 플렉스 규칙에 따라 배치되는 대상입니다. 박스모델을 기반으로 독립적인 크기를 가지며 컨테이너 내에서 유연하게 반응합니다.

### 3. Flex-Wrap (줄 바꿈 제어)의 메커니즘

`flex-wrap` 속성은 플렉스 아이템들이 컨테이너의 너비를 초과했을 때, 한 줄에 강제로 배치할지 아니면 다음 줄로 넘길지를 결정합니다.

**속성값 분석:**

- **nowrap (기본값):** 모든 아이템을 한 줄에 강제로 배치합니다. 아이템의 너비 합이 부모보다 크면 아이템이 압축되거나 범위를 벗어납니다(Overflow).
- **wrap (줄 바꿈):** 아이템의 너비 합이 컨테이너 너비를 초과하면, 초과한 아이템을 다음 행(Next Line)으로 배치합니다.
- **wrap-reverse:** 아이템을 역순으로 줄 바꿈하여 배치합니다.

### 4. 실무형 레이아웃 구성 원리

**다단(Multi-column) 레이아웃 공식:**

한 줄에 특정 개수의 박스를 배치하고 싶을 때, 컨테이너와 아이템은 다음 속성을 상호작용시킵니다.

- **부모(Container):** `display: flex;`와 `flex-wrap: wrap;`을 동시에 선언하여 행 흐름을 활성화합니다.
- **자식(Items):** `width` 또는 `flex-basis`를 통해 점유율을 설정합니다. (예: 2단 구성 시 `width: 50%;`)

##요약 및 결론
**Container**: 레이아웃 문맥 생성 (아이템들의 전체적인 배치 규칙과 간격을 결정)

**Flex**: 정렬 및 공간 배분 (요소 간의 가변적인 관계 정의)

**Wrap**: 흐름 제어(Flow Control)	(컨테이너 경계에 따른 적응형 배치 구현)

*팁: 실무에서는 카드 레이아웃을 만들 때 Container에 `gap` 속성을 추가하여 아이템 간의 간격을 박스모델 계산 없이 편리하게 관리하기도 합니다.*

```python
%%writefile ex.html

<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <title>박스모델과 플렉스 실습</title>
    <style>
        /* 1. 박스모델 초기화 */
        * {
            box-sizing: border-box;
            margin: 0;
            padding: 0;
        }

        /* 2. 최상위 부모(body) 설정 */
        body {
            display: flex;          /* 자식인 container를 플렉스로 관리 */
            justify-content: center; /* 가로 중앙 정렬 */
            align-items: center;     /* 세로 중앙 정렬 */
            min-height: 100vh;      /* 화면 전체 높이 확보 */
            background-color: #e9ecef;
        }

        /* 3. 플렉스 컨테이너 (부모) */
        .container {
            display: flex;          /* 자식 아이템들을 가로로 나열 */
            flex-wrap: wrap;        /* 자리가 모자라면 줄 바꿈(Wrap) */
            width: 600px;           /* 전체 컨테이너 너비 고정 */
            gap: 20px;              /* 박스 사이의 물리적 간격 */
            padding: 20px;
            background-color: white;
            border-radius: 10px;
            box-shadow: 0 4px 6px rgba(0,0,0,0.1);
        }

        /* 4. 플렉스 아이템 (자식 4개) */
        .item {
            /* 한 줄에 2개씩 배치 (50%에서 간격의 절반인 10px 제외) */
            width: calc(50% - 10px);
            height: 150px;

            /* 박스모델 상세 정의 */
            background-color: #3498db;
            border: 4px solid #2980b9; /* 테두리 */
            padding: 20px;             /* 안쪽 여백 */
            color: white;

            /* 내용물 중앙 정렬 */
            display: flex;
            justify-content: center;
            align-items: center;
            font-size: 1.2rem;
            font-weight: bold;
        }
    </style>
</head>
<body>

    <div class="container">
        <div class="item">BOX 1</div>
        <div class="item">BOX 2</div>
        <div class="item">BOX 3</div>
        <div class="item">BOX 4</div>
    </div>

</body>
</html>
```

### Flexbox를 활용한 양방향 정렬 헤더 레이아웃

### 1. 주축(Main Axis) 정렬을 통한 배치 분리

부모 요소인 `.header-container`에 `justify-content: space-between;`을 선언하여, 좌측에는 **로고**, 우측에는 **내비게이션 메뉴**가 위치하도록 양 끝으로 밀어냈습니다. 이는 별도의 여백(Margin) 계산 없이 요소 간의 간격을 최대화하는 가장 효율적인 방식입니다.

### 2. 교차축(Cross Axis) 정렬을 통한 수직 중앙 배치

`align-items: center;` 속성을 사용하여 높이 `80px`인 헤더 안에서 로고와 메뉴 텍스트들이 세로 방향으로 정확히 가운데에 오도록 정렬했습니다. 박스모델의 `padding-top` 등을 조절하지 않아도 시각적 균형을 맞춰줍니다.

### 3. 중첩 플렉스(Nested Flex) 구조의 활용

- **1차 플렉스:** `.header-container`가 로고와 내비게이션 덩어리를 좌우로 배치합니다.
- **2차 플렉스:** `.nav-menu`(`ul`)에 다시 `display: flex;`를 주어, 세로로 나열되던 리스트 항목(`li`)들을 가로로 나란히 세웠습니다. `gap: 30px;`을 통해 메뉴 간의 간격도 규격화했습니다.

```python
%%writefile ex2.html

<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <title>Flex Header 실습</title>
    <style>
        /* 1. 박스모델 및 기본 초기화 */
        * {
            box-sizing: border-box;
            //실제 박스 크기는 딱 200px로 유지됩니다. (내부 콘텐츠 영역이 알아서 140px로 줄어듭니다.)
            margin: 0;
            padding: 0;
        }

        /* 2. 플렉스 컨테이너 (Header) */
        .header-container {
            display: flex;            /* 플렉스 레이아웃 선언 */
            justify-content: space-between; /* 자식 요소들을 양 끝으로 배치 */
            align-items: center;      /* 세로축 중앙 정렬 */
            padding: 0 40px;          /* 좌우 내 여백 */
            height: 80px;             /* 헤더 높이 */
            background-color: #ffffff;
            box-shadow: 0 2px 5px rgba(0,0,0,0.1);
        }

        /* 3. 로고 스타일 */

		.logo img {
		    height: 30px;          /* 헤더 높이(80px)보다 작게 설정하여 여유 공간 확보 */
		    width: auto;           /* 가로 세로 비율 유지 */
		    display: block;        /* 하단 미세 공백 제거 */
		}


        /* 4. 내비게이션 (메뉴 항목들도 Flex로 정렬) */
        .nav-menu {
            display: flex;            /* 메뉴 아이템들도 가로로 나열 */
            gap: 30px;                /* 메뉴 사이 간격 */
            list-style: none;         /* 불렛 제거 */
        }

        .nav-menu li a {
            text-decoration: none;
            color: #333;
            font-weight: 500;
        }

        .nav-menu li a:hover {
            color: #3498db;
        }
    </style>
</head>
<body>

    <header class="header-container">

		<div class="logo">
	        <a href="#">
				<img src="logo.png" alt="서비스 로고">
		    </a>
		</div>


        <nav>
            <ul class="nav-menu">
                <li><a href="#">Home</a></li>
                <li><a href="#">About</a></li>
                <li><a href="#">Services</a></li>
                <li><a href="#">Contact</a></li>
            </ul>
        </nav>
    </header>

</body>
</html>
```

## **CSS3 기초 II: 포지셔닝(Relative/Absolute)과 레이아웃 기초**

웹 페이지의 요소들을 배치하는 것은 마치 빈 방에 가구를 배치하는 것과 같습니다. 어떤 가구는 벽에 붙이고, 어떤 소품은 책상 위에 겹쳐 올려두기도 하죠. CSS의 **position** 속성은 바로 이러한 요소들의 배치 전략과 기준점을 결정하는 핵심 도구입니다.

---

### **1. 레이아웃의 흐름과 포지셔닝 기초**

### **1.1 문서의 일반적인 흐름 (Normal Flow)**

모든 HTML 요소는 별도의 설정을 하지 않으면 일반적인 흐름에 따라 배치됩니다. 이는 우리가 문서를 작성할 때 위에서 아래로, 왼쪽에서 오른쪽으로 채워나가는 방식과 같습니다.

- **블록 요소(Block):** `<div>`, `<h1>`, `<p>` 등은 새로운 줄에서 시작하며, 부모의 전체 너비를 차지하며 수직으로 쌓입니다.
- **인라인 요소(Inline):** `<span>`, `<a>` 등은 줄 바꿈 없이 콘텐츠 크기만큼만 공간을 차지하며 수평으로 나열됩니다.
- **특징:** 요소들이 서로의 공간을 존중하며, 차례대로 밀려나거나 쌓이는 가장 기본적인 상태입니다.

### **1.2 position 속성 개요**

**position** 속성은 요소가 위치를 결정할 때 누구를 기준점으로 삼을지를 정의합니다. 5가지 주요 값의 특징을 이해하는 것이 중요합니다.

- **static (기본값):** 일반적인 흐름을 따릅니다. 좌표 속성이 적용되지 않는 상태입니다.
- **relative (상대 위치):** 요소가 원래 있어야 할 자리를 기준으로 이동합니다. 원래 차지하던 공간은 그대로 유지됩니다.
- **absolute (절대 위치):** 일반 흐름에서 완전히 벗어나, 포지셔닝된(static이 아닌) 가장 가까운 조상 요소를 기준으로 배치됩니다.
- **fixed (고정 위치):** 브라우저 화면(Viewport)을 기준으로 고정되어, 스크롤을 내려도 항상 같은 위치에 나타납니다.
- **sticky (끈적이는 위치):** 처음에는 일반 흐름을 따르다가, 스크롤이 특정 지점에 도달하면 화면 상단 등에 붙습니다.

### **1.3 좌표 속성의 활용: top, right, bottom, left와 방향성**

**position**이 **static**이 아닌 요소들은 좌표 속성을 통해 구체적인 이동 거리를 설정할 수 있습니다.

- **방향의 원리:** `top: 20px`은 위로 이동하는 것이 아니라, 기준점의 상단(Top)으로부터 아래로 20px만큼 밀어내라는 의미입니다.
- **좌표 이동:**
    - **top / bottom:** 수직 방향 이동 (기준점 상단/하단으로부터의 거리)
    - **left / right:** 수평 방향 이동 (기준점 좌측/우측으로부터의 거리)
- **특징:** **relative**에서는 원래 위치가 기준이 되고, **absolute**에서는 지정된 부모 요소의 경계선이 기준이 됩니다.

```python
%%writefile ex3.html

<html>
<head>
<style>
body{

	padding: 0;
	margin: 0;
	display: flex;          /* 자식 아이템들을 가로로 나열 */
    flex-wrap: wrap;        /* 자리가 모자라면 줄 바꿈(Wrap) */
    justify-content: center; /* 가로 중앙 정렬 */
    align-items: center;      /* 세로축 중앙 정렬 */
    min-height: 100vh;		  /* 높이를 지정해야 세로축 중앙 정렬이 의미 있음 */

}
.container {
    display: flex;          /* 자식 아이템들을 가로로 나열 */
    flex-wrap: wrap;        /* 자리가 모자라면 줄 바꿈(Wrap) */
    justify-content: center; /* 가로 중앙 정렬 */
    align-items: center;      /* 세로축 중앙 정렬 */
    width: 800px;           /* 전체 컨테이너 너비 고정, 꽉 채우고 싶다면 100% */
    gap: 20px;              /* 박스 사이의 물리적 간격 */
    padding: 20px;
    background-color: white;
    border-radius: 10px;
    box-shadow: 0 4px 6px rgba(0,0,0,0.1);
}

.parent {
  /* 1.2 position 속성: 자식의 기준점이 되기 위해 relative를 부여 */
  position: relative; // relative는 .parent 부모에 flex를 주지 않으면 좌측 위에 고정된다.
  width: 300px;
  height: 150px;
  background-color: #f0f0f0;
  border: 2px solid #333;
}

.child {
  /* 1.2 position 속성: 부모를 기준으로 흐름에서 이탈(유령 상태) */
  position: absolute;

  /* 1.3 좌표 속성: 부모의 '천장'과 '오른쪽 벽'을 기준으로 위치 고정 */
  top: 10px;    /* 위쪽 경계선에서 10px 아래로 밀어냄 */
  right: 10px;  /* 오른쪽 경계선에서 10px 왼쪽으로 밀어냄 */

  width: 80px;
  height: 40px;
  background-color: #3498db;
  color: white;
}

.next-content {
  /* 1.1 일반적인 흐름: .child가 absolute로 빠졌기 때문에 이 박스는 .parent 바로 아래에 붙습니다. */
  background-color: #e74c3c;
  height: 50px;
}
</style>
</head>

<body>
<div class="container">
    <div class="parent">
        부모 박스 (Relative)
        <div class="child">
            자식 박스 (Absolute)
        </div>
    </div>

    <div class="next-content">
        다음 콘텐츠 (Normal Flow)
    </div>
</div>
</body>
</html>
```

## **CSS3 기초 II: 포지셔닝(Relative/Absolute)의 상호작용**

웹 페이지의 요소들을 배치하는 것은 마치 빈 방에 가구를 배치하는 것과 같습니다. 어떤 가구는 벽에 붙이고, 어떤 소품은 책상 위에 겹쳐 올려두기도 하죠. CSS의 position 속성은 바로 이러한 요소들의 배치 전략과 기준점을 결정하는 핵심 도구입니다.

---

### **1. Relative & Absolute: 상호작용의 핵심**

### **1.1 Relative (상대 위치): 자기 자신을 기준으로 한 미세 조정과 '빈자리'의 유지**

**relative**는 요소가 원래 있어야 할 자리를 지키면서 위치만 살짝 이동하는 상태입니다.

- **자기 자신 기준:** 좌표 속성(`top`, `left` 등)을 사용하면 '원래 본인이 있던 위치'를 기준으로 이동합니다.
- **공간 점유:** 요소가 이동하더라도 원래 차지하고 있던 물리적인 공간은 그대로 유지됩니다. 따라서 다음에 오는 형제 요소들의 배치에 영향을 주지 않습니다.
- **기준점 역할:** 본인이 움직이는 것보다, 자식 요소인 **absolute의 기준점** 역할을 수행하기 위해 부모에게 주로 선언합니다.

### **1.2 Absolute (절대 위치): 일반 흐름에서의 이탈과 가장 가까운 조상 기준점 찾기**

**absolute**는 일반적인 문서의 흐름을 완전히 무시하고 공중으로 붕 떠오르는 상태입니다.

- **흐름 이탈:** 요소가 떠오르면서 원래 차지하던 공간이 즉시 사라집니다. 뒤에 오던 요소들이 해당 자리를 차지하기 위해 위로 밀고 올라옵니다.
- **기준점 찾기:** `static`이 아닌 포지션(`relative`, `absolute`, `fixed`)을 가진 가장 가까운 조상을 기준으로 배치됩니다.
- **특징:** 조상 중 기준이 될 요소가 하나도 없다면 최상위 요소인 `<body>`나 브라우저 화면 전체를 기준으로 배치됩니다.

### **1.3 [실무 조합]: 부모 relative + 자식 absolute 공식 완벽 이해**

실무 레이아웃 설계에서 가장 강력하고 빈번하게 사용되는 세트 구성입니다.

- **울타리와 좌표:** 부모를 **울타리(relative)**로 만들고 자식을 그 안에서 **절대 좌표(absolute)**로 고정하는 방식입니다.
- **이미지 위 배지(Badge) 배치:** 상품 카드 이미지의 우측 상단에 'SALE'이나 'HOT' 같은 배지를 겹쳐서 배치할 때 사용합니다.
- **커스텀 디자인:** 박스 내부 특정 위치에 아이콘이나 화살표를 고정할 때 유용하며, 부모 박스가 이동해도 자식은 항상 지정된 상대적 위치를 유지합니다.

### **1.4 예시 코드**

가장 빈번하게 사용되는 **relative**와 **absolute**의 상호작용을 구현한 예시와 코드입니다.

```python
%%writefile ex4.html

<html>
<head>
<style>
body{
.card {
  position: relative;   /* 자식의 기준점이 됨 */
  width: 250px;
  height: 300px;
  background-color: #f0f0f0;
  border: 1px solid #ddd;
}

.badge {
  position: absolute;   /* 일반 흐름에서 이탈하여 공중에 뜸 */
  top: 15px;            /* 부모 상단 경계에서 15px 아래로 */
  right: 15px;          /* 부모 우측 경계에서 15px 왼쪽으로 */

  padding: 5px 10px;
  background-color: #ff4757;
  color: white;
  font-weight: bold;
}
</style>
</head>
<body>
<div class="card">
  <div class="photo">상품 이미지</div>
  <span class="badge">HOT</span>
</div>
</body>
</html>
```

### **💡 코드 활용 팁**

- **기준점의 일치:** `.card`에 적용된 `position: relative`는 하위 요소인 `.badge`가 위치를 잡을 때 참고할 시작점(0,0)을 이 박스로 고정하는 역할을 합니다.
- **독립적인 배치:** `.badge`에 `position: absolute`를 사용하면, 다른 요소와의 배치 순서와 상관없이 부모 박스 내에서 독자적인 위치를 가질 수 있게 됩니다.
- **여백의 조절:** `top: 15px`과 `right: 15px` 수치를 변경함으로써 부모 박스의 모서리로부터 배지가 얼마나 떨어져 있을지를 정밀하게 제어할 수 있습니다.

# 토스뱅크 실습 코드 리뷰 노트

## 잘 한 것
- `* { box-sizing: border-box; }` — 전역 설정, 실무 표준
- Flexbox로 헤더 / 메인 / 카드 레이아웃 전부 처리
- `position: relative + absolute` 조합으로 뱃지·닫기버튼 위치 잡음
- `transition: transform 0.2s` 호버 효과 적용

## 알아둘 것

**`flex: 2.5 / flex: 1`**
- 전체 너비를 3.5 등분 → 메인 2.5칸, 사이드 1칸
- 숫자만 바꾸면 비율 조절 가능

**`height: fit-content`**
- 고정값 없이 내용물 높이에 맞게 자동 조절
- 콘텐츠 양이 달라져도 레이아웃 안 깨짐

## **CSS3 기초 II: Fixed & Sticky: 스크롤에 반응하는 UI**

웹 페이지를 스크롤 하더라도 특정 메뉴가 상단에 계속 붙어 있거나, 화면 구석에 'TOP' 버튼이 떠 있는 것을 본 적이 있을 것입니다. 이는 요소가 스크롤 흐름을 따르지 않고 사용자 화면을 기준으로 위치를 고정하기 때문입니다.

---

### **1. Fixed (고정 위치): 뷰포트(브라우저 창) 기준 배치와 페이지 고정 레이아웃**

**fixed**는 요소를 현재 눈에 보이는 브라우저 화면(Viewport)의 특정 지점에 박제하는 속성입니다.

- **화면 기준 배치:** 부모 요소가 누구든 상관없이 브라우저 창 전체를 기준으로 좌표(`top`, `right` 등)를 잡습니다.
- **스크롤 무시:** 페이지가 아무리 길어서 스크롤을 끝까지 내려도, 요소는 화면의 정해진 위치에 항상 고정되어 나타납니다.
- **공간 이탈:** 일반적인 문서 흐름에서 완전히 빠지기 때문에 원래 차지하던 공간은 사라집니다.
- **주요 사례:** 상단 고정 내비게이션 바, 화면 하단 우측의 'TOP' 버튼, 화면 중앙에 나타나는 모달 팝업 배경 등에 사용됩니다.

### **2. Sticky (끈적이는 위치): 조건부 고정(스크롤 임계점)의 이해와 활용**

**sticky**는 평소에는 일반적인 흐름을 따르다가, 스크롤이 특정 지점에 도달하면 화면에 끈질기게 달라붙는 속성입니다.

- **임계점 기준:** `top: 0`과 같은 좌표 설정이 필수입니다. 스크롤이 이 설정값에 닿기 전까지는 `relative`처럼 행동하고, 닿는 순간 `fixed`처럼 고정됩니다.
- **부모 내부 동작:** 고정된 상태는 오직 부모 요소의 영역 안에서만 유효합니다. 부모 박스가 스크롤되어 화면 밖으로 완전히 사라지면 sticky 요소도 함께 따라 올라가며 사라집니다.
- **주요 사례:** 긴 표의 제목 줄(Header) 고정, 블로그 본문 옆의 사이드바 목차(Index) 고정 등에 유용합니다.

### **3. Fixed와 Sticky의 핵심 차이점**

두 속성 모두 화면에 고정된다는 점은 비슷하지만, 기준과 동작 범위에서 명확한 차이가 있습니다.

- **기준점:** `fixed`는 오직 브라우저 창(Viewport)이 기준이지만, `sticky`는 일반 흐름을 따르며 부모 요소를 기준으로 고정됩니다.
- **영향력:** `fixed`는 공간을 아예 차지하지 않아 주변 요소와 겹치지만, `sticky`는 고정되기 전까지는 자신의 원래 자리를 차지하고 있습니다.
- **해제 조건:** `fixed`는 해제 조건이 없으나, `sticky`는 부모의 경계선을 만나면 고정 상태가 해제됩니다.

### **4. Fixed & Sticky 활용 예시 코드 및 활용 팁**

브라우저 화면에 고정되는 `fixed`와 부모 안에서 끈질기게 붙어있는 `sticky`의 구현 예시입니다.

```python
%%writefile sticky.html

<html>
<body>

<head>
<style>
/* 상단 고정 바 */
.navbar {
  position: fixed;
  top: 0; left: 0;
  width: 100%; height: 60px;
  background: #333; color: #fff;
  z-index: 1000;
  display: flex; align-items: center; padding-left: 20px;
}

/* 전체 컨테이너 (상단바 높이만큼 마진 확보) */
.container {
  margin-top: 80px; /* navbar(60px) + 여유공간(20px) */
  padding: 0 20px;
}

/* 사이드바와 본문의 가로 배치 */
.content-box {
  display: flex;
  gap: 20px;
  align-items: flex-start; /* sticky 작동을 위한 필수 설정 */
}

/* 끈질기게 따라오는 사이드바 */
.sidebar {
  position: sticky;
  top: 80px; /* 화면 상단 80px 지점에서 고정 */
  width: 200px;
  height: 300px;
  background: #f4f4f4;
  padding: 15px;
}

/* 본문 영역 (추가된 부분) */
.main-text {
  flex: 1;           /* 남는 가로 공간을 모두 점유 */
  min-height: 1500px; /* 스크롤을 만들기 위한 강제 높이 */
  background: white;
  padding: 20px;
  border: 1px solid #ddd;
}

/* 우측 하단 고정 버튼 */
.top-btn {
  position: fixed;
  bottom: 30px; right: 30px;
  padding: 10px 20px;
  background: orange; border: none; cursor: pointer;
}
</style>
</head>

<body>
<header class="navbar">상단 고정 메뉴 (Fixed)</header>

<main class="container">
  <div class="content-box">
    <aside class="sidebar">사이드바 (Sticky)</aside>
    <section class="main-text">본문 내용 (스크롤 가능 영역)</section>
  </div>
</main>

<button class="top-btn">TOP</button>
</body>
</html>
```

# Sticky & Fixed 포지션 실습 코드 : 기능(동작) 설명
###*: 전체 페이지를 `상단 고정 바` / `사이드바` / `본문` / `우측 하단 버튼`으로 구성*

## 1단계 - 상단 고정 바 (.navbar)
- `position: fixed`로 스크롤과 무관하게 항상 화면 상단에 고정
- `top: 0; left: 0; width: 100%`로 화면 전체 너비를 차지
- `z-index: 1000`으로 다른 요소들 위에 항상 올라와 있음
- `display: flex; align-items: center`로 내부 텍스트를 세로 중앙 정렬

## 2단계 - 전체 컨테이너 (.container)
- `margin-top: 80px`으로 `navbar(60px)` + 여유공간`(20px)`만큼 아래에서 시작
- `navbar`가 `fixed`라 일반 흐름에서 벗어나 있어, 본문이 navbar에 가리지 않도록 밀어내는 역할

## 3단계 - 사이드바와 본문 가로 배치 (.content-box)
- `display: flex; gap: 20px`으로 사이드바와 본문을 가로로 나란히 배치
- `align-items: flex-start` 필수 설정 - 이게 없으면 `sticky`가 작동하지 않음
- (기본값 `stretch`면 자식 높이가 부모에 맞춰져 sticky 고정 기준이 사라짐)

## 4단계 - 사이드바 (.sidebar)
- `position: sticky; top: 80px`으로 스크롤 시 화면 상단 80px 지점에서 멈춰 따라옴
- `fixed`와 달리 부모 요소(`.content-box`) 안에서만 고정되고, 부모 영역을 벗어나면 함께 사라짐

## 5단계 - 본문 영역 (.main-text)
- `flex: 1`로 사이드바`(200px)`를 제외한 나머지 가로 공간을 모두 차지
- `min-height: 1500px`으로 스크롤이 생기도록 강제로 높이를 늘림
- (스크롤이 없으면 `sticky` 동작을 확인할 수 없음)

## 6단계 - 우측 하단 고정 버튼 (.top-btn)
- `position: fixed; bottom: 30px; right: 30px`으로
스크롤과 무관하게 항상 화면 우측 하단에 고정

###*: 최종적으로 스크롤 시 `navbar`는 상단에 고정, `사이드바`는 본문 옆에서 따라오고, `TOP 버튼`은 우측 하단에 고정된 레이아웃이 완성~!*

```python
%%writefile FixedSticky.html

<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <title>Fixed & Sticky 실습 완성본</title>
  <style>
    /* 0. 기본 초기화 */
    * { margin: 0; padding: 0; box-sizing: border-box; }
    body { font-family: sans-serif; line-height: 1.6; background: #f9f9f9; }

    /* ③ 반투명 상단 고정 헤더 (Fixed) */
    .navbar {
      position: fixed;
      top: 0; left: 0;
      width: 100%; height: 60px;
      background: rgba(51, 51, 51, 0.8); /* 반투명 배경 */
      color: white;
      display: flex; align-items: center; padding: 0 20px;
      backdrop-filter: blur(5px); /* 배경 흐림 효과 */
      z-index: 1000;
    }

    /* 헤더 때문에 가려지는 공간 확보 */
    .container { margin-top: 80px; padding: 0 50px; display: flex; gap: 30px; }

    /* ② 부모 박스를 넘지 않는 스크롤 목차 (Sticky) */
    .content-box { display: flex; flex: 1; gap: 20px; align-items: flex-start; }

    .sidebar-toc {
      position: sticky;
      top: 80px; /* 헤더 아래에서 고정 시작 */
      width: 200px;
      background: #eee; padding: 15px;
      border-radius: 8px;
    }

    .main-text { flex: 1; background: white; padding: 30px; min-height: 2000px; }

    /* ④ 표(Table)의 첫 번째 행 고정 (Sticky) */
    table { width: 100%; border-collapse: collapse; margin-top: 20px; }
    th, td { border: 1px solid #ddd; padding: 10px; text-align: left; }
    th {
      position: sticky;
      top: 60px; /* 헤더 바로 아래에 붙도록 설정 */
      background: #3498db; color: white;
    }

    /* ① 스크롤 시 따라오는 퀵 메뉴 (Fixed) */
    .quick-menu {
      position: fixed;
      top: 50%; right: 20px;
      transform: translateY(-50%);
      width: 80px; background: orange; color: white;
      padding: 10px; text-align: center; border-radius: 10px;
      font-size: 12px; cursor: pointer; z-index: 999;
    }

    /* ⑤ 하단 쿠키 동의 바 (Fixed) */
    .cookie-bar {
      position: fixed;
      bottom: 0; left: 0;
      width: 100%; height: 50px;
      background: #222; color: #ccc;
      display: flex; justify-content: space-between; align-items: center;
      padding: 0 30px; font-size: 13px; z-index: 1001;
    }
    .cookie-bar button { background: orange; border: none; padding: 5px 10px; cursor: pointer; }
  </style>
</head>
<body>

  <header class="navbar">MY WEBSITE (Fixed Header)</header>

  <div class="quick-menu">상담하기<br>SNS</div>

  <main class="container">
    <div class="content-box">
      <aside class="sidebar-toc">
        <h4>목차 (Sticky)</h4>
        <ul>
          <li>소개</li>
          <li>특징</li>
          <li>데이터 표</li>
        </ul>
      </aside>

      <section class="main-text">
        <h2>본문 영역 (이 영역이 길어야 실습 확인 가능)</h2>
        <p>스크롤을 내리면 왼쪽 목차는 부모 박스가 끝날 때까지만 따라오고,</p>
        <p>상단 헤더와 퀵 메뉴는 화면에 계속 붙어 있습니다.</p>

        <table>
          <thead>
            <tr><th>번호</th><th>항목</th><th>내용</th></tr>
          </thead>
          <tbody>
            <tr><td>1</td><td>데이터</td><td>스크롤 시 파란색 줄이 고정됩니다.</td></tr>
            <script>
              for(let i=2; i<=30; i++) {
                document.write(`<tr><td>${i}</td><td>데이터 항목</td><td>상세 내용 ${i}</td></tr>`);
              }
            </script>
          </tbody>
        </table>
      </section>
    </div>
  </main>

  <div class="cookie-bar">
    <span>이 사이트는 쿠키를 사용합니다. 동의하시겠습니까?</span>
    <button>확인</button>
  </div>

</body>
</html>
```
