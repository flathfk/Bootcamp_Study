```python
%%writefile ex4.html

<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <title>DOM 실습 - 이미지 슬라이더</title>
  <style>
    /* CSS: 슬라이더 외형 디자인 */
    body { display: flex; justify-content: center; align-items: center; height: 100vh; margin: 0; background: #f4f4f4; font-family: sans-serif; }
    .slider-card { width: 500px; background: white; padding: 20px; border-radius: 15px; box-shadow: 0 10px 30px rgba(0,0,0,0.1); text-align: center; }

    #slider-img {
      width: 100%; height: 300px; object-fit: cover; border-radius: 10px;
      border: 4px solid #333; transition: transform 0.3s;
    }

    #counter { font-size: 1.2rem; font-weight: bold; margin: 15px 0; color: #555; }

    .btn-group button {
      padding: 10px 25px; font-size: 16px; cursor: pointer; border: none;
      border-radius: 5px; background: #333; color: white; transition: 0.2s;
    }
    .btn-group button:hover { background: #555; }
    .btn-group button:active { transform: scale(0.95); }
  </style>
</head>
<body>

  <div class="slider-card">
    <img id="slider-img" src="https://picsum.photos/id/10/500/300" alt="슬라이드">
    <div id="counter">1 / 5</div>

    <div class="btn-group">
      <button id="prev-btn">이전</button>
      <button id="next-btn">다음</button>
    </div>
  </div>

  <script>
    // 2. DOM 요소 선택 (객체 가져오기)
    const imgObj = document.getElementById('slider-img');
    const countObj = document.getElementById('counter');

    // 3. 데이터 준비 (이미지 주소 5개 배열)
    const images = [
      "https://picsum.photos/id/10/500/300",
      "https://picsum.photos/id/20/500/300",
      "https://picsum.photos/id/30/500/300",
      "https://picsum.photos/id/40/500/300",
      "https://picsum.photos/id/50/500/300"
    ];

    let index = 0;

    // 4. 화면 업데이트 함수 (객체 속성 조작 핵심)
    function updateSlider() {
      // img 객체의 src 속성을 변경 (이미지 교체)
      imgObj.src = images[index];
      // div 객체의 텍스트 내용을 변경 (숫자 교체)
      countObj.textContent = `${index + 1} / ${images.length}`;
    }

    // 5. 버튼 이벤트 연결
    document.getElementById('next-btn').onclick = () => {
      index = (index + 1) % images.length; // 0->1->2->3->4->0 순환
      updateSlider();
    };

    document.getElementById('prev-btn').onclick = () => {
      index = (index - 1 + images.length) % images.length; // 역순 순환
      updateSlider();
    };
  </script>

</body>
</html>
```

```python
%%writefile ex5.html

<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <style>
    body { font-family: 'Pretendard', sans-serif; display: flex; flex-direction: column; align-items: center; padding: 50px; background: #f8f9fa; }
    .card-container { display: flex; gap: 20px; margin-bottom: 30px; }

    /* 기본 카드 스타일 */
    .card {
      width: 150px; padding: 20px; background: white; border: 2px solid #ddd;
      border-radius: 12px; cursor: pointer; transition: 0.3s; text-align: center;
    }

    /* 1. classList로 제어할 '선택됨' 스타일 */
    .card.selected { border-color: #007bff; background-color: #e7f1ff; transform: translateY(-5px); box-shadow: 0 5px 15px rgba(0,123,255,0.2); }

    #display-area { font-size: 1.5rem; font-weight: bold; color: #333; padding: 20px; border-top: 2px solid #333; }
    .price-tag { color: #007bff; }
  </style>
</head>
<body>

  <h2>원하시는 요금제를 선택하세요</h2>

  <div class="card-container">
    <div class="card" data-plan="베이직" data-price="9,900">
      <h3>Basic</h3>
    </div>
    <div class="card" data-plan="스탠다드" data-price="19,900">
      <h3>Standard</h3>
    </div>
    <div class="card" data-plan="프리미엄" data-price="29,900">
      <h3>Premium</h3>
    </div>
  </div>

  <div id="display-area">요금제를 선택해주세요.</div>

  <script>
    // [1] querySelectorAll: 모든 카드 요소를 한꺼번에 선택 (NodeList)
    const cards = document.querySelectorAll('.card');
    const display = document.querySelector('#display-area');

    cards.forEach(card => {
      card.onclick = function() {
        // [2] classList: 기존에 선택된 카드가 있다면 선택 해제
        // querySelector는 첫 번째 요소만 찾으므로 현재 'selected'인 요소 하나만 찾아 제거함
        const currentSelected = document.querySelector('.card.selected');
        if (currentSelected) {
          currentSelected.classList.remove('selected');
        }

        // [3] classList: 클릭된 카드에 'selected' 클래스 추가
        this.classList.add('selected');

        // [4] Attribute: 클릭된 객체의 사용자 정의 속성(data-) 읽기
        const planName = this.getAttribute('data-plan');
        const planPrice = this.getAttribute('data-price');

        // [5] 화면 업데이트
        display.innerHTML = `선택된 플랜: <span class="price-tag">${planName} (월 ${planPrice}원)</span>`;
      };
    });
  </script>

</body>
</html>
```

# 요금제 선택 카드 UI : 기능(동작) 설명
###*: 카드 클릭 시 선택 스타일이 토글되고, 선택된 요금제 정보를 화면에 출력하는 인터랙티브 UI*

## 1단계 - 카드 요소 전체 선택
- `querySelectorAll('.card')`로 카드 3개를 한꺼번에 NodeList로 가져옴
- `querySelector('#display-area')`로 하단 텍스트 출력 영역을 가져옴

## 2단계 - 각 카드에 클릭 이벤트 등록
- `cards.forEach(card => { card.onclick = function() {...} })`로
  3개 카드 각각에 클릭 이벤트를 붙임

## 3단계 - 기존 선택 카드 해제
- `document.querySelector('.card.selected')`로 현재 `selected` 클래스가 붙은 카드를 찾음
- 있으면 `classList.remove('selected')`로 선택 스타일을 제거함
- (한 번에 하나만 선택되도록 하는 로직)

## 4단계 - 클릭된 카드에 선택 스타일 적용
- `this.classList.add('selected')`로 클릭된 카드에 `selected` 클래스를 추가함
- CSS에서 `.card.selected`에 파란 테두리 / 배경색 / 위로 뜨는 효과가 정의돼 있어 시각적으로 선택됨을 표시함

## 5단계 - data 속성에서 요금제 정보 읽기
- `this.getAttribute('data-plan')`으로 클릭된 카드의 요금제 이름을 읽음
- `this.getAttribute('data-price')`으로 가격을 읽음
- `data-*` 속성은 HTML 요소에 JS에서 쓸 커스텀 데이터를 저장하는 방식임

## 6단계 - 화면 업데이트
- `display.innerHTML`에 템플릿 리터럴로 요금제 이름과 가격을 조합해 출력함

###*: 최종적으로 카드 클릭 시 이전 선택이 해제되고 새 카드가 강조되며, 하단에 선택된 요금제 정보가 출력되는 UI가 완성~!*
