````python
%%writefile server.js

const express = require("express");
const path = require("path");
const app = express();

app.use(express.static(path.join(__dirname, "static")));

app.get("/", (req, res) => {
  res.sendFile(path.join(__dirname, "templates", "index.html"));
});
app.get("/about", (req, res) => {
  res.sendFile(path.join(__dirname, "templates", "about.html"));
});
app.get("/deposit", (req, res) => {
  res.sendFile(path.join(__dirname, "templates", "deposit.html"));
});
app.get("/loan", (req, res) => {
  res.sendFile(path.join(__dirname, "templates", "loan.html"));
});
app.get("/card", (req, res) => {
  res.sendFile(path.join(__dirname, "templates", "card.html"));
});
app.get("/support", (req, res) => {
  res.sendFile(path.join(__dirname, "templates", "support.html"));
});

app.listen(3000, () => {
  console.log("Server running on port 3000");
});
````

## 📁 폴더 구조 개선 & 보안 강화

### 변경 전 (어제)
```
/content/
├── server.js
├── index.html
├── about.html
├── deposit.html
├── loan.html
├── card.html
├── support.html
└── logo.png
```

### 변경 후 (오늘)
```
/content/
├── server.js
├── templates/
│   ├── index.html
│   ├── about.html
│   ├── deposit.html
│   ├── loan.html
│   ├── card.html
│   └── support.html
└── static/
    └── logo.png
```

### 개선 이유
- static/ : 이미지 등 누구나 요청 가능한 파일만 공개
- templates/ : HTML은 서버 라우트가 직접 제어
- server.js 등 민감한 파일이 외부에 노출되지 않음
