### **2. 중앙 정렬의 기술: `absolute` 요소를 가로/세로 정중앙에 배치하기 (`transform` 활용)**

요소를 부모의 정중앙에 배치하는 것은 디자인에서 가장 자주 쓰이는 테크닉 중 하나입니다.

- **좌표의 한계:** `top: 50%; left: 50%;`만 주면 요소의 '왼쪽 상단 꼭짓점'이 중앙에 오기 때문에 실제로는 오른쪽 아래로 치우쳐 보입니다.
- **`transform` 보정:** `transform: translate(-50%, -50%);`를 추가하면 요소의 중심점을 자기 크기의 절반만큼 역방향으로 이동시켜 완벽한 정중앙 정렬을 완성합니다.
- **장점:** 요소의 가로/세로 크기를 정확히 모르더라도 언제나 중앙을 유지할 수 있습니다.


### 2-1 Absolute 중앙 정렬 기술

이 기술은 부모의 50% 지점으로 요소를 보낸 뒤, `transform`을 통해 요소 자기 자신의 크기 절반-50%만큼 역방향으로 이동시켜 정중앙을 맞추는 기법입니다.

- HTML 구성

기준점이 될 부모 박스와 그 안에 중앙에 위치할 자식 박스입니다.

```python
%%writefile test.html

<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <style>
    /* CSS 여기 */
    .parent-box {
      position: relative;
      width: 100%;
      height: 300px;
      background-color: #f8f9fa;
      border: 2px dashed #adb5bd;
    }
    .center-item {
      position: absolute;
      top: 50%;
      left: 50%;
      transform: translate(-50%, -50%);
      padding: 15px 30px;
      background-color: #3498db;
      color: white;
      font-weight: bold;
      border-radius: 8px;
    }
  </style>
</head>
<body>

  <!-- HTML 여기 -->
  <div class="parent-box">
    <div class="center-item">
      정중앙 정렬
    </div>
  </div>

</body>
</html>
```

### 핵심 요약

- **`top: 50%`, `left: 50%`**: 기준을 부모의 절반 지점으로 잡습니다.
- **`transform: translate(-50%, -50%)`**: 요소가 어느 쪽으로도 치우치지 않게 '자기 몸집'만큼 보정합니다.

### **3. Float vs Flexbox: 언제 어떤 도구를 선택해야 하는가?**

과거의 표준인 `float`과 현대의 표준인 `flexbox`는 각각의 강점이 다릅니다.

- **Float을 쓰는 경우:** 신문처럼 텍스트가 이미지 주위를 감싸야 하는 디자인을 구현할 때 여전히 유일하고 강력한 도구입니다.
- **Flexbox를 쓰는 경우:** 단순 정렬을 넘어 요소 간의 간격 배분, 순서 변경, 반응형 가변 크기 조절 등이 필요한 거의 모든 현대적 웹 레이아웃에 최적화되어 있습니다.
- **결론:** 단순한 흐름 제어는 `float`을, 복잡한 구조의 시스템 레이아웃은 `flexbox`를 선택하는 것이 효율적입니다.

### 3-1 Float vs Flexbox 사례 비교

- HTML 구성

두 가지 사례를 각각의 컨테이너로 나누어 구성했습니다.

```python
%%writefile test2.html

<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <style>
    /* CSS 여기 */
    /* --- Float: 흐름 제어 --- */
.float-container {
  width: 100%;
  padding: 20px;
  background-color: white;
  /* 부모가 자식의 height를 인지하게 함 (clearfix 대용) */
  overflow: hidden;
}

.float-image {
  float: left;        /* 핵심: 왼쪽으로 밀고 텍스트를 오른쪽으로 보냄 */
  width: 100px;
  height: 100px;
  background-color: #3498db;
  margin-right: 15px; /* 텍스트와의 간격 */
  display: flex; align-items: center; justify-content: center; color: white;
}

/* --- Flexbox: 시스템 정렬 --- */
.flex-container {
  display: flex;      /* 핵심: 정렬 시스템 가동 */
  justify-content: space-between; /* 요소 간 간격 균등 배분 */
  align-items: center;          /* 수직 중앙 정렬 (매우 간편함) */
  padding: 20px;
  background-color: #eee;
}

.flex-item {
  width: 30%;
  height: 80px;
  background-color: #2ecc71;
  display: flex; align-items: center; justify-content: center;
  color: white; font-weight: bold;
}
  </style>
</head>
<body>

  <!-- HTML 여기 -->
  <div class="float-container">
  <div class="float-image">IMAGE</div>
  <p class="float-text">
    이 텍스트는 이미지 옆에서 시작하여 이미지 아래로 자연스럽게 흐릅니다.
    신문이나 블로그 기사처럼 이미지를 글자가 감싸야 할 때 여전히 유일하고 강력한 도구입니다.
  </p>
</div>

<hr>

<div class="flex-container">
  <div class="flex-item">카드 1</div>
  <div class="flex-item">카드 2</div>
  <div class="flex-item">카드 3</div>
</div>

</body>
</html>
```

### 핵심 요약

- **Float**: "글자가 그림을 피해가게 하고 싶을 때" 씁니다. (레이아웃보다는 **콘텐츠 배치**용)
- **Flexbox**: "박스들을 나란히 배치하고, 간격을 맞추고, 높이를 정렬하고 싶을 때" 씁니다. (**전체 레이아웃**용)

### **5. 레이아웃 심화 실습 주제 5개**

포지셔닝 심화 기술인 `z-index` 계층 구조와 `transform` 중앙 정렬, 그리고 상황에 맞는 레이아웃 도구 선택 능력을 기르기 위한 실습 과제입니다.

1. **화면 정중앙에 위치하는 모달 창(Modal) 만들기**
    - **내용:** 배경은 반투명하게 처리하고, 그 위에 흰색 공지 사항 창이 화면 가로/세로 정중앙에 오도록 배치하세요.
    - **핵심:** `position: absolute`와 `transform: translate(-50%, -50%)` 공식을 사용하여 완벽한 중앙 정렬을 구현합니다.
2. **이미지 위에 겹쳐진 'Sold Out' 배지**
    - **내용:** 상품 이미지 우측 상단에 원형으로 된 '품절' 배지를 겹쳐서 표시하세요.
    - **핵심:** 부모 요소에 `relative`, 배지 요소에 `absolute`를 부여하고 `z-index`를 이용해 이미지보다 위에 보이도록 설정합니다.
3. **겹쳐진 카드 덱(Card Deck) 효과**
    - **내용:** 3장의 카드가 조금씩 어긋나게 겹쳐져 있는 디자인을 만드세요.
    - **핵심:** 각 카드에 서로 다른 `z-index` 값을 부여하여 앞뒤 순서를 제어하고, `hover` 시 특정 카드가 가장 위로 올라오도록 구현합니다.
4. **반응형 기사형 레이아웃 (Float 활용)**
    - **내용:** 신문 사이트처럼 텍스트 중간에 이미지를 왼쪽 혹은 오른쪽에 배치하고, 글자가 이미지의 곡선을 따라 흐르듯 배치되는 페이지를 구성하세요.
    - **핵심:** `flex`로는 불가능한 `float` 고유의 텍스트 감싸기 기능을 실습하고, 문단 끝에서 `clear`를 통해 흐름을 끊어주는 연습을 합니다.
5. **내비게이션 바 위의 알림 숫자 배지**
    - **내용:** 상단 메뉴 아이콘(예: 종 모양)의 오른쪽 위에 작게 '9+'라고 쓰인 빨간색 알림 숫자를 붙이세요.
    - **핵심:** 아이콘을 기준으로 숫자의 위치를 세밀하게 조정하는 `absolute` 포지셔닝 기술을 익힙니다.

### **6. 레이아웃 심화 실습 주제별 모범 코드**

**5번**에서 제시한 심화 실습 과제들을 구현하기 위한 핵심 코드입니다. `z-index`와 `transform`을 활용한 정교한 배치 방법을 확인해 보세요.

```python
%%writefile test3.html

<!DOCTYPE html>
<html lang="ko">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>레이아웃 심화 실습 (1-5)</title>
  <style>
    /* [공용 스타일] */
    * { margin: 0; padding: 0; box-sizing: border-box; }
    body { font-family: 'Pretendard', sans-serif; background: #f8f9fa; color: #333; padding: 100px 20px 50px; }
    h3 { margin-bottom: 15px; padding-bottom: 5px; border-bottom: 2px solid #dee2e6; }
    section { max-width: 800px; margin: 0 auto 80px; background: white; padding: 30px; border-radius: 12px; shadow: 0 4px 6px rgba(0,0,0,0.05); }

    /* ------------------------------------------------------
       ① 화면 정중앙 모달 창 (Modal)
       핵심: absolute + top/left 50% + transform 보정
    ------------------------------------------------------ */
    .modal-overlay {
      position: fixed; top: 0; left: 0; width: 100%; height: 100%;
      background: rgba(0,0,0,0.5); z-index: 9999;
    }
    .modal-window {
      position: absolute;
      top: 50%; left: 50%;
      transform: translate(-50%, -50%); /* 내 몸집의 절반만큼 역방향 이동 */
      width: 300px; padding: 20px; background: white;
      border-radius: 10px; text-align: center;
      box-shadow: 0 10px 30px rgba(0,0,0,0.3);
    }
    .close-btn { margin-top: 15px; padding: 5px 15px; cursor: pointer; }

    /* ------------------------------------------------------
       ② 이미지 위 'Sold Out' 배지
       핵심: 부모 relative + 자식 absolute + z-index
    ------------------------------------------------------ */
    .product-box {
      position: relative; /* 배지의 기준점 */
      width: 200px; height: 200px; background: #e9ecef;
      display: flex; align-items: center; justify-content: center;
      border-radius: 8px; font-weight: bold; color: #adb5bd;
    }
    .sold-out-badge {
      position: absolute;
      top: 10px; right: 10px;
      background: #fa5252; color: white; padding: 4px 12px;
      border-radius: 20px; font-size: 12px; z-index: 10;
    }

    /* ------------------------------------------------------
       ③ 겹쳐진 카드 덱 (Card Deck) 효과
       핵심: z-index 계층 제어 + hover 시 우선순위 변경
    ------------------------------------------------------ */
    .card-deck { position: relative; height: 220px; }
    .card {
      position: absolute; width: 140px; height: 180px;
      border-radius: 10px; border: 3px solid white;
      display: flex; align-items: center; justify-content: center;
      color: white; font-weight: bold; transition: 0.3s;
    }
    .card-1 { background: #4dabf7; left: 0; z-index: 1; }
    .card-2 { background: #51cf66; left: 40px; top: 20px; z-index: 2; }
    .card-3 { background: #fcc419; left: 80px; top: 40px; z-index: 3; }
    /* hover 시 해당 카드가 가장 위로(z-index) 올라옴 */
    .card:hover { z-index: 100; transform: translateY(-15px); cursor: pointer; }

    /* ------------------------------------------------------
       ④ 반응형 기사형 레이아웃 (Float 활용)
       핵심: float으로 텍스트 감싸기 + clear로 흐름 해제
    ------------------------------------------------------ */
    .article-container { line-height: 1.7; color: #495057; }
    .float-image {
      float: left; /* 이미지 주변으로 텍스트가 흐름 */
      width: 150px; height: 120px; background: #339af0;
      margin-right: 20px; margin-bottom: 10px;
      display: flex; align-items: center; justify-content: center; color: white;
    }
    .clear-fix { clear: both; } /* float의 영향력이 다음 요소에 미치지 않도록 차단 */

    /* ------------------------------------------------------
       ⑤ 내비게이션 바 알림 숫자 배지
       핵심: 아이콘 내부 relative + 숫자 absolute 미세 조정
    ------------------------------------------------------ */
    .nav-bar {
      position: fixed; top: 0; left: 0; width: 100%; height: 60px;
      background: #212529; display: flex; align-items: center; padding: 0 30px;
      z-index: 1000;
    }
    .nav-icon {
      position: relative; /* 알림 숫자의 기준점 */
      font-size: 24px; color: white; cursor: pointer;
    }
    .notification {
      position: absolute;
      top: -5px; right: -8px; /* 아이콘 우측 상단에 걸치도록 조정 */
      background: #f03e3e; color: white; font-size: 10px;
      padding: 2px 5px; border-radius: 10px; font-weight: bold;
      border: 2px solid #212529; /* 배경색과 동일한 테두리로 구분감 생성 */
    }
  </style>
</head>
<body>

  <nav class="nav-bar">
    <div class="nav-icon">
      🔔
      <span class="notification">9+</span>
    </div>
  </nav>

  <div class="modal-overlay" id="modal">
    <div class="modal-window">
      <h3>① 중앙 모달</h3>
      <p>transform: translate(-50%, -50%)로<br>완벽한 정중앙에 위치합니다.</p>
      <button class="close-btn" onclick="document.getElementById('modal').style.display='none'">닫기</button>
    </div>
  </div>

  <section>
    <h3>② 상품 배지 겹치기</h3>
    <div class="product-box">
      상품 이미지
      <span class="sold-out-badge">Sold Out</span>
    </div>
  </section>

  <section>
    <h3>③ 카드 덱 z-index 제어</h3>
    <div class="card-deck">
      <div class="card card-1">1</div>
      <div class="card card-2">2</div>
      <div class="card card-3">3</div>
    </div>
    <p style="font-size: 13px; color: #868e96; margin-top: 10px;">* 카드 위에 마우스를 올려보세요.</p>
  </section>

  <section>
    <h3>④ Float 텍스트 감싸기</h3>
    <div class="article-container">
      <div class="float-image">기사 이미지</div>
      <p>
        이 텍스트는 float 속성이 적용된 이미지 옆에서 흐르기 시작합니다.
        이미지의 높이가 끝나면 텍스트는 다시 왼쪽 끝으로 돌아와 전체 너비를 채우게 됩니다.
        이것은 flex나 grid로는 구현할 수 없는 float 고유의 텍스트 흐름 제어 기능입니다.
      </p>
      <div class="clear-fix"></div> </div>
  </section>

</body>
</html>
```

### **💡 실무 포인트 (Insight)**

1. **z-index의 계층 설계:** `fixed` 헤더는 서비스 전체에서 가장 위에 있어야 하므로 `9999` 같은 큰 값을, 배지는 그보다 낮은 값을 주어 계층 구조를 명확히 합니다.
2. **Float vs Sticky:** `float`은 큰 틀(왼쪽 사이드바, 오른쪽 본문)을 잡는 데 쓰고, `sticky`는 그 사이드바 안의 **콘텐츠가 스크롤을 따라오게** 만드는 데 사용했습니다.
3. **사용자 경험(UX):** `fixed` 헤더에 `background: rgba(...)`와 `backdrop-filter: blur()`를 섞어주면 스크롤 시 뒤쪽 콘텐츠가 고급스럽게 비쳐 보입니다.
