# JavaScript 문법 요점 정리

## 1. 실행 메커니즘과 호이스팅 (Hoisting)

- **개념**: 엔진이 코드를 실행하기 전, 모든 선언문(변수, 함수)을 메모리에 우선 등록하는 현상.
- **전통적 관행과의 충돌 (변칙성)**:
    - **`var`**: 선언과 동시에 `undefined`로 초기화됨. 정의 전 호출 시 에러가 아닌 `undefined`를 반환하는 논리적 허점 존재.
    - **`let / const`**: 선언은 되나 초기화되지 않음. 정의 전 호출 시 **TDZ(일시적 사각지대)** 에러 발생으로 전통적 안전성 확보.
- **학습 요점**: "정의되지 않은 것은 호출할 수 없다"는 원칙을 위해 `const` 사용을 강제함.

## 2. 함수의 체계

### 2.1 함수 선언문 (Function Declaration)

- **형식**: `function name() { ... }`
- **특징**: 전체가 호이스팅되어 정의 위치와 상관없이 코드 어디서든 호출 가능.

### 2.2 함수 표현식 (Function Expression)

- **형식**: `const name = function() { ... }`
- **특징**: 변수 할당 시점이 곧 정의 시점임. 반드시 정의부 이후에만 호출 가능.

### 2.3 화살표 함수 (Arrow Function)

- **형식**: `const name = () => { ... }`
- **변칙성**: `function` 키워드 생략, 자신만의 `this`를 생성하지 않고 상위 스코프를 계승(Lexical this).

### 2.4 콜백 함수 (Callback Function)

- **본질**: 함수 호출 권한을 다른 함수나 시스템에 위임.
- **전달**: `sayHello`처럼 이름(정의부)만 전달하고, **괄호(호출부)**는 시스템이 결정함.

## 3. 데이터 가공: 고차 함수 (Higher-Order Function)

- **선언적 프로그래밍**: '어떻게'가 아닌 '무엇을' 할지에 집중하여 데이터를 가공하는 방식.
- **레코드 단위 처리 (Row-oriented)**:
    - 원본 배열을 수정하지 않고 새로운 레코드 배열을 반환(비변경성).
- **핵심 메서드**:
    1. **`map`**: 레코드 1:1 변환 (새 배열 생성).
    2. **`filter`**: 조건에 맞는 레코드 추출 (새 배열 생성).
    3. **`reduce`**: 데이터를 하나의 값으로 응축.
- **구조적 결합**: 고차 함수 내부로 전달되는 **화살표 함수가 처리 정의부**이며, 고차 함수가 이를 반복 **호출**함.

## 4. 비동기 구조와 이벤트 루프 (Event Loop)

- **싱글 스레드 제약**: 자바스크립트는 한 번에 하나의 **호출부**만 처리 가능.
- **작동 기제**:
    1. **콜 스택(Call Stack)**: 현재 실행 중인 호출부 관리.
    2. **태스크 큐(Task Queue)**: 실행 대기 중인 콜백 함수 적재.
    3. **이벤트 루프**: 스택이 비었을 때 큐의 콜백을 스택으로 이동시켜 호출 실행.

## 5. 데이터 안전성 및 방어적 코딩

- **선택적 체이닝 (`?.`)**: 객체 속성 부재 시 에러 중단 없이 `undefined` 반환.
- **Null 병합 (`??`)**: 데이터 로딩 실패 시 유효한 기본값 할당.
- **학습 요점**: 데이터 무결성 확보를 위한 현대적 예외 처리 관행.

## 6. 실전 적용: 라이브러리 호출 (Chart.js 예시)

- **호출부의 실체**: `new Chart(ctx, { ... })`
- **분석**:
    - `Chart`는 외부 라이브러리에 정의된 생성자(정의부)임.
    - `new` 키워드를 통해 메모리 생성과 동시에 기능을 **호출**함.
    - 괄호 내부의 객체는 호출 시 전달되는 실제 데이터(Argument)임.

## 실습 4. 레코드 데이터 가공 (Higher-Order Function)

- **목표**: 행(Row) 기반 레코드 데이터를 고차 함수로 호출하여 가공하기.
- **과제**: 제시된 `timeline` 배열을 활용하여 아래 결과값을 출력하는 코드를 작성하시오.

```python
%%writefile ex4.js

const timeline = [
    { month: '1월', loan: 1200, deposit: 4000 },
    { month: '2월', loan: 1900, deposit: 4200 },
    { month: '3월', loan: 1700, deposit: 4500 }
];

// [미션 1] map()을 호출하여 'month' 정보만 담긴 새로운 배열을 만드시오.
const labels = timeline.map(record => record.month);

// [미션 2] filter()를 호출하여 loan이 1500 이상인 레코드만 추출하시오.
const highLoans = timeline.filter(record => record.loan >= 1500);

console.log(labels);
console.log(highLoans);
```

```python
!node ex4.js
```

- **분석 포인트**: map 내부의 화살표 함수가 ' 정의부 '이고, map 자체가 ' 호출부 '임을 구분할 것.

## [자가 점검 리스트]

- 1. 정의부만 있고 호출부가 없는 코드의 화면 변화 여부
    - **답변**: **No**
    - **이유**:
        - **정의부**는 메모리에 실행 로직(설계도)을 등록하는 과정일 뿐, 엔진에 실행 명령을 내리는 단계가 아님.
        - 화면에 변화를 주거나 연산을 시작하는 것은 반드시 호출부(Invocation)가 엔진을 가동해야만 가능함.
        - 따라서 호출부 없는 정의부는 실행되지 않는 '잠든 코드'와 같음.
- 2. 화살표 함수를 변수에 할당(정의)한 뒤 호출해야 하는 이유
    - **답변**: **함수 이름이 없는 익명 함수이며, 호이스팅되지 않기 때문**
    - **이유**:
        - 전통적인 함수 선언문과 달리 화살표 함수는 **함수 표현식**의 일종임.
        - 함수 자체가 이름이 없으므로 호출하기 위해서는 반드시 변수에 담아 식별자(이름)를 부여해야 함.
        - 자바스크립트 엔진은 변수 할당문이 실행되기 전까지 해당 함수의 존재를 알지 못하므로, 정의 전 호출 시 에러가 발생함.
- 3. 콜백 함수 전달 시 괄호`()`를 붙였을 때의 주도권
    - **답변**: **사용자(프로그래머)**
    - **이유**:
        - 함수 이름 뒤에 괄호를 붙이는 행위는 그 즉시 **호출**하겠다는 명령임.
        - 콜백의 본질은 호출 주도권을 시스템(예: `setTimeout`, `map`)에 **위임**하는 것인데, 괄호를 붙이면 그 시점에 사용자가 직접 실행해버리고 그 결과값(Return value)만 넘겨주게 됨.
        - 따라서 시스템이 원하는 시점에 함수를 실행하게 하려면 괄호 없이 **정의부(이름)**만 전달해야 함.
- 4. 이벤트 루프가 태스크 큐의 콜백을 실행(호출)하는 조건
    - **답변**: **콜 스택(Call Stack)이 텅 비어 있는 상태일 때**
    - **이유**:
        - 자바스크립트는 싱글 스레드 언어이므로 한 번에 하나의 **호출부**만 처리할 수 있음.
        - 현재 실행 중인 작업(스택 상단)이 모두 끝나고 스택이 완전히 비워져야만, 이벤트 루프가 대기 중인 **태스크 큐**의 콜백을 스택으로 올려 **호출**을 개시함.
        - 이 메커니즘 때문에 비동기 함수는 아무리 실행 시간이 0초라 하더라도 현재 진행 중인 동기 작업이 끝나기 전에는 호출되지 않음.
