# Visualization을 위한 JavaScript framework Chart.js

### 1. 개요

**Chart.js**는 HTML5의 **캔버스(Canvas)** 요소를 기반으로 하는 오픈 소스 JavaScript 차트 라이브러리입니다. 복잡한 데이터 세트를 시각적으로 매력적이고 이해하기 쉬운 차트로 변환해 주는 도구로, 전 세계적으로 가장 많이 사용되는 도구 중 하나입니다.

- **공식 사이트**: https://www.chartjs.org/
- **깃허브 저장소**: [https://github.com/chartjs/Chart.js](https://www.google.com/search?q=https://github.com/chartjs/Chart.js)
- **최신 버전**: 현재 v4.x 대가 주로 사용되며, 성능 최적화와 트리 쉐이킹(Tree-shaking)을 통한 모듈화가 강화되었습니다.

**Chart.js**는 "단순함"과 "유연함"을 동시에 잡은 라이브러리입니다.

- **핵심 특징**:
    - **다양한 차트 타입**: 막대(Bar), 선(Line), 파이(Pie), 레이더(Radar), 산점도(Scatter) 등 8가지 기본 차트 타입을 제공합니다.
    - **반응형(Responsive)**: 창 크기에 맞춰 차트 크기가 자동으로 조절됩니다.
    - **애니메이션**: 차트가 렌더링될 때 세련된 기본 애니메이션이 적용됩니다.
    - **경량성**: 용량이 작아 웹 페이지 로딩 성능에 미치는 영향이 적습니다.
    - **Canvas 기반**: SVG 방식인 D3.js나 ECharts와 달리 HTML5 Canvas를 사용하여 대량의 데이터를 렌더링할 때 효율적입니다.

### 3. 사용 시나리오별 실전 예제

각 예제는 CDN을 통해 라이브러리를 로드하므로, HTML 파일을 만들어 복사해 넣으면 즉시 확인 가능합니다.

### **(1) 비즈니스 매출 및 트렌드 분석 (Line Chart)**

시간의 흐름에 따른 데이터의 변화를 파악할 때 사용합니다. 매출 추이, 방문자 수 변화 등을 표현하기에 가장 적합합니다.

```python
%%writefile ex1.html

<div style="width: 500px;"><canvas id="lineChart"></canvas></div>

<script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
<script>
    const lineCtx = document.getElementById('lineChart').getContext('2d');
    new Chart(lineCtx, {
        type: 'line',
        data: {
            labels: ['1월', '2월', '3월', '4월', '5월'],
            datasets: [{
                label: '월별 매출 (백만 원)',
                data: [45, 52, 48, 70, 65],
                borderColor: 'rgb(75, 192, 192)',
                tension: 0.3, // 선의 곡선화
                fill: true,
                backgroundColor: 'rgba(75, 192, 192, 0.2)'
            }]
        }
    });
</script>
```

### **(2) 항목 간 수치 비교 및 랭킹 (Bar Chart)**

서로 다른 항목의 값을 한눈에 비교할 때 사용합니다. 제품별 판매량, 지역별 성과 지표 등에 주로 활용됩니다.

```python
%%writefile ex2.html

<div style="width: 500px;"><canvas id="barChart"></canvas></div>

<script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
<script>
    const barCtx = document.getElementById('barChart').getContext('2d');
    new Chart(barCtx, {
        type: 'bar',
        data: {
            labels: ['제품A', '제품B', '제품C', '제품D'],
            datasets: [{
                label: '판매 수량',
                data: [300, 450, 200, 500],
                backgroundColor: ['#ff6384', '#36a2eb', '#cc65fe', '#ffce56']
            }]
        },
        options: {
            scales: { y: { beginAtZero: true } }
        }
    });
</script>
```

### **(3) 점유율 및 비중 분석 (Doughnut Chart)**

전체 대비 각 항목이 차지하는 비율을 보여줄 때 사용합니다. 시장 점유율, 예산 지출 구조 등을 시각화하기 좋습니다.

```python
%%writefile ex3.html

<div style="width: 300px;"><canvas id="doughnutChart"></canvas></div>

<script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
<script>
    const doughnutCtx = document.getElementById('doughnutChart').getContext('2d');
    new Chart(doughnutCtx, {
        type: 'doughnut',
        data: {
            labels: ['검색광고', 'SNS', '이메일', '기타'],
            datasets: [{
                data: [40, 30, 20, 10],
                backgroundColor: ['#4bc0c0', '#ff9f40', '#9966ff', '#ffcd56']
            }]
        },
        options: {
            cutout: '60%' // 도넛 중앙 구멍 크기 조절
        }
    });
</script>
```

### **(4) 다각도 역량 및 성향 분석 (Radar Chart)**

여러 지표를 한 번에 비교하여 전체적인 균형이나 특징을 파악할 때 유용합니다. 게임 캐릭터 스탯, 직무 역량 평가 등에 쓰입니다.

```python
%%writefile ex4.html

<div style="width: 400px;"><canvas id="radarChart"></canvas></div>

<script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
<script>
    const radarCtx = document.getElementById('radarChart').getContext('2d');
    new Chart(radarCtx, {
        type: 'radar',
        data: {
            labels: ['의사소통', '기술력', '협업', '창의성', '문제해결'],
            datasets: [{
                label: '개발자 A 역량',
                data: [85, 95, 80, 70, 90],
                borderColor: 'red',
                backgroundColor: 'rgba(255, 99, 132, 0.2)'
            }]
        }
    });
</script>
```

### 4. 핵심 개발 팁 (Pro Tips)

1. **반응형 설정**: `options: { responsive: true, maintainAspectRatio: false }`를 설정하면 부모 컨테이너 크기에 맞춰 차트가 유연하게 변합니다.
2. **데이터 동적 업데이트**:
    - `chart.data.datasets[0].data = [new_data];` 와 같이 값을 수정합니다.
    - 반드시 **`chart.update();`** 명령어를 실행해야 애니메이션과 함께 화면이 갱신됩니다.
3. **인스턴스 관리**: 동일한 **캔버스(Canvas)** 엘리먼트에 새 차트를 다시 그리려면, 반드시 기존 차트 객체의 `chart.destroy();`를 먼저 호출하여 메모리 충돌 및 잔상 현상을 방지해야 합니다.

## 금융 데이터 로딩 및 시각화

### 1. 개요 (Overview)

실무에서는 차트 코드 안에 데이터를 직접 넣지 않습니다. 특히 금융권 대시보드는 수천 개의 트랜잭션 데이터를 다루기 때문에, 데이터를 별도의 모듈(JS 파일)로 관리하거나 API로 호출하는 방식을 사용합니다.

- **핵심 컨셉**: 데이터 전용 파일(`data.js`)을 먼저 로드하여 전역 변수나 객체에 담고, 차트 설정 파일에서 이를 참조(Reference)하여 렌더링합니다.

---

### 2. 사용 시나리오 및 장단점

- **시나리오**: 토스뱅크 내 가계 대출 추이 분석, 일일 송금 거래량 모니터링, 사용자 자산 분포도 확인 등.
- **장점**: 데이터가 아무리 많아도 HTML 코드가 깔끔하게 유지되며, 데이터만 교체하면 차트가 즉시 업데이트됩니다.
- **단점**: 외부 파일 로딩 순서가 중요합니다. 반드시 데이터 파일이 차트 실행 스크립트보다 먼저 로드되어야 합니다.

---

### 4. 핵심 개발 팁 (Pro Tips)

1. **금융권 컬러 가이드**: 금융 기업(토스, 카카오뱅크 등)은 신뢰감을 주는 **블루(#0050FF)** 계열을 주력으로 사용합니다. 차트의 `borderColor`나 `backgroundColor`에 이를 반영하면 더 전문적인 느낌을 줍니다.
2. **대량 데이터 처리 시 성능**: 데이터 포인트가 수천 개를 넘어갈 경우, `options: { parsing: false, animation: false }` 설정을 통해 렌더링 성능을 극대화할 수 있습니다.
3. **데이터 업데이트 루틴**: `data.js` 파일만 주기적으로 업데이트하거나, 서버 측에서 이 파일을 생성하게 만들면 별도의 DB 연동 로직 없이도 최신 데이터를 화면에 뿌려줄 수 있습니다.
