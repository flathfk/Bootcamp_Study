```python
import os
os.makedirs("/content/project/templates", exist_ok=True)
os.makedirs("/content/project/static", exist_ok=True)
```

```python
%%writefile /content/project/server.js

const express = require("express");
const path = require("path");

const app = express();
const PORT = 3000;

const products = [
    { id: 1, name: "MacBook Pro", price: 2890000, category: "Laptop", stock: 12 },
    { id: 2, name: "Mechanical Keyboard", price: 189000, category: "Accessory", stock: 30 },
    { id: 3, name: "4K Monitor", price: 620000, category: "Display", stock: 8 },
    { id: 4, name: "Wireless Mouse", price: 99000, category: "Accessory", stock: 24 }
];

app.use("/static", express.static(path.join(__dirname, "static")));

app.get("/", (req, res) => {
    res.sendFile(path.join(__dirname, "templates", "index.html"));
});

app.get("/api/products", (req, res) => {
    setTimeout(() => {
        res.json(products);
    }, 700);
});

app.listen(PORT, () => {
    console.log(`server running on http://localhost:${PORT}`);
});
```
