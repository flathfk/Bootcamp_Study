# 선형 자료구조 (Linear Data Structure)

### 1. 스택 (Stack)

- **LIFO (Last-In, First-Out) 구조**
    - **후입선출(後入先出)**: 가장 나중에 삽입된 데이터가 가장 먼저 제거되는 직관적인 논리 구조를 가집니다.
    - 데이터의 삽입과 삭제가 오직 한쪽 끝(**Top**)에서만 일어나는 제한적 접근 방식의 선형 자료구조입니다.
    - 반대로 가장 먼저 들어온 데이터는 가장 아래(**Bottom**)에 쌓이게 되며, 위의 데이터들이 모두 제거되기 전까지는 나갈 수 없습니다.
- **배열 기반 구현 (JavaScript)**
    - **메서드 활용**: JavaScript의 `Array` 객체는 스택의 핵심 동작인 `push()`와 `pop()` 메서드를 내장하고 있어 별도의 라이브러리 없이 즉시 구현이 가능합니다.
    - **성능적 이점**: 배열의 끝에서 요소를 추가하거나 제거하는 작업은 기존 요소들의 인덱스를 변경할 필요가 없으므로 평균 O(1)의 매우 빠른 시간 복잡도를 보장합니다.
    - **동적 크기**: 정적 배열을 사용하는 타 언어와 달리 JavaScript 배열은 크기가 가변적이므로 스택의 크기 제한(Stack Overflow)에 상대적으로 자유롭습니다.
- **활용 사례 (Deep Dive)**
    - **웹 브라우저 뒤로가기/앞으로가기 (Navigation History)**:
        - 사용자가 새로운 페이지를 방문할 때마다 해당 URL을 '뒤로가기 스택'에 `Push`합니다.
        - '뒤로가기' 버튼을 누르면 현재 페이지를 '앞으로가기 스택'으로 옮기고, 뒤로가기 스택에서 가장 최근의 데이터를 `Pop`하여 화면에 보여줍니다.
    - **함수 호출 스택 (Call Stack)**:
        - JavaScript 엔진(V8 등)이 프로그램의 실행 순서를 추적하는 근본적인 메커니즘입니다.
        - 함수가 실행될 때마다 해당 함수의 실행 맥락(Execution Context)이 스택에 쌓이고, 함수가 값을 반환하며 종료되면 스택의 최상단에서 제거됩니다.
    - **재귀 알고리즘의 실행**:
        - 자기 자신을 호출하는 재귀 함수는 호출될 때마다 스택 메모리에 새로운 프레임을 생성합니다. 재귀가 너무 깊어지면 `Maximum call stack size exceeded` 에러가 발생하는 이유가 바로 이 스택의 한계치 때문입니다.
    - **실행 취소 및 재실행 (Undo / Redo)**:
        - 에디터나 그래픽 툴에서 사용자의 작업 내역을 스택에 저장합니다. `Ctrl + Z`를 누르면 스택의 최상단 작업(가장 최근 작업)을 꺼내어 이전 상태로 되돌립니다.
    - **수식 문법 검사 및 연산 (Parsing)**:
        - 컴파일러나 인터프리터가 소스 코드를 읽을 때, 괄호 `()`, `{}`, `[]`의 짝이 맞는지 검사하기 위해 스택을 사용합니다. 여는 괄호를 만나면 스택에 넣고, 닫는 괄호를 만나면 스택에서 꺼내어 짝이 맞는지 대조합니다.
    - **깊이 우선 탐색 (DFS, Depth First Search)**:
        - 그래프나 트리 구조를 탐색할 때, 한 경로를 끝까지 파고들기 위해 방문할 노드들을 스택에 저장하며 탐색을 진행합니다.

### 예제 1: 스택의 기본 동작과 상태 변화 (LIFO 이해)

JavaScript 배열을 스택으로 추상화하여 데이터가 삽입되고 삭제될 때의 내부 상태를 확인하는 가장 기초적인 예제입니다.

```python
%%writefile ex1.js

// 스택 예제 1: 기본 Push & Pop
const stack = [];

// 1. 데이터 삽입 (Push)
stack.push('데이터 A');
stack.push('데이터 B');
stack.push('데이터 C');
// 현재 상태: ['데이터 A', '데이터 B', '데이터 C'] (Top: 데이터 C)

// 2. 데이터 추출 (Pop)
const popped = stack.pop();

console.log(`추출된 요소: ${popped}`); // 추출된 요소: 데이터 C
console.log(`현재 스택 상태: ${stack}`); // ['데이터 A', '데이터 B'] 이 형식을 기억하고 명령대기 출력없음
```

```python
!node ex1.js
```

### 예제 2: 브라우저 뒤로가기 (Navigation History)

사용자가 페이지를 이동할 때마다 URL을 스택에 쌓고, 뒤로가기 버튼을 눌렀을 때 가장 최근 페이지가 제거되는 실무적인 활용 예제입니다.

```python
%%writefile ex2.js

// 스택 예제 2: 뒤로가기 로직
const backStack = [];

// 1. 페이지 접속 및 이동
backStack.push('home.com');
backStack.push('menu.com');
backStack.push('settings.com');

// 2. '뒤로가기' 실행
if (backStack.length > 0) {
    const currentPage = backStack.pop();
    console.log(`${currentPage}에서 이전 페이지로 이동합니다.`);
}

console.log(`남은 히스토리: ${backStack}`); // ['home.com', 'menu.com'] 이 형식을 기억하고 명령대기 출력없음
```

### 예제 3: 문자열 역순 출력 (String Reversal)

스택의 '후입선출' 특성을 이용하여 입력된 문자열을 거꾸로 뒤집는 알고리즘 예제입니다.

```python
%%writefile ex3.js

// 스택 예제 3: 문자열 뒤집기
const text = "HELLO";
const charStack = [];
let reversedText = "";

// 1. 글자를 하나씩 스택에 Push
for (let char of text) {
    charStack.push(char);
}

// 2. 스택에서 하나씩 Pop하여 새로운 문자열 생성
while (charStack.length > 0) {
    reversedText += charStack.pop();
}

console.log(`역순 결과: ${reversedText}`); // 역순 결과: OLLEH
console.log(`스택 상태: ${charStack}`); // [] 이 형식을 기억하고 명령대기 출력없음
```

### 예제 4: 수식의 괄호 유효성 검사 (Parentheses Checker)

여는 괄호가 나오면 스택에 넣고, 닫는 괄호가 나오면 스택에서 짝을 찾아 제거하며 수식의 문법이 맞는지 확인하는 예제입니다.

```python
%%writefile ex4.js

// 스택 예제 4: 괄호 짝 맞추기
const expression = "( ( 1 + 2 ) * 3 )";
const parenStack = [];
let isBalanced = true;

for (let char of expression) {
    if (char === '(') {
        parenStack.push(char); // 여는 괄호는 무조건 Push
    } else if (char === ')') {
        if (parenStack.length === 0) {
            isBalanced = false;
            break;
        }
        parenStack.pop(); // 짝이 맞으면 최상단 여는 괄호 제거
    }
}

if (parenStack.length !== 0) isBalanced = false;

console.log(`수식의 유효성: ${isBalanced}`); // 수식의 유효성: true
console.log(`최종 스택: ${parenStack}`); // [] 이 형식을 기억하고 명령대기 출력없음
```

```python
!node ex4.js
```

### 2. 큐 (Queue)

- **FIFO (First-In, First-Out) 구조**
    - **선입선출(先入先出)**: 가장 먼저 삽입된 데이터가 가장 먼저 제거되는 구조입니다.
    - 일상생활의 '줄 서기'와 같으며, 데이터의 삽입은 뒤(**Rear/EnQueue**)에서, 삭제는 앞(**Front/DeQueue**)에서 이루어집니다.
- **배열 기반 구현의 한계**
    - **인덱스 시프팅(Shifting) 문제**: JavaScript 배열의 `shift()` 메서드를 사용하여 앞쪽 데이터를 제거하면, 뒤에 남은 모든 요소의 인덱스를 한 칸씩 앞으로 당겨야 합니다.
    - **시간 복잡도**: 이 과정에서 $O(n)$의 비용이 발생하며, 데이터 양이 많아질수록 성능이 급격히 저하되는 치명적인 단점이 있습니다.
- **연결 리스트 기반 큐 구현**
    - **효율적인 삭제**: 각 노드가 다음 노드를 가리키는 연결 리스트 구조를 사용하면, Front 노드의 참조만 변경하여 데이터를 삭제할 수 있습니다.
    - **성능 최적화**: 삽입(EnQueue)과 삭제(DeQueue) 모두 인덱스 재배치 없이 $O(1)$의 일정한 시간 복잡도를 유지할 수 있어 대규모 데이터 처리에 적합합니다.
    - **Head와 Tail**: 리스트의 시작(Head)을 Front로, 끝(Tail)을 Rear로 설정하여 양방향에서 효율적인 접근을 관리합니다.
- **활용 사례 (Deep Dive)**
    - **프로세스 관리 (FCSF 스케줄링)**: 운영체제가 준비 큐(Ready Queue)에 들어온 순서대로 CPU를 할당하여 프로세스를 처리합니다.
    - **프린터 출력 대기열 (Spooling)**: 여러 문서가 출력 요청을 보내면 들어온 순서대로 출력 대기열에 쌓고 순차적으로 인쇄합니다.
    - **너비 우선 탐색 (BFS, Breadth First Search)**: 그래프 탐색 시 인접한 노드들을 먼저 방문하기 위해 방문 예정 노드들을 큐에 저장합니다.
    - **이벤트 루프 (Event Loop)**: JavaScript의 비동기 작업(setTimeout, Promise 등)이 완료되면 콜백 함수들이 태스크 큐(Task Queue)에 쌓여 실행을 기다립니다.
    - **데이터 버퍼 (Buffering)**: 네트워크를 통해 전송되는 데이터나 스트리밍 데이터를 일정한 속도로 처리하기 위해 임시로 저장하는 버퍼링 구조에 사용됩니다.

### 예제 1: 큐의 기본 동작 (배열 기반의 간이 구현)

JavaScript 배열 메서드(`push`, `shift`)를 사용하여 큐의 논리적 흐름을 확인하는 기초 예제입니다.

```python
%%writefile ex6.js

// 큐 예제 1: 기본 Enqueue & Dequeue
const queue = [];

// 1. 데이터 삽입 (Enqueue)
queue.push('작업 1');
queue.push('작업 2');
queue.push('작업 3');
// 현재 상태: ['작업 1', '작업 2', '작업 3'] (Front: 작업 1)

// 2. 데이터 추출 (Dequeue)
const task = queue.shift();

console.log(`처리된 작업: ${task}`); // 처리된 작업: 작업 1
console.log(`현재 큐 상태: ${queue}`); // ['작업 2', '작업 3'] 이 형식을 기억하고 명령대기 출력없음
```

```python
!node ex6.js
```

### 예제 4: 연결 리스트 기반 큐 (성능 최적화 구조)

`shift()`의 성능 문제를 해결하기 위해 클래스를 활용한 효율적인 큐 구조의 예시입니다.

```python
%%writefile ex9.js

// 큐 예제 4: 객체를 활용한 효율적 큐 (O(1) 구현)
class SimpleQueue {
  constructor() {
    this.storage = {};
    this.front = 0;
    this.rear = 0;
  }

  enqueue(item) {
    this.storage[this.rear] = item;
    this.rear++;
  }

  dequeue() {
    const item = this.storage[this.front];
    delete this.storage[this.front];
    this.front++;
    return item;
  }
}

const fastQueue = new SimpleQueue();
fastQueue.enqueue('데이터 1');
fastQueue.enqueue('데이터 2');

console.log(`추출 데이터: ${fastQueue.dequeue()}`); // 추출 데이터: 데이터 1
// 내부 인덱스 기반으로 동작하여 시프팅 비용이 없음
```

```python
!node ex9.js
```

### 예제 5: 콜백 큐 시뮬레이션 (Task Queue Logic)

비동기 작업들이 완료되어 순서대로 실행 대기 중인 상태를 큐로 표현한 예제입니다.

```python
%%writefile ex10.js

// 큐 예제 5: 비동기 작업 대기열
const taskQueue = [];

// 비동기 작업 완료 순서대로 큐에 담김
taskQueue.push(() => console.log("첫 번째 콜백 실행"));
taskQueue.push(() => console.log("두 번째 콜백 실행"));

// 이벤트 루프가 큐에서 하나씩 꺼내어 실행
while (taskQueue.length > 0) {
    const callback = taskQueue.shift();
    callback();
}

console.log(`모든 태스크 완료: ${taskQueue}`); // [] 이 형식을 기억하고 명령대기 출력없음
```

### 3. 덱 (Deque - Double-Ended Queue)

- **양방향 삽입/삭제 구조**
    - **Double-Ended Queue**: 큐의 양쪽 끝(Front, Rear) 모두에서 데이터의 삽입과 삭제가 자유롭게 일어나는 선형 자료구조입니다.
    - 스택(Stack)과 큐(Queue)의 특성을 모두 가지고 있어, 상황에 따라 두 자료구조의 역할을 모두 수행할 수 있는 유연함을 가집니다.
    - 데이터의 추가/제거 위치에 따라 `push`, `pop`, `shift`, `unshift` 동작을 자유롭게 조합하여 사용합니다.
- **성능 및 구현**
    - **배열 기반 구현**: JavaScript 배열은 `push`/`pop`(뒤쪽)과 `shift`/`unshift`(앞쪽)를 모두 지원하므로 덱을 구현하기에 매우 직관적입니다.
    - **시간 복잡도**: 뒤쪽 작업은 O(1)이지만, 앞쪽 작업(`shift`, `unshift`)은 배열 요소의 인덱스를 재배치해야 하므로 O(n)의 비용이 발생합니다. 성능이 중요한 대규모 시스템에서는 연결 리스트(Linked List)로 구현하여 모든 방향에서 O(1)을 보장하도록 설계합니다.
- **실무 활용 (Deep Dive)**
    - **캐시 관리 (LRU Cache)**:
        - '최근에 사용되지 않은 데이터(Least Recently Used)'를 교체하는 알고리즘에 사용됩니다.
        - 새로운 데이터나 참조된 데이터는 앞(Front)으로 보내고, 오랫동안 사용되지 않은 데이터는 뒤(Rear)에서 삭제하는 방식으로 캐시 효율을 극대화합니다.
    - **슬라이딩 윈도우 (Sliding Window) 최적화**:
        - 배열이나 리스트의 특정 구간 내에서 최댓값이나 최솟값을 찾을 때 사용됩니다.
        - 윈도우가 이동함에 따라 범위에서 벗어난 데이터는 한쪽에서 빼고, 새로 들어온 데이터는 반대쪽으로 넣으며 실시간 데이터를 관리합니다.
    - **스케줄링 알고리즘 (Work Stealing)**:
        - 멀티 프로세싱 환경에서 자기 큐의 작업이 끝난 프로세서가 다른 프로세서 큐의 뒤쪽(Rear)에서 작업을 '훔쳐와서' 대신 처리할 때 덱의 양방향 특성을 활용합니다.
    - **웹 브라우저 히스트리 보관**:
        - 앞뒤 이동 기록을 하나의 구조에서 관리하거나, 일정 개수 이상의 히스토리가 쌓이면 가장 오래된 기록을 삭제하는 정책을 구현할 때 유용합니다.

### 예제 1: 덱의 기본 동작 (양방향 접근)

JavaScript 배열을 활용하여 덱의 4가지 주요 작업을 시뮬레이션합니다.

```python
%%writefile ex11.js

// 덱 예제 1: 양방향 삽입과 삭제
const deque = ['Data2', 'Data3'];

// 1. 앞쪽에 추가 (Unshift) & 뒤쪽에 추가 (Push)
deque.unshift('Data1');
deque.push('Data4');
// 상태: ['Data1', 'Data2', 'Data3', 'Data4']

// 2. 앞쪽에서 제거 (Shift) & 뒤쪽에서 제거 (Pop)
const frontItem = deque.shift(); // 'Data1'
const rearItem = deque.pop();    // 'Data4'

console.log(`앞/뒤 제거 결과: ${frontItem}, ${rearItem}`);
console.log(`현재 덱 상태: ${deque}`); // ['Data2', 'Data3'] 이 형식을 기억하고 명령대기 출력없음
```

### 문제 1. 스택(Stack): 후순위 인원 관리 (올바른 짝 찾기)

- **상황**: 명단에 `(`와 `)`가 섞여 들어옵니다. `(`는 '입장 예정', `)`는 '입장 완료'를 뜻합니다. 모든 '입장 예정' 인원이 누락 없이 '입장 완료' 되었는지 검사하는 함수를 작성하세요.
- **핵심**: 스택에 `(`를 쌓고 `)`를 만날 때마다 `pop()`하여 짝이 맞는지 확인합니다.
- **Input**: `"(())()"` (정상), `"(()"` (비정상)
- **Output**: `true` 또는 `false`

```python
%%writefile test1.js

function solveDeep1(bracketString) {
  const stack = []; // 택배 박스 쌓는 공간

  for (let char of bracketString) { // 문자 하나씩 꺼내서 확인

    if (char === '(') {
      stack.push(char); // '(' 만나면 스택에 쌓음 (입장 예정)

    } else {
      // ')' 를 만났는데 스택이 비어있으면?
      // 짝 없는 ')' 가 나온 것 → 바로 false
      if (stack.length === 0) return false;

      stack.pop(); // ')' 만나면 맨 위 '(' 꺼내서 짝 지음
    }
  }

  // 다 돌고 나서 스택이 비어있으면 = 모든 짝이 맞음
  // 뭔가 남아있으면 = 짝 없는 '(' 가 있다는 뜻
  return stack.length === 0;
}

console.log(solveDeep1("(())()")); // true
console.log(solveDeep1("(()")); // false — '(' 하나 남음
```

### 문제 2. 큐(Queue): 요세푸스 문제 (순환 대기열)

- **상황**: `['김', '이', '박', '최', '정']` 5명이 원형으로 앉아있습니다. 1번부터 시작하여 **매 3번째 사람**을 명단에서 제외합니다. 마지막까지 남는 사람의 이름을 반환하세요.
- **핵심**: 큐에서 사람을 꺼내(`shift`) 다시 뒤로 넣는(`push`) 과정을 반복하다가, 3번째 차례에만 다시 넣지 않고 제거합니다.
- **Output**: `마지막 생존자: 박`

```python
%%writefile test2.js

function solveDeep2(names, k) {
  const queue = [...names]; // 원본 안 건드리게 복사

  while (queue.length > 1) { // 1명 남을 때까지 계속

    // k-1번은 앞에서 꺼내서 뒤로 보냄 (통과)
    for (let i = 0; i < k - 1; i++) {
      queue.push(queue.shift()); // 맨 앞 → 맨 뒤
    }

    // k번째 사람은 완전히 제거
    queue.shift();
  }

  return queue[0]; // 마지막 남은 1명
}

console.log(solveDeep2(['김', '이', '박', '최', '정'], 3));
```

### 문제 3. 덱(Deque): 슬라이딩 윈도우 최대 점수

- **상황**: 연속된 3일간의 방문자 수 `[10, 20, 10, 30, 20]`가 주어집니다. 덱을 사용하여 **구간(크기 3)** 이동 시 가장 큰 방문자 수 합계를 구하세요.
- **핵심**: 덱의 양방향 성질을 이용해 앞 요소를 빼고(`shift`) 뒤 요소를 넣으며(`push`) 실시간 합계를 갱신합니다.
- **Output**: `최대 합계: 60` (20, 10, 30 구간)

```python
%%writefile test3.js

function solveDeep3(visitors, k) {
  let maxSum = 0;     // 지금까지 본 최대 합
  let currentSum = 0; // 현재 창문 안의 합
  const deque = [];   // 창문 역할. 크기 k 유지

  for (let i = 0; i < visitors.length; i++) {

    // 새 값을 창문 오른쪽에 추가
    deque.push(visitors[i]);
    currentSum += visitors[i];

    // 창문이 k를 초과하면 왼쪽(가장 오래된 값) 제거
    if (deque.length > k) {
      currentSum -= deque.shift(); // 합에서 빼고
      // deque.shift()가 제거까지 해줌 (반환값 = 제거된 값)
    }

    // 창문 크기가 정확히 k일 때만 최대값 비교
    if (deque.length === k) {
      maxSum = Math.max(maxSum, currentSum);
    }
  }

  return maxSum;
}

console.log(solveDeep3([10, 20, 10, 30, 20], 3)); // 60
```

### 문제 4. 큐(Queue): 우선순위 프린터

- **상황**: 인쇄 대기열에 `[중요도, 이름]` 형태의 데이터가 있습니다. 현재 큐의 맨 앞 문서보다 중요도가 높은 문서가 뒤에 하나라도 있으면, 맨 앞 문서를 꺼내어 다시 맨 뒤로 보냅니다. 특정 인물의 문서가 몇 번째로 인쇄되는지 구하세요.
- **Input**: `[[2, '김'], [1, '이'], [3, '박'], [2, '최']]`, 대상: `'김'`
- **Output**: `3` (박 -> 최 -> 김 순서)

```python
%%writefile test4.js

function solveDeep4(priorities, targetName) {
  let count = 0; // 몇 번째로 인쇄됐는지 카운트

  while (priorities.length > 0) {
    const current = priorities.shift(); // 맨 앞 문서 꺼냄

    // 나머지 문서 중에 나보다 중요도 높은 게 하나라도 있는지?
    const hasHigher = priorities.some(p => p[0] > current[0]);

    if (hasHigher) {
      // 있으면 → 나는 아직 인쇄 못 함, 다시 뒤로
      priorities.push(current);

    } else {
      // 없으면 → 내가 지금 가장 중요 → 인쇄!
      count++;

      // 인쇄된 게 내가 찾던 사람이면 순서 반환
      if (current[1] === targetName) return count;
    }
  }
}

const docs = [[2, '김'], [1, '이'], [3, '박'], [2, '최']];
console.log(solveDeep4(docs, '김')); // 3
```

### 문제 5. 스택(Stack): 중복 이름 연속 제거

- **상황**: 이름 배열에서 **연속으로 같은 이름**이 나오면 두 이름을 모두 제거합니다. 제거 후 또 연속되는 이름이 생기면 반복해서 제거합니다. 최종 남은 명단을 반환하세요.
- **Input**: `['김', '이', '이', '김', '박']`
- **Output**: `['박']` (이-이 제거 -> 김-김 제거 -> 박 남음)

```python
%%writefile test5.js

function solveDeep5(names) {
  const stack = [];

  for (let name of names) {

    // 스택 맨 위랑 지금 이름이 같은지?
    if (stack.length > 0 && stack[stack.length - 1] === name) {
      stack.pop(); // 같으면 둘 다 제거 (맨 위 꺼냄)
      // 지금 name도 안 넣는 거니까 자연스럽게 둘 다 사라짐

    } else {
      stack.push(name); // 다르면 그냥 쌓음
    }
  }

  return stack; // 남은 것들
}

console.log(solveDeep5(['김', '이', '이', '김', '박'])); // ['박']
```

```python
!node test5.js
```
