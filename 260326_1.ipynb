## 연결 자료구조

### **연결 리스트 (Linked List) Deep Dive**

- **추상 자료형(ADT)의 물리적 구현:** 논리적으로는 일렬로 나열된 선형 구조이지만, 실제 메모리(RAM)상에서는 여기저기 흩어져 있는 노드들을 '참조(Reference)'로 엮어 만든 동적 자료구조입니다.
- **노드의 구성 요소:**
    - **Data:** 저장하려는 실제 값 (숫자, 문자열, 객체 등).
    - **Pointer (Next):** 다음 노드의 메모리 주소를 가리키는 이정표. 마지막 노드의 포인터는 `null`을 가리켜 리스트의 끝임을 알립니다.
- **동적 확장성:** 배열처럼 미리 크기를 예약(Capacity)할 필요가 없어, 메모리가 허용하는 한 무한히 확장 가능하며 '메모리 단편화' 문제에서 상대적으로 자유롭습니다.

### **유형별 상세 분석 및 JavaScript 구현**

### **1) 단일 연결 리스트 (Singly Linked List)**

- **특징:** 한 방향으로만 흐르는 편도 통행 구조입니다.
- **구현 핵심:** 각 노드는 `next` 속성만 가집니다. `Head` 노드를 놓치면 전체 리스트를 유실하므로 관리가 엄격해야 합니다.
- **장점:** 메모리 오버헤드가 적고 구현이 단순합니다.
- **단점:** 특정 노드의 이전 노드를 찾으려면 다시 `Head`부터 탐색해야 하는 비효율이 발생합니다 ($O(n)$).

### **2) 이중 연결 리스트 (Doubly Linked List)**

- **특징:** 양방향 통행이 가능한 구조입니다.
- **구현 핵심:** 노드가 `prev`와 `next` 두 개의 포인터를 가집니다.
- **장점:** 특정 노드 삭제 시 이전 노드를 즉시 알 수 있어 삭제 작업이 매우 유연하며, 뒤에서부터 역방향 탐색이 가능합니다.
- **단점:** 포인터 변수를 하나 더 저장해야 하므로 노드당 메모리 사용량이 약 1.5~2배 증가하며, 삽입/삭제 시 변경해야 할 포인터가 많아 코드가 복잡해집니다.

### **배열(Array) vs 연결 리스트(Linked List) 성능 분석**

| **비교 항목** | **배열 (Array / Vector)** | **연결 리스트 (Linked List)** |
| --- | --- | --- |
| **조회 (Access)** | **O(1)** (인덱스로 즉시 접근) | **O(n)** (처음부터 찾아야 함) |
| **맨 앞 삽입/삭제** | **O(n)** (전체 데이터 이동 필요) | **O(1)** (Head 포인터만 변경) |
| **맨 뒤 삽입/삭제** | **O(1)** (빈 공간이 있을 때) | **O(n)** (Tail까지 가야 함) / **O(1)** (Tail 보유 시) |
| **중간 삽입/삭제** | **O(n)** (데이터 밀어내기) | **O(1)** (위치를 알 때 포인터만 교체) |
| **메모리 할당** | **정적/연속적** (컴파일/할당 시 결정) | **동적/비연속적** (런타임에 실시간 할당) |
| **캐시 효율** | **높음** (Locality가 좋아 성능 우수) | **낮음** (데이터가 파편화되어 있음) |

### **활용 및 주의사항**

- **JavaScript의 배열은 특별하다:** 사실 JavaScript의 `Array`는 내부적으로 해시 테이블이나 연결 리스트의 특성을 결합하여 최적화되어 있습니다. 따라서 아주 정밀한 성능 최적화가 필요한 알고리즘(예: 대규모 큐 구현)이 아니라면 표준 `Array`를 사용하는 것이 생산성 면에서 유리합니다.
- **가비지 컬렉션:** 연결 리스트에서 노드를 삭제할 때, 해당 노드를 가리키는 참조를 모두 끊어주어야 JavaScript의 가비지 컬렉터(GC)가 메모리를 정상적으로 회수합니다.

2) 맨 앞에 노드 삽입 (prepend) - O(1)

```python
%%writefile ex2.js

function createNode(data) {
  return { data: data, next: null };
}

// 초기 상태: 빈 리스트
const list = { head: null, size: 0 };

function prepend(list, data) {
  const newNode = createNode(data);
  newNode.next = list.head;
  list.head = newNode;
  list.size++;

  console.log(`[prepend] '${data}' 삽입 후 리스트 상태:`, JSON.stringify(list, null, 2));
}

prepend(list, "데이터1");
prepend(list, "데이터2");
```

4) 특정 값 삭제 (remove) - O(n)

```python
%%writefile ex4.js

function createNode(data) {
  return { data: data, next: null };
}

// 초기 상태 세팅: 데이터2 → 데이터1 → 데이터3
const list = { head: null, size: 3 };
const node2 = createNode('데이터2');
const node1 = createNode('데이터1');
const node3 = createNode('데이터3');
node2.next = node1;
node1.next = node3;
list.head = node2;

function remove(list, data) {
  if (!list.head) return;

  if (list.head.data === data) {
    list.head = list.head.next;
    list.size--;
    console.log(`[remove] Head 노드 '${data}' 삭제 후 리스트 상태:`, JSON.stringify(list, null, 2));
    return;
  }

  let curr = list.head;
  while (curr.next && curr.next.data !== data) { curr = curr.next; }

  if (curr.next) {
    curr.next = curr.next.next;
    list.size--;
    console.log(`[remove] 중간/끝 노드 '${data}' 삭제 후 리스트 상태:`, JSON.stringify(list, null, 2));
  } else {
    console.log(`[remove 실패] '${data}'를 찾을 수 없습니다.`);
  }
}

remove(list, "데이터1");
```

### **이중 연결 리스트 (Doubly Linked List)**

- **5) 이중 연결 노드 생성**

```python
%%writefile ex5.js

const list = { head: null, tail: null, size: 0 };

const createDoublyNode = (data) => ({
  data,
  prev: null,
  next: null
});

// prev는 순환 참조 발생하므로 '[Ref]'로 대체해서 출력
const replacer = (key, val) => key === 'prev' ? '[Ref]' : val;

function prependDoubly(list, data) {
  const newNode = createDoublyNode(data);

  if (!list.head) {
    list.head = newNode;
    list.tail = newNode;
  } else {
    newNode.next = list.head;
    list.head.prev = newNode;
    list.head = newNode;
  }
  list.size++;

  console.log(`[Doubly-Prepend] '${data}' 삽입 후 상태:`, JSON.stringify(list, replacer, 2));
}

prependDoubly(list, "이중데이터1");
prependDoubly(list, "이중데이터2");
prependDoubly(list, "이중데이터3");
```

6) 중간 삽입 로직 (포인터 4개 조작)

```python
%%writefile ex6.js

const createDoublyNode = (data) => ({
  data,
  prev: null,
  next: null
});

// nodeA <-> nodeB 초기 연결 상태 세팅
const nodeA = createDoublyNode("데이터1");
const nodeB = createDoublyNode("데이터2");
nodeA.next = nodeB;  // A → B
nodeB.prev = nodeA;  // A ← B

function insertBetween(nodeA, nodeB, data) {
  const newNode = createDoublyNode(data);

  newNode.next = nodeB;   // 1. 새 노드의 next → B
  newNode.prev = nodeA;   // 2. 새 노드의 prev → A
  nodeA.next = newNode;   // 3. A의 next → 새 노드
  nodeB.prev = newNode;   // 4. B의 prev → 새 노드

  console.log(`[InsertBetween] '${nodeA.data}'와 '${nodeB.data}' 사이에 '${data}' 삽입 완료`);
  console.log(`결과 구조: ${nodeA.data} <-> ${newNode.data} <-> ${nodeB.data}`);
}

insertBetween(nodeA, nodeB, "NEW");
```

8) 연결 리스트의 맨 앞 삽입 성능 (O(1))         

```python
%%writefile ex8.js

// [단계 1] 노드 생성 함수 정의
const createNode = (data) => ({
  data,
  next: null
});

// [단계 2] 리스트 초기 상태 설정
let list = {
  head: null,
  size: 0
};

// [단계 3] 초기 상태 출력 확인
console.log("--- [초기화 상태] ---");
console.log("리스트 객체:", list);
console.log("Head 연결 여부:", list.head === null ? "비어 있음" : "연결됨");
console.log("현재 크기:", list.size);

// [실행 전] 초기 상태: { head: null, size: 0 }

function prepend(list, data) {
  const newNode = createNode(data);
  newNode.next = list.head; // 새 노드가 기존 head를 가리킴
  list.head = newNode;      // head를 새 노드로 교체
  list.size++;

  // console.log(`[prepend] '${data}' 삽입 후 리스트 상태:`, JSON.stringify(list, null));
}


// [연결 리스트의 성능 측정] 데이터 10만 개를 맨 앞에 삽입할 때

let myList = { head: null };

console.time('List prepend'); // 타이머 시작

for (let i = 0; i < 100000; i++) {
  prepend(myList, i); // 참조값 1개(head)만 바꾸고 끝냄 (O(1))
}

console.timeEnd('List prepend'); // 타이머 종료 및 소요 시간 출력

/* 출력 예시 (시스템 환경에 따라 다름):
List prepend: 3.456ms
- 설명: 연결 리스트는 데이터가 10만 개든 1억 개든,
  단순히 새 노드의 next를 기존 head에 연결하고 head를 교체하는 작업만 수행하므로
  배열(unshift)에 비해 압도적으로 빠른 성능을 보여줍니다.
*/
```

### **실전 응용 및 기타**

- **9) 연결 리스트 기반 큐(Queue)의 Dequeue**

```python
%%writefile /content/ex9.js

const createNode = (data) => ({ data, next: null });

// 초기 상태 세팅: 데이터2 → 데이터3
const list = { head: null, size: 2 };
const node2 = createNode('데이터2');
const node3 = createNode('데이터3');
node2.next = node3;
list.head = node2;

function dequeue(list) {
  if (!list.head) return null;

  const removedData = list.head.data;
  list.head = list.head.next;  // head를 다음 노드로 교체
  list.size--;

  console.log(`[dequeue] '${removedData}' 추출 완료. 새로운 Head: ${list.head ? list.head.data : 'null'}`);
  console.log('최종 리스트 상태:', JSON.stringify(list, null, 2));
  return removedData;
}

dequeue(list);
```

10) 순환(Circular) 구조 만들기       

```python
%%writefile ex10.js

const createNode = (data) => ({ data, next: null });

// 초기 상태: 데이터3 하나만 있는 리스트
const list = { head: null, size: 1 };
const node3 = createNode('데이터3');
list.head = node3;

function makeCircular(list) {
  if (!list.head) return;

  let last = list.head;

  // 1. 마지막 노드까지 이동
  while (last.next) {
    last = last.next;
  }

  // 2. 마지막 노드의 next → head로 연결 (순환 고리 형성)
  last.next = list.head;

  console.log(`[Circular] 마지막 노드 '${last.data}'의 next를 Head '${list.head.data}'로 연결 완료`);
  console.log(`확인: list.head.next... (무한 반복 가능 구조)`);
}

makeCircular(list);
```

## 연결 리스트 구현

### **Node 구조 설계**

- **데이터와 참조의 결합:** 연결 리스트의 최소 단위인 노드(Node)는 실제 저장할 값인 `data`와 다음 노드의 위치를 가리키는 참조 변수인 `next`로 구성됩니다.
- **객체 리터럴 활용:** 클래스를 사용하지 않을 경우, JavaScript의 객체 `{}`를 사용하여 동적으로 노드를 생성합니다.
- **추상화:** `next`가 `null`인 상태는 해당 노드가 리스트의 끝(Tail)임을 의미하며, 리스트 전체의 시작점은 별도의 `head` 변수에 저장하여 관리합니다.

### **주요 동작 구현**

### **1) 삽입 (Insertion)**

- **맨 앞 삽입 (O(1)):** 새로운 노드를 생성하고, 새 노드의 `next`가 현재의 `head`를 가리키게 한 뒤, `head`를 새 노드로 교체합니다.
- **중간/끝 삽입 (O(n)):** 삽입하려는 위치의 바로 앞 노드(Previous)까지 탐색한 후, `newNode.next = prev.next`와 `prev.next = newNode`를 통해 연결 고리를 재구성합니다.
- **특징:** 배열과 달리 기존 요소들을 한 칸씩 뒤로 밀어내는 데이터 이동(Shift) 과정이 전혀 없습니다.

### **2) 삭제 (Deletion)**

- **노드 건너뛰기 (O(n)):** 삭제할 노드의 이전 노드(`prev`)를 찾아, `prev.next`가 삭제할 노드의 다음 노드(`target.next`)를 가리키도록 변경합니다.
- **메모리 해제:** 참조가 끊긴 노드는 JavaScript의 가비지 컬렉터(GC)에 의해 자동으로 메모리에서 제거됩니다.
- **특징:** 중간 요소를 삭제해도 뒤에 있는 데이터들의 인덱스를 수정할 필요가 없어 삽입과 마찬가지로 연결 수정만으로 완료됩니다.

### **3) 탐색 (Search)**

- **순차 탐색 (O(n)):** 특정 데이터를 찾기 위해 `head`부터 시작하여 `next` 포인터를 따라 원하는 값을 만날 때까지 노드를 하나씩 방문합니다.
- **한계:** 배열처럼 인덱스를 이용한 즉시 접근(Random Access)이 불가능하며, 최악의 경우 리스트 전체를 순회해야 합니다.

### **코드 예시**

### **1) Node 구조 설계 및 리스트 초기화**

데이터와 다음 연결을 위한 기본 틀을 정의합니다.

```python
%%writefile nodeEx1.js

// 노드 생성 함수: 데이터와 next 포인터를 가진 객체 반환
const createNode = (data) => ({
  data: data,
  next: null
});

// 연결 리스트 초기 상태 (Head가 비어있음)
let myLinkedList = { head: null, size: 0 };

console.log("--- 리스트 초기화 완료 ---");
console.log(myLinkedList);
```

```python
!node nodeEx1;
```

```python
%%writefile nodeEx2.js

// 맨 앞에 삽입 (Prepend): O(1)
function insertAtFront(list, data) {
  const newNode = createNode(data);
  newNode.next = list.head; // 새 노드가 기존 head를 가리킴
  list.head = newNode;      // head를 새 노드로 교체
  list.size++;
  console.log(`[삽입] 맨 앞에 ${data} 추가 완료`);
}

// 맨 뒤에 삽입 (Append): O(n)
function insertAtBack(list, data) {
  const newNode = createNode(data);
  if (!list.head) {
    list.head = newNode;
  } else {
    let curr = list.head;
    while (curr.next) { curr = curr.next; } // 끝까지 탐색
    curr.next = newNode;
  }
  list.size++;
  console.log(`[삽입] 맨 뒤에 ${data} 추가 완료`);
}

insertAtFront(myLinkedList, "노드1");
insertAtBack(myLinkedList, "노드2");
insertAtBack(myLinkedList, "노드3");
```

```python
%%writefile nodeEx3.js

function findNode(list, targetData) {
  let curr = list.head;
  let index = 0;

  while (curr) {
    if (curr.data === targetData) {
      console.log(`[탐색 성공] '${targetData}'를 인덱스 ${index}에서 찾았습니다.`);
      return curr;
    }
    curr = curr.next;
    index++;
  }
  console.log(`[탐색 실패] '${targetData}'가 리스트에 없습니다.`);
  return null;
}

findNode(myLinkedList, "노드2");
findNode(myLinkedList, "노드99"); // 실패 케이스
```

```python
%%writefile nodeEx4.js

function deleteNode(list, targetData) {
  if (!list.head) return;

  // 1. 삭제할 대상이 Head일 때
  if (list.head.data === targetData) {
    list.head = list.head.next;
    list.size--;
    console.log(`[삭제] Head 노드 '${targetData}' 삭제 완료`);
    return;
  }

  // 2. 삭제할 대상의 이전 노드(prev) 찾기
  let prev = list.head;
  while (prev.next && prev.next.data !== targetData) {
    prev = prev.next;
  }

  // 3. 연결 변경 (건너뛰기)
  if (prev.next) {
    prev.next = prev.next.next;
    list.size--;
    console.log(`[삭제] 중간 노드 '${targetData}' 삭제 완료`);
  } else {
    console.log(`[삭제 실패] '${targetData}'를 찾을 수 없습니다.`);
  }
}

deleteNode(myLinkedList, "노드2"); // 중간 삭제
```

```python
%%writefile nodeEx5.js

function printList(list) {
  let curr = list.head;
  let result = "Head -> ";

  while (curr) {
    result += `[${curr.data}] -> `;
    curr = curr.next;
  }

  result += "null";
  console.log("--- 현재 리스트 상태 ---");
  console.log(result);
  console.log(`총 노드 개수: ${list.size}`);
}

printList(myLinkedList);
```

### **문제 1. 단일 연결 리스트: 마지막 인원 확인 (Tail 탐색)**

- **상황:** 연결 리스트 형태로 줄을 서 있는 명단이 주어집니다. 각 노드는 다음 사람을 가리킵니다. 명단의 가장 마지막에 서 있는 사람의 이름을 찾아내는 함수를 작성하세요.
- **핵심:** `while` 문을 사용하여 `curr.next`가 `null`인 노드(마지막 노드)에 도달할 때까지 포인터를 이동시킵니다.
- **Input:** `head: { data: "철수", next: { data: "영희", next: { data: "민수", next: null } } }`
- **Output:** `"민수"`

```python
%%writefile quiz1.js

// 마지막 사람 찾기
function findLastPerson(head) {

  // 리스트가 비어있으면 null 반환
  if (!head) return null;

  // 지금 내가 보고 있는 사람 - 맨 앞부터 시작
  let currentPerson = head;

  // 다음 사람이 없을 때까지 계속 앞으로 이동
  while (currentPerson.next !== null) {
    currentPerson = currentPerson.next; // 한 칸 앞으로
  }

  // 여기까지 왔으면 currentPerson이 마지막 사람
  return currentPerson.data;
}

// 테스트: 철수 -> 영희 -> 민수 -> 끝
const list1 = {
  data: '철수',
  next: {
    data: '영희',
    next: {
      data: '민수',
      next: null
    }
  }
};

console.log(findLastPerson(list1)); // '민수'
```

### **문제 2. 단일 연결 리스트: 중간 새치기 (Node 삽입)**

- **상황:** `nodeA`와 `nodeB`가 연결되어 있습니다. 두 사람 사이에 새로운 사람(`newNode`)이 새치기를 하려고 합니다. 연결 관계를 수정하여 `nodeA -> newNode -> nodeB`가 되도록 만드세요.
- **핵심:** `newNode.next`에 기존의 `nodeA.next`를 먼저 연결한 뒤, `nodeA.next`를 `newNode`로 교체하여 연결 고리를 재구성합니다.
- **Input:** `nodeA` (앞사람), `newNode` (새치기객)
- **Output:** `nodeA.next.data`가 `newNode`의 데이터와 일치해야 함

```python
%%writefile quiz2.js

// 두 사람 사이에 새치기시키기
function cutInLine(nodeA, newNode) {

  // 1단계: 새치기할 사람의 다음을 먼저 기존 뒷사람으로 연결
  //        (이 순서 꼭 지켜야 함 - 안 그러면 뒷사람 위치를 잃어버림)
  newNode.next = nodeA.next;

  // 2단계: 앞사람의 다음을 새치기할 사람으로 교체
  nodeA.next = newNode;
}

// 초기 상태: 앞사람 -> 기존뒷사람
let nodeB   = { data: '기존뒷사람', next: null };
let nodeA   = { data: '앞사람', next: nodeB };
let newNode = { data: '새치기객', next: null };

// 새치기 실행
cutInLine(nodeA, newNode);
// 결과: 앞사람 -> 새치기객 -> 기존뒷사람

console.log(nodeA.next.data);       // '새치기객'
console.log(nodeA.next.next.data);  // '기존뒷사람'
```

### **문제 3. 단일 연결 리스트: 명단 합산 (전체 탐색)**

- **상황:** 각 노드에 숫자가 적힌 쪽지를 들고 사람들이 서 있습니다. `head`부터 시작하여 마지막 사람까지 이동하며 모든 쪽지에 적힌 숫자의 총합을 구하세요.
- **핵심:** `curr` 변수가 `null`이 될 때까지 반복하며 `curr.data`를 누적 변수에 더하고, `curr = curr.next`로 한 칸씩 전진합니다.
- **Input:** `10 -> 20 -> 30 -> null`
- **Output:** `60`

```python
%%writefile quiz3.js

// 모든 숫자 더하기
function addAllNumbers(head) {

  let totalSum    = 0;    // 지금까지 더한 합계
  let currentNode = head; // 지금 보고 있는 사람 - 맨 앞부터 시작

  // currentNode가 null이 될 때까지 (끝까지) 반복
  while (currentNode !== null) {
    totalSum += currentNode.data;   // 현재 사람의 숫자를 합계에 더함
    currentNode = currentNode.next; // 다음 사람으로 이동
  }

  return totalSum;
}

// 테스트: 10 -> 20 -> 30 -> 끝
const list3 = {
  data: 10,
  next: {
    data: 20,
    next: {
      data: 30,
      next: null
    }
  }
};

console.log(addAllNumbers(list3)); // 60
```
