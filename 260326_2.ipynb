# 원시 타입과 참조 타입의 메모리 할당, Deep Copy

### 1. 배경

자바스크립트 엔진이 데이터를 관리하는 방식은 효율적인 메모리 사용과 프로그램의 안정성에 직결됩니다.

- 메모리 공간의 분리: 고정된 크기의 데이터를 위한 스택(Stack)과 동적인 데이터를 위한 힙(Heap)으로 나뉩니다.
- 복사 메커니즘의 차이: 타입에 따라 값 자체가 복사되는지, 아니면 데이터가 저장된 위치(주소)가 복사되는지가 결정됩니다.

이 구조를 명확히 이해해야만 참조 타입 데이터를 다룰 때 발생하는 의도치 않은 원본 변경 사고를 방지할 수 있습니다.

### 2. 설명

### 2.1. 원시 타입(Primitive Types)과 Stack 할당

- 종류: Number, String, Boolean, null, undefined, Symbol, BigInt
- 동작: 변수 선언 시 스택 메모리 영역에 실제 값이 직접 저장됩니다.
- 복사: 변수를 대입하면 값 자체가 복제되므로, 복사본의 값을 변경해도 원본에는 아무런 영향을 주지 않습니다.

### 2.2. 참조 타입(Reference Types)과 Heap 할당

- 종류: Object, Array, Function, Map, Set 등
- 동작: 실제 데이터는 힙(Heap) 메모리에 저장되고, 스택(Stack)에는 해당 힙 메모리의 시작 주소(Reference)가 저장됩니다.
- 복사: 변수를 대입하면 주소값만 복제됩니다. 따라서 두 변수가 서로 다른 이름을 가졌더라도 메모리 상에서는 같은 데이터를 바라보게 됩니다.

### 2.3. Shallow Copy(얕은 복사) vs Deep Copy(깊은 복사)

- Shallow Copy: 객체의 최상위 수준만 복사합니다. 내부에 또 다른 객체가 들어있다면 그 객체의 주소만 복사되어 원본과 공유됩니다.
- Deep Copy: 객체의 모든 계층을 재귀적으로 복사하여 힙 메모리에 완전히 새로운 구조를 생성합니다. 원본과 복사본은 어떤 데이터도 공유하지 않습니다.

### 3. 예제

### 3.1. 원시 타입의 독립적 복사

```python
%%writefile ex1.js

let scoreKim = 100;
let scoreLee = scoreKim; // 값 자체가 복사됨

scoreLee = 80;

console.log(scoreKim); // 100
console.log(scoreLee); // 80
```

```python
!node ex1.js
```

3.2. 참조 타입의 주소 공유 (Shallow Copy)

```python
%%writefile ex2.js

let memberPark = {
    name: "박지민",
    skills: ["JS", "CSS"]
};

let memberChoi = memberPark; // 주소값만 복사됨
memberChoi.name = "최현우";

console.log(memberPark.name); // "최현우" (원본이 변경됨)
console.log(memberPark === memberChoi); // true
```

```python
!node ex2.js
```

3.3. 전개 연산자의 한계 (중첩 객체 얕은 복사)

```python
%%writefile ex3.js

let userJung = {
    name: "정다은",
    info: { city: "서울" }
};

let clonedJung = { ...userJung }; // 1단계만 복사
clonedJung.info.city = "부산";

console.log(userJung.info.city); // "부산" (내부 객체 주소는 공유됨)
```

```python
!node ex3.js
```

### 실습 3: JSON 메서드를 활용한 Deep Copy

**문제:** 아래 중첩 객체를 Deep Copy 하고 원본의 보존을 확인하세요.



```python
%%writefile ex6.js

let studentHan = {
    name: "한유진",
    profile: { city: "대구" }
};

// JSON으로 깊은 복사 — 중첩 객체까지 완전히 새로 만들어짐
let studentCopy = JSON.parse(JSON.stringify(studentHan));

// 복사본 수정
studentCopy.profile.city = "서울";

// 원본 확인
console.log(studentHan.profile.city);  // "대구" — 원본 유지됨
console.log(studentCopy.profile.city); // "서울" — 복사본만 바뀜
```

```python
!node ex6.js
```

상세 설명: JSON.stringify는 객체를 문자열로 직렬화하며, 이때 참조 관계가 모두 끊깁니다. 이를 다시 JSON.parse로 객체화하면 힙 메모리에 완전히 새로운 데이터 구조가 생성됩니다. 일반적인 전개 연산자({...})로는 해결할 수 없는 2단계 깊이의 객체(profile)까지 복사하기 위해 이 방법을 사용합니다.

### 실습 4: structuredClone 활용

**문제:** 객체 내부의 배열까지 완벽하게 복사하는 현대적 방법을 실습하세요.

```python
%%writefile ex7.js

let userLim = {
    name: "임윤아",
    tags: ["운동", "독서"]
};

// structuredClone — 중첩 배열/객체까지 완전히 새로 복사
let userCopy = structuredClone(userLim);

// 복사본 수정
userCopy.tags.push("요리");

// 원본 확인
console.log(userLim.tags);   // ["운동", "독서"] — 원본 유지됨
console.log(userCopy.tags);  // ["운동", "독서", "요리"] — 복사본만 바뀜
```

```python
!node ex7.js
```

상세 설명: structuredClone은 최신 자바스크립트 엔진이 제공하는 표준 딥 카피 API입니다. JSON 방식과 달리 Date 객체나 RegExp 등 특수 객체도 안정적으로 복사하며, 중첩된 배열(tags)까지 힙 메모리에 새로 할당하여 원본과의 연결을 완벽히 차단합니다.

### 실습 5: 재귀 함수를 이용한 커스텀 Deep Copy

**문제:** 모든 계층을 복사하는 함수를 완성하여 객체 내부의 배열 데이터를 보호하세요.

```python
%%writefile ex8.js

function performDeepCopy(target) {
    // 원시 타입이거나 null이면 그냥 반환 (더 이상 파고들 필요 없음)
    if (target === null || typeof target !== "object") {
        return target;
    }

    // 배열이면 빈 배열, 객체면 빈 객체로 시작
    let result = Array.isArray(target) ? [] : {};

    for (let key in target) {
        if (Object.prototype.hasOwnProperty.call(target, key)) {
            // 각 속성값에 대해 재귀 호출 — 중첩 객체/배열도 다시 이 함수로 들어옴
            result[key] = performDeepCopy(target[key]);
        }
    }
    return result;
}

// 테스트
let original = {
    name: "한유진",
    profile: { city: "대구" },
    tags: ["운동", "독서"]
};

let copy = performDeepCopy(original);

copy.profile.city = "서울";
copy.tags.push("요리");

console.log(original.profile.city); // "대구" — 원본 유지
console.log(original.tags);         // ["운동", "독서"] — 원본 유지
console.log(copy.profile.city);     // "서울"
console.log(copy.tags);             // ["운동", "독서", "요리"]
```

```python
!node ex8.js
```

상세 설명: 이 함수는 객체의 모든 속성을 순회합니다. 속성 값이 다시 객체나 배열일 경우, 자기 자신(performDeepCopy)을 다시 호출하는 재귀(Recursion) 방식을 사용합니다. 이를 통해 아무리 깊은 단계의 중첩 구조라도 힙 메모리 최하단까지 내려가 새로운 복사본을 만들어내어 완벽한 독립성을 보장합니다.

# 일급 객체로서의 함수, 고차 함수(Higher-order Function)와 순수 함수

## 1. 배경

자바스크립트가 현대 웹 개발에서 가장 강력한 언어 중 하나로 자리 잡은 이유는 함수를 단순한 실행 단위가 아닌 데이터와 동일하게 취급하는 설계 덕분입니다. 1950년대 리스프(LISP) 언어에서 기원한 함수형 프로그래밍 패러다임을 자바스크립트가 수용하면서 다음과 같은 이점을 얻게 되었습니다.

- 추상화의 수준 향상: 반복되는 제어 로직을 함수 내부로 숨기고 행위 자체를 주입받아 처리할 수 있습니다.
- 상태 관리의 안정성: 순수 함수를 통해 외부 환경에 영향을 주지 않는 예측 가능한 코드를 작성할 수 있습니다.
- 선언적 프로그래밍: 어떻게(How) 할 것인가보다 무엇을(What) 할 것인가에 집중하여 코드의 가독성을 높입니다.

## 2. 설명

### 2.1. 일급 객체로서의 함수

자바스크립트에서 함수는 객체입니다. 따라서 일반적인 객체가 할 수 있는 모든 일을 할 수 있으며 이를 일급 객체라고 부릅니다.

- 변수 할당: 함수를 변수나 객체의 속성에 할당할 수 있습니다.
- 인자 전달: 다른 함수의 매개변수로 함수를 전달할 수 있습니다.
- 결과 반환: 함수의 반환값으로 함수를 사용할 수 있습니다.

### 2.2. 고차 함수 (Higher-order Function)

고차 함수는 일급 객체의 특성을 활용하여 다음과 같은 조건을 하나 이상 만족하는 함수를 의미합니다.

- 함수를 인자로 받는 경우: 인자로 전달되는 함수를 콜백 함수(Callback Function)라고 합니다.
- 함수를 결과로 반환하는 경우: 함수를 생성하여 반환하는 팩토리 함수나 커링(Currying) 기법에 사용됩니다.

### 2.3. 순수 함수

함수형 프로그래밍의 가장 작은 단위이며 다음 두 가지 조건을 엄격하게 준수합니다.

- 참조 투명성: 동일한 인자가 주어지면 항상 동일한 값을 반환합니다. 외부 상태(전역 변수, 무작위 값 등)에 의존하지 않습니다.
- 부작용(Side Effect) 없음: 함수 외부의 변수를 수정하거나, 입출력(I/O)을 수행하거나, 매개변수로 받은 객체를 직접 변경하지 않습니다.

## 3. 예제

### 3.1. 일급 객체의 특성 활용

```python
%%writefile eex1.js

const greetUser = (name) => {
    return `안녕하세요, ${name}님!`;
};

function displayMessage(formatFn, userName) {
    // 함수를 인자로 전달받아 실행
    console.log(formatFn(userName));
}

displayMessage(greetUser, "김철수"); // 안녕하세요, 김철수님!
```

```python
!node eex1.js
```

3.2. 고차 함수 (함수 반환 및 커링)

```python
%%writefile eex2.js

function createMultiplier(factor) {
    // 내부에서 함수를 생성하여 반환
    return function(number) {
        return number * factor;
    };
}

const doubleValue = createMultiplier(2);
console.log(doubleValue(100)); // 200 (이영희의 점수 계산 예시)
```

```python
!node eex2.js
```

3.3. 순수 함수 vs 비순수 함수

```python
%%writefile eex3.js

let bonusPoint = 10;

// 비순수 함수: 외부 변수 bonusPoint에 의존하며 이를 변경함
function calculateTotalImpure(score) {
    bonusPoint += 5;
    return score + bonusPoint;
}

// 순수 함수: 오직 인자로만 계산하고 외부 상태를 변경하지 않음
function calculateTotalPure(score, bonus) {
    return score + bonus;
}

console.log(calculateTotalPure(80, 10)); // 90
```

```python
!node eex3.js
```

### 실습 2: 고차 함수 - 콜백 함수를 통한 데이터 필터링

문제: 배열과 조건 함수를 인자로 받아 조건에 맞는 이름만 추출하는 customFilter 함수를 완성하세요.

```python
%%writefile eex5.js

const userNames = ["이진우", "박서윤", "이민호", "최현우"];

function customFilter(array, callback) {
    const result = [];
    for (let i = 0; i < array.length; i++) {
        if (callback(array[i])) {  // 콜백 함수에 현재 요소를 넣어서 true면 추가
            result.push(array[i]);
        }
    }
    return result;
}

const startsWithLee = (name) => name.startsWith("이");
const filteredList = customFilter(userNames, startsWithLee);

console.log(filteredList); // ['이진우', '이민호']
```

```python
!node eex5.js
```

상세 설명: customFilter는 함수를 인자로 받는 고차 함수입니다. 이진우, 이민호와 같은 데이터를 걸러내는 구체적인 기준(콜백)을 외부에서 주입받아 로직을 수행합니다.

### 실습 4: 고차 함수 - 커링(Currying) 구현

문제: 성씨를 고정한 뒤 이름을 나중에 받아 풀네임을 반환하는 고차 함수를 작성하세요.

**커링**(**Currying**)은 여러 개의 인자를 가진 함수를 **단 하나의 인자만 받는 함수들의 체인**으로 만드는 기법을 말합니다.

```python
%%writefile eex7.js

function setLastName(lastName) {
    // 성씨를 기억한 채로 이름을 받는 함수를 반환
    return function(firstName) {
        return lastName + firstName;
    };
}

const kimFamily = setLastName("김"); // "김"을 고정
console.log(kimFamily("태형")); // "김태형"
console.log(kimFamily("민준")); // "김민준"
console.log(kimFamily("지수")); // "김지수"
```

```python
!ndoe eex7.js
```

상세 설명: 함수가 함수를 반환하는 고차 함수의 특징을 활용합니다. 김태형, 김남준과 같은 데이터를 만들기 위해 성씨를 미리 닫힌 스코프에 저장해두고 나중에 이름을 받아 완성합니다.

### 실습 5: 고차 함수 체이닝 (Map & Filter)

문제: 학생 객체 배열에서 80점 이상인 학생의 이름만 추출하여 "우수학생: 이름" 형태의 배열을 만드세요.

```python
%%writefile eex8.js

const students = [
    { name: "정다은", score: 85 },
    { name: "강하늘", score: 70 },
    { name: "임윤아", score: 95 }
];

const result = students
    .filter(student => student.score >= 80)   // 80점 이상만 걸러내기
    .map(student => `우수학생: ${student.name}`); // 이름만 추출해서 형식 맞추기

console.log(result); // ["우수학생: 정다은", "우수학생: 임윤아"]
```

```python
!node eex8.js
```

상세 설명: filter와 map은 모두 고차 함수입니다. 정다은, 임윤아 학생의 데이터를 선별하고 가공하는 과정을 체이닝하여 선언적으로 표현했습니다.
