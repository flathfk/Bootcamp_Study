## 트리 (Tree)

### 1. 트리 기본 개념

트리는 노드들이 부모-자식 관계로 연결된 **비선형 계층적 자료구조**입니다. 데이터가 일렬로 나열된 선형 구조와 달리, 데이터 간의 상하 관계를 표현하는 데 최적화되어 있습니다.

### 1.1 노드 (Node)

- **정의**: 트리를 구성하는 가장 기본이 되는 원소입니다.
- **구성**: 실제 저장되는 **데이터**(Value)와 자식 노드를 가리키는 **참조**(Reference)값으로 이루어집니다.

### 1.1.2 루트 노드 (Root Node)

- **정의**: 트리의 최상단에 위치한 단 하나의 노드입니다.
- **특징**: 모든 노드의 출발점이며, 부모 노드가 없는 유일한 노드입니다. 트리 탐색은 항상 루트에서 시작됩니다.

### 1.1.3 리프 노드 (Leaf Node)

- **정의**: 자식 노드가 하나도 없는 최하단의 노드입니다.
- **별칭**: 단말 노드(Terminal Node) 또는 잎 노드라고도 부릅니다.

### 1.1.4 부모(Parent) / 자식(Child) / 형제(Sibling) 노드

- **부모 노드**: 특정 노드보다 한 단계 위에 위치하며 그 노드와 연결된 노드입니다.
- **자식 노드**: 특정 노드보다 한 단계 아래에 위치하며 그 노드로부터 연결된 노드입니다.
- **형제 노드**: 동일한 부모 노드를 가진 노드들의 집합입니다.

### 1.1.5 간선 (Edge)

- **정의**: 노드와 노드 사이를 잇는 연결선입니다.
- **특징**: 부모 노드와 자식 노드를 연결하며, 트리의 관계를 형성하는 물리적인 선입니다.

### 1.1.6. 깊이와 높이

트리의 규모를 측정하기 위한 핵심 지표입니다.

### 1.1.7. 깊이 (Depth)

- **정의**: 루트 노드에서 특정 노드까지 도달하기 위해 거쳐야 하는 **간선(Edge)의 수**입니다.
- **기준**: 루트 노드의 깊이는 **0**입니다. 아래로 내려갈수록 1씩 증가합니다.

### 1.1.8. 높이 (Height)

- **정의**: 특정 노드에서 가장 멀리 떨어져 있는 리프 노드까지의 **간선의 수**입니다.
- **기준**: 리프 노드의 높이는 **0**입니다. 트리 전체의 높이는 루트 노드의 높이를 의미합니다.

### 1.1.9. 레벨 (Level)

- **정의**: 트리의 특정 깊이를 가지는 노드들의 집합을 층으로 나눈 것입니다.
- **기준**: 보통 루트를 Level 0(또는 1)로 설정하며, 깊이와 유사한 개념으로 쓰입니다.

### 1.2. 트리의 주요 특징

- **비순환성 (Acyclic)**: 사이클이 존재하지 않아 한 번 내려간 경로는 다시 돌아올 수 없습니다.
- **재귀적 구조**: 모든 노드는 그 자체로 하나의 '서브트리(Subtree)'의 루트가 될 수 있습니다.

```python
%%writefile ex1.js

// < 기본 노드(Node) 객체 정의 >
// 트리의 가장 기본 단위인 노드를 JavaScript 클래스로 정의

class TreeNode {
  constructor(value) {
    this.value = value;      // 1.1 노드가 담고 있는 데이터
    this.children = [];      // 1.1 자식 노드들에 대한 '참조' 배열
  }
}

// 사용 예시
const tempNode = new TreeNode("데이터");
console.log(tempNode.value); // "데이터"
console.log(tempNode.children); // [] (빈 배열)

// ----------------------------------------------------------------

// < 수동으로 트리 구조 생성 >
// 생성된 노드 객체들을 children 배열에 넣어 연결
// 실제 트리 형태를 메모리에 구성

// 1. 노드들 생성
const root = new TreeNode("컴퓨터"); // 1.2 루트 노드
const folder1 = new TreeNode("문서");
const folder2 = new TreeNode("사진");
const file1 = new TreeNode("보고서.docx"); // 1.3 리프 노드

// 2. 부모-자식 관계 연결
root.children.push(folder1); // 루트의 자식으로 문서 폴더 추가
root.children.push(folder2); // 루트의 자식으로 사진 폴더 추가
folder1.children.push(file1); // 문서 폴더의 자식으로 파일 추가

// ----------------------------------------------------------------

// < 재귀(Recursion)를 이용한 트리 전체 탐색 >
// 트리의 모든 노드를 하나씩 방문하여 값을 출력
// 트리의 재귀적 특성을 활용하는 가장 기본적인 알고리즘

function printAllNodes(node) {
  console.log(node.value); // 현재 노드의 값 출력

  // 자식 노드가 있다면, 각각의 자식에 대해 이 함수를 다시 호출
  // 자식이 없으면(리프 노드) 자동으로 반복문이 종료됨
  node.children.forEach(child => {
    printAllNodes(child); // 재귀 호출
  });
}

// 실행 (수동 생성한 트리 사용)
printAllNodes(root);
// 출력 순서: "컴퓨터" -> "문서" -> "보고서.docx" -> "사진"

// ----------------------------------------------------------------

// < 특정 노드의 존재 여부 확인 (검색) >
// 트리를 탐색하여 특정 값을 가진 노드가 있는지 확인

function containsNode(node, targetValue) {
  // 1. 현재 노드의 값이 찾는 값과 같으면 true 반환
  if (node.value === targetValue) {
    return true;
  }

  // 2. 자식 노드들을 돌며 검색
  for (let child of node.children) {
    if (containsNode(child, targetValue)) {
      return true; // 자식 서브트리에서 찾았으면 true 반환
    }
  }

  // 3. 어디서도 찾지 못했다면 false 반환
  return false;
}

// 실행 예시
console.log(containsNode(root, "보고서.docx")); // true
console.log(containsNode(root, "게임.exe"));   // false

// ----------------------------------------------------------------

// < 자식 노드 삭제 >
// 특정 부모 노드 아래의 자식 노드를 참조 해제하여 삭제
// JavaScript의 가비지 컬렉터가 더 이상 참조되지 않는 객체를 메모리에서 해제

function removeChildNode(parentNode, valueToDelete) {
  // 1. 기존 자식 노드 개수를 확인합니다.
  const originalCount = parentNode.children.length;

  // 2. filter를 사용하여 삭제할 값을 제외한 나머지 노드들로 배열을 재구성합니다.
  // 이 과정에서 삭제 대상 노드에 대한 '참조'가 끊어집니다.
  parentNode.children = parentNode.children.filter(child => {
    return child.value !== valueToDelete;
  });

  // 3. 삭제 결과에 따른 피드백을 출력합니다.
  if (parentNode.children.length < originalCount) {
    console.log(`알림: '${valueToDelete}' 노드가 성공적으로 삭제되었습니다.`);
  } else {
    console.log(`알림: 삭제하려는 '${valueToDelete}' 노드를 찾을 수 없습니다.`);
  }
}

// root -> [folder1("문서"), folder2("사진")]

removeChildNode(root, "사진");

console.log(root.children);
// 결과: folder2("사진") 노드가 삭제되고 folder1("문서")만 남음
```

코드 흐름 정리:

- `root`(컴퓨터) → `children.push(folder1, folder2)` 로 두 자식 연결

- `folder1`(문서) → `children.push(file1)` 로 리프 노드 연결

- `folder2`(사진) → `children` 비어 있음 → 리프 취급

`folder2`에 `children`이 없는 건 아니고 `children = []` (빈 배열) 상태 (`push`를 안 했으니 자식이 없는 것)

```python
%%writefile ex2.js

class TreeNode {
  constructor(value) {
    this.value = value;
    this.children = [];
  }
}

const root = new TreeNode("사장");
const jeon = new TreeNode("전무");
const lee  = new TreeNode("이사");
const bu1  = new TreeNode("부장1");
const bu2  = new TreeNode("부장2");
const gae1 = new TreeNode("개발부장1");
const gae2 = new TreeNode("개발부장2");

root.children.push(jeon);
root.children.push(lee);
jeon.children.push(bu1);
jeon.children.push(bu2);
lee.children.push(gae1);
lee.children.push(gae2);

function printAllNodes(node) {
  console.log(node.value);
  node.children.forEach(child => {
    printAllNodes(child);
  });
}

function removeChildNode(parentNode, valueToDelete) {
  const originalCount = parentNode.children.length;
  parentNode.children = parentNode.children.filter(child => {
    return child.value !== valueToDelete;
  });
  if (parentNode.children.length < originalCount) {
    console.log(`'${valueToDelete}' 노드가 삭제되었습니다.`);
  } else {
    console.log(`'${valueToDelete}' 노드를 찾을 수 없습니다.`);
  }
}

removeChildNode(lee, "개발부장2");
printAllNodes(root);
```

### 2. 이진 트리 (Binary Tree)

모든 노드가 **최대 2개**의 자식 노드만을 가질 수 있는 트리 구조입니다. 자식 노드는 보통 '왼쪽 자식'과 '오른쪽 자식'으로 구분합니다.

- **특징**: 자식 노드의 개수가 0개, 1개, 혹은 2개까지만 허용됩니다.
- **중요성**: 데이터 검색, 정렬 알고리즘의 기초가 되며 컴퓨터 메모리 상에서 가장 효율적으로 다룰 수 있는 트리 형태입니다.

### 2.1. 완전 이진 트리 (Complete Binary Tree)

트리의 마지막 레벨을 제외하고 모든 레벨이 노드로 가득 차 있으며, 마지막 레벨의 노드들은 **왼쪽부터 차례대로** 채워져 있는 트리입니다.

- **특징**:
    - 중간에 빈 곳이 있으면 안 됩니다. (반드시 왼쪽부터 채워야 함)
    - 배열을 이용해 트리를 구현할 때 메모리 낭비가 없어 매우 효율적입니다.
- **활용**: 힙(Heap) 자료구조를 구현할 때 주로 사용됩니다.

### 2.2. 포화 이진 트리 (Full Binary Tree / Perfect Binary Tree)

모든 레벨이 노드로 꽉 차 있는 트리입니다. 즉, 모든 리프 노드(1.3)가 같은 깊이(2.1)를 가지며, 리프 노드를 제외한 모든 노드가 2개의 자식을 가집니다.

- **특징**:
    - 트리의 높이가 h일 때, 전체 노드의 개수는 항상 2^{h+1} - 1개가 됩니다.
    - 이진 트리 중에서 가장 균형 잡히고 완벽한 삼각형 형태를 띱니다.
- **차이점**: 모든 '포화 이진 트리'는 '완전 이진 트리'이지만, 그 역은 성립하지 않습니다.

### 2.3. 트리의 주요 특징

- **이진 트리의 효율성**: 일반 트리와 달리 자식 수가 제한되어 있어 탐색 알고리즘 설계 시 복잡도가 낮아집니다.
- **배열 구현**: 이진 트리는 부모 인덱스가 n일 때 왼쪽 자식은 2n, 오른쪽 자식은 2n+1이라는 수학적 규칙을 가져 배열로 변환하기 쉽습니다.

```python
%%writefile ex3.js

// ----------------------------------------------------------------
// 이진 트리 전용 노드 클래스
// 일반 트리와 달리 자식이 최대 2개 -> children 배열 대신 left, right 사용
// ----------------------------------------------------------------
class BinaryTreeNode {
  constructor(value) {
    this.value = value;
    this.left = null;  // 왼쪽 자식 참조
    this.right = null; // 오른쪽 자식 참조
  }
}

// ----------------------------------------------------------------
// 포화 이진 트리 구성 (모든 레벨이 가득 찬 상태)
// ----------------------------------------------------------------

// 루트 노드 (Level 0)
const root = new BinaryTreeNode("A");

// Level 1 - 루트의 왼쪽/오른쪽 자식
root.left  = new BinaryTreeNode("B");
root.right = new BinaryTreeNode("C");

// Level 2 - 리프 노드 (자식 없음, 모든 깊이 동일)
root.left.left   = new BinaryTreeNode("D");
root.left.right  = new BinaryTreeNode("E");
root.right.left  = new BinaryTreeNode("F");
root.right.right = new BinaryTreeNode("G");

// ----------------------------------------------------------------
// 완전 이진 트리 규칙 확인
// 완전 이진 트리는 왼쪽부터 빈틈없이 채워져야 함
// 왼쪽 자식이 없는데 오른쪽 자식이 있으면 규칙 위반
// ----------------------------------------------------------------
function isLeaf(node) {
  return node.left === null && node.right === null;
}

// isLeaf 확인
console.log("D는 리프 노드?", isLeaf(root.left.left));  // true
console.log("B는 리프 노드?", isLeaf(root.left));       // false

function checkCompleteRule(node) {
  if (node.left === null && node.right !== null) {
    console.log("경고: 왼쪽 자식이 비어있으므로 완전 이진 트리가 아닙니다.");
    return false;
  }
  console.log("정상: 완전 이진 트리 규칙을 만족합니다.");
  return true;
}

checkCompleteRule(root);

// ----------------------------------------------------------------
// 트리의 높이 계산
// 루트부터 가장 먼 리프까지의 거리 (간선 기준)
// 빈 노드 -1 -> 리프 노드 0 -> 위로 올라갈수록 +1
// ----------------------------------------------------------------
function getBinaryTreeHeight(node) {
  if (node === null) return -1; // 빈 노드는 -1

  const leftHeight  = getBinaryTreeHeight(node.left);
  const rightHeight = getBinaryTreeHeight(node.right);

  // 왼쪽/오른쪽 중 더 깊은 쪽 + 1
  return Math.max(leftHeight, rightHeight) + 1;
}

console.log("트리의 전체 높이:", getBinaryTreeHeight(root)); // 2

// ----------------------------------------------------------------
// 특정 노드의 레벨 출력
// 루트 = 0, 아래로 내려갈수록 +1
// ----------------------------------------------------------------
function printLevel(node, currentLevel = 0) {
  if (node === null) return;

  console.log(`${node.value} 노드의 레벨: ${currentLevel}`);

  // 자식으로 내려갈 때 레벨 1 증가
  printLevel(node.left,  currentLevel + 1);
  printLevel(node.right, currentLevel + 1);
}

printLevel(root);
// A: 0, B: 1, C: 1, D: 2, E: 2, F: 2, G: 2
```

### 2.5. 트리의 주요 특징

- **이진 트리의 효율성**: 자식 노드가 최대 2개로 제한되므로, 배열의 인덱스($2n, 2n+1$)를 사용하여 부모-자식 관계를 수학적으로 즉시 계산할 수 있습니다.
- **균형의 중요성**: 포화 이진 트리처럼 균형이 잘 잡힌 트리는 탐색 속도가 매우 빠르지만, 한쪽으로 치우친 트리는 연결 리스트와 다를 바 없는 성능을 보입니다.

### 3. 이진 탐색 트리 (Binary Search Tree, BST)

이진 탐색 트리는 이진 트리의 일종으로, 데이터를 저장할 때 특정 '크기 규칙'을 적용하여 탐색 속도를 극대화한 자료구조입니다.

### 3.1 정렬 구조의 이해

BST의 핵심은 모든 노드에 대해 다음의 규칙이 항상 성립한다는 점입니다.

- **왼쪽 자식 노드**: 부모 노드의 값보다 **작은** 값들만 위치합니다.
- **오른쪽 자식 노드**: 부모 노드의 값보다 **큰** 값들만 위치합니다.
- **중복 불가**: 일반적으로 트리 내에 동일한 값을 가진 노드는 존재하지 않습니다.
- **장점**: 이 규칙 덕분에 데이터를 찾을 때마다 탐색 범위를 절반씩 줄일 수 있어 매우 효율적($O(\log n)$)입니다.

### 3.2 탐색 (Search)

원하는 값을 찾는 과정은 루트 노드부터 시작하여 크기를 비교하며 내려갑니다.

1. 찾는 값이 현재 노드보다 **작으면 왼쪽**으로 이동합니다.
2. 찾는 값이 현재 노드보다 **크면 오른쪽**으로 이동합니다.
3. 일치하는 값을 찾거나, 더 이상 갈 곳이 없으면(null) 종료합니다.

### 3.3 삽입 (Insertion)

새로운 데이터를 추가할 때도 탐색과 동일한 논리를 사용합니다.

1. 루트부터 비교를 시작하여 자기 자리를 찾아 내려갑니다.
2. 비교 중 비어있는 자리가 생기면 그곳에 새로운 노드를 연결합니다.
3. **주의**: 삽입 순서에 따라 트리의 모양이 결정되며, 이미 정렬된 데이터를 순서대로 넣으면 트리가 한쪽으로 치우쳐 성능이 저하될 수 있습니다.

### 3.4 삭제 (Deletion)

삭제는 BST에서 가장 복잡한 연산이며, 삭제 후에도 BST의 규칙이 유지되어야 합니다.

- **자식이 없는 경우 (리프 노드)**: 단순히 해당 노드를 삭제(null 처리)합니다.
- **자식이 1개인 경우**: 삭제될 노드의 부모와 자식을 직접 연결해 줍니다.
- **자식이 2개인 경우**:
    1. 삭제할 노드와 가장 값이 비슷한 노드(오른쪽 서브트리의 가장 작은 값 혹은 왼쪽 서브트리의 가장 큰 값)를 찾아 삭제할 위치로 옮깁니다.
    2. 그 후 대체된 노드의 원래 위치를 정리합니다.

```python
%%writefile ex4.js

// ----------------------------------------------------------------
// BST 전용 노드 클래스
// 이진 트리와 구조는 같지만, 삽입할 때 정렬 규칙을 지켜야 함
// 왼쪽 < 현재 < 오른쪽
// ----------------------------------------------------------------
class BSTNode {
  constructor(value) {
    this.value = value;
    this.left = null;
    this.right = null;
  }
}

// ----------------------------------------------------------------
// BST 클래스
// 노드들을 정렬된 상태로 관리하는 트리 구조
// ----------------------------------------------------------------
class BinarySearchTree {
  constructor() {
    this.root = null;
  }

  // ----------------------------------------------------------------
  // 삽입 (Insertion)
  // 루트부터 값을 비교하며 빈 자리를 찾아 내려감
  // 작으면 왼쪽, 크면 오른쪽
  // ----------------------------------------------------------------
  insert(value) {
    const newNode = new BSTNode(value);

    // 트리가 비어있으면 루트로 설정
    if (!this.root) {
      this.root = newNode;
      return;
    }

    let current = this.root;
    while (true) {
      if (value < current.value) {
        // 현재보다 작으면 왼쪽으로
        if (!current.left) { current.left = newNode; break; }
        current = current.left;
      } else {
        // 현재보다 크면 오른쪽으로
        if (!current.right) { current.right = newNode; break; }
        current = current.right;
      }
    }
  }

  // ----------------------------------------------------------------
  // 탐색 (Search)
  // 매 비교마다 탐색 범위가 절반으로 줄어듦
  // ----------------------------------------------------------------
  find(value) {
    if (!this.root) return false;

    let current = this.root;
    while (current) {
      if (value === current.value) return true; // 찾음
      // 작으면 왼쪽, 크면 오른쪽으로 이동
      current = value < current.value ? current.left : current.right;
    }
    return false; // 끝까지 못 찾음
  }

  // ----------------------------------------------------------------
  // 최소값 / 최대값
  // BST 구조상 가장 왼쪽 끝 = 최소, 가장 오른쪽 끝 = 최대
  // ----------------------------------------------------------------
  getMin(node = this.root) {
    while (node && node.left) node = node.left;
    return node ? node.value : null;
  }

  getMax(node = this.root) {
    while (node && node.right) node = node.right;
    return node ? node.value : null;
  }

  // ----------------------------------------------------------------
  // 리프 노드 삭제 (Deletion)
  // 자식이 없는 노드만 삭제 (자식이 있으면 연결 유지)
  // ----------------------------------------------------------------
  removeLeaf(value) {
    this.root = this._removeNode(this.root, value);
  }

  _removeNode(node, value) {
    if (!node) return null;

    if (value < node.value) {
      // 찾는 값이 더 작으면 왼쪽 서브트리로
      node.left = this._removeNode(node.left, value);
      return node;
    } else if (value > node.value) {
      // 찾는 값이 더 크면 오른쪽 서브트리로
      node.right = this._removeNode(node.right, value);
      return node;
    } else {
      // 값을 찾음 - 자식이 없으면 null 반환 (연결 끊기)
      if (!node.left && !node.right) return null;
      // 자식이 있으면 삭제하지 않음
      return node;
    }
  }
}

// ----------------------------------------------------------------
// 실행
// ----------------------------------------------------------------
const bst = new BinarySearchTree();

// 값 삽입
bst.insert(10);
bst.insert(5);
bst.insert(15);
bst.insert(3);
bst.insert(7);
bst.insert(12);
bst.insert(18);

// 삽입 후 루트 확인
console.log("루트 노드:", bst.root.value); // 10

// 탐색
console.log("7 존재?", bst.find(7));   // true
console.log("99 존재?", bst.find(99)); // false

// 최소 / 최대
console.log("최소값:", bst.getMin()); // 3
console.log("최대값:", bst.getMax()); // 18

// 리프 노드 삭제
bst.removeLeaf(3);
console.log("3 삭제 후 존재?", bst.find(3)); // false

// 자식 있는 노드 삭제 시도 (삭제 안 됨)
bst.removeLeaf(5);
console.log("5 삭제 시도 후 존재?", bst.find(5)); // true (자식 있어서 유지)
```

### 4. 트리 순회 (Tree Traversal)

트리 순회란 트리의 각 노드를 특정 규칙에 따라 한 번씩 방문하는 과정을 말합니다. 크게 **깊이 우선 탐색**(DFS)과 너비 우선 탐색(BFS)으로 나뉩니다.

### 4.1 깊이 우선 탐색 (DFS, Depth-First Search)

뿌리(Root)에서 시작하여 한 방향의 자식들을 끝까지 파고든 뒤, 더 이상 갈 곳이 없으면 옆 줄기로 이동하는 방식입니다. 재귀(Recursion)를 이용해 구현하며, 방문 순서에 따라 세 가지로 분류됩니다.

- **4.1.1 전위 순회 (Preorder Traversal)**
    - **순서**: **Root** → 왼쪽 자식 → 오른쪽 자식
    - **특징**: 부모 노드를 먼저 처리해야 하는 트리 복사나 구조 파악에 사용됩니다.
- **4.1.2 중위 순회 (Inorder Traversal)**
    - **순서**: 왼쪽 자식 → **Root** → 오른쪽 자식
    - **특징**: **이진 탐색 트리**(BST)에서 중위 순회를 하면 데이터가 **오름차순**(작은 값→큰 값)으로 정렬되어 출력됩니다.
- **4.1.3 후위 순회 (Postorder Traversal)**
    - **순서**: 왼쪽 자식 → 오른쪽 자식 → **Root**
    - **특징**: 자식 노드를 먼저 처리해야 하는 트리 삭제(하단부터 삭제)나 폴더 용량 계산 등에 사용됩니다.

### 4.2 너비 우선 탐색 (BFS, Breadth-First Search)

루트 노드부터 시작하여 인접한 노드(형제 노드)를 먼저 방문하는 방식입니다.

- **4.2.1 레벨 순회 (Level Order Traversal)**
    - **방식**: 위에서 아래로, 왼쪽에서 오른쪽으로 층(Level)별로 방문합니다.
    - **구현**: 재귀 대신 **큐(Queue)** 자료구조를 사용하여 방문할 노드들을 순서대로 저장하고 꺼냅니다.
    - **특징**: 두 노드 사이의 최단 경로를 찾거나, 가까운 노드부터 탐색해야 할 때 유리합니다.

```python
%%writefile ex5.js

class BinaryTreeNode {
  constructor(value) {
    this.value = value;
    this.left = null;
    this.right = null;
  }
}

// 4.3.1 전위 순회 (Preorder: Root-L-R)
function preorder(node) {
  if (!node) return;
  console.log(node.value); // 부모 먼저
  preorder(node.left);    // 왼쪽
  preorder(node.right);   // 오른쪽
}

// 4.3.2 중위 순회 (Inorder: L-Root-R)
function inorder(node) {
  if (!node) return;
  inorder(node.left);     // 왼쪽 먼저 끝까지
  console.log(node.value); // 부모
  inorder(node.right);    // 오른쪽
}

// 4.3.3 후위 순회 (Postorder: L-R-Root)
function postorder(node) {
  if (!node) return;
  postorder(node.left);   // 왼쪽
  postorder(node.right);  // 오른쪽
  console.log(node.value); // 부모 나중에
}

// 4.3.4 레벨 순회 (Level Order: BFS)
function levelOrder(root) {
  if (!root) return;
  const queue = [root]; // 큐에 루트 삽입

  while (queue.length > 0) {
    const node = queue.shift(); // 앞에서 하나 꺼냄
    console.log(node.value);

    // 왼쪽, 오른쪽 자식을 큐의 뒤에 추가 (나중에 방문)
    if (node.left) queue.push(node.left);
    if (node.right) queue.push(node.right);
  }
}

// 1. 테스트를 위한 이진 트리 생성
//      1 (Root)
//     / \
//    2   3
//   / \
//  4   5

const root = new BinaryTreeNode(1);
root.left = new BinaryTreeNode(2);
root.right = new BinaryTreeNode(3);
root.left.left = new BinaryTreeNode(4);
root.left.right = new BinaryTreeNode(5);

// 2. 각 순회 방식 실행

console.log("--- 4.3.1 전위 순회 (Preorder) ---");
preorder(root);
// 출력: 1 -> 2 -> 4 -> 5 -> 3
// (부모 1 출력 후, 왼쪽 줄기 2-4-5 다 보고, 마지막에 오른쪽 3)

console.log("--- 4.3.2 중위 순회 (Inorder) ---");
inorder(root);
// 출력: 4 -> 2 -> 5 -> 1 -> 3
// (가장 왼쪽 4부터 시작해서 부모 2, 그다음 5... 순으로 올라옴)

console.log("--- 4.3.3 후위 순회 (Postorder) ---");
postorder(root);
// 출력: 4 -> 5 -> 2 -> 3 -> 1
// (자식들 4, 5를 먼저 처리하고 부모 2를 본 뒤, 최종적으로 루트 1)

console.log("--- 4.3.4 레벨 순회 (Level Order) ---");
levelOrder(root);
// 출력: 1 -> 2 -> 3 -> 4 -> 5
// (층별로 1층의 1, 2층의 2-3, 3층의 4-5 순서대로 출력)
```
