# ES6 Class 문법, 상속, super 키워드 및 정적 메서드

## 1. 배경

자바스크립트는 전통적으로 프로토타입 기반의 객체 지향 언어였습니다. 하지만 Java나 C++ 같은 클래스 기반 언어에 익숙한 개발자들에게 프로토타입 체인을 직접 다루는 방식은 다소 생소하고 복잡하게 느껴졌습니다.

- **가독성과 편의성:** 2015년(ES6), 자바스크립트는 더 명확하고 간결한 문법을 제공하기 위해 클래스(Class) 문법을 도입했습니다.
- **문법적 설탕(Syntactic Sugar):** ES6 클래스가 새로운 객체 모델을 만든 것은 아닙니다. 내부적으로는 여전히 프로토타입을 기반으로 동작하지만, 문법적으로 훨씬 직관적인 인터페이스를 제공합니다.
- **표준화된 상속:** 이전에는 부모의 생성자를 호출하거나 프로토타입을 연결하는 과정이 번거로웠으나, 클래스 문법을 통해 `extends`와 `super`를 사용하는 표준화된 상속 구조를 갖추게 되었습니다.

## 2. 설명

### 2.1. ES6 Class 문법

클래스는 객체를 생성하기 위한 **청사진**(Blueprint)입니다.

- **constructor:** 인스턴스를 생성하고 초기화하기 위한 특수 메서드입니다. 클래스당 하나만 존재할 수 있습니다.
- **메서드 정의:** 클래스 몸체에 정의한 메서드는 자동으로 해당 클래스의 `prototype`에 등록됩니다.

### 2.2. 상속(extends)과 super 키워드

상속은 기존 클래스의 기능을 물려받아 새로운 클래스를 만드는 기능입니다.

- **extends:** 부모 클래스의 모든 속성과 메서드를 자식 클래스가 상속받도록 연결합니다.
- **super:**
    - **함수로 호출 시:** 부모 클래스의 `constructor`를 호출합니다. 자식 클래스에서 `this`를 사용하기 전에 반드시 호출해야 합니다.
    - **참조로 사용 시:** 부모 클래스의 메서드에 직접 접근할 때 사용합니다. (`super.methodName()`)

### 2.3. 정적 메서드 (Static Methods)

- **정의:** `static` 키워드를 사용하여 정의하며, 클래스에 직접 묶인 메서드입니다.
- **특징:** 인스턴스를 생성하지 않고 클래스 이름으로 바로 호출합니다. 주로 유틸리티 함수나 고정된 데이터를 다룰 때 사용됩니다.

## 3. 예제

### 3.1. 기본 클래스와 인스턴스 생성

```python
%%writefile ex1.js

class User {
    constructor(name) {
        this.name = name; // 인스턴스 프로퍼티
    }

    // 프로토타입 메서드
    sayHello() {
        return `안녕하세요, 저는 ${this.name}입니다.`;
    }
}

const kim = new User("김철수");
console.log(kim.sayHello()); // 안녕하세요, 저는 김철수입니다.
```

3.2. 상속과 super의 활용

```python
%%writefile ex2.js

// 부모 클래스 -- User가 없으면 Manager를 만들 수 없음
class User {
  constructor(name) {
    this.name = name;
  }
  sayHello() {
    return `안녕하세요, 저는 ${this.name}입니다.`;
  }
}

// 자식 클래스 -- User를 상속받음
class Manager extends User {
  constructor(name, department) {
    super(name);  // 부모(User)의 constructor 실행
    this.department = department;
  }

  // 부모 메서드 덮어쓰기 (Overriding)
  sayHello() {
    // super.sayHello() -- 부모의 sayHello 결과를 가져와서 뒤에 붙임
    return `${super.sayHello()} 소속은 ${this.department}입니다.`;
  }
}

const lee = new Manager('이영희', '개발팀');
console.log(lee.sayHello());
// '안녕하세요, 저는 이영희입니다. 소속은 개발팀입니다.'
```

```python
!node ex2.js
```

3.3. 정적 메서드 호출

```python
%%writefile ex3.js

class MathUtils {
    static calculateTax(price) {
        return price * 0.1;
    }
}

// 인스턴스 생성 없이 클래스에서 직접 호출
console.log(MathUtils.calculateTax(10000)); // 1000 (박지민의 결제 예시)
```

### 실습 1: 클래스 기본 구조 설계

**문제:** `Student` 클래스를 만드세요. `name`과 `grade`를 인자로 받아 초기화하고, `introduce` 메서드를 통해 "저는 x학년 x입니다"를 반환하도록 작성하세요.

```python
%%writefile ex_student.js

// Student 클래스 -- 학생 객체를 찍어내는 틀
class Student {

  // constructor -- new Student() 할 때 자동 실행
  // name(이름)과 grade(학년)를 받아서 이 객체에 저장
  constructor(name, grade) {
    this.name  = name;   // 이 학생의 이름
    this.grade = grade;  // 이 학생의 학년
  }

  // introduce 메서드 -- 자기소개 문장 반환
  introduce() {
    return `저는 ${this.grade}학년 ${this.name}입니다.`;
  }
}

// new Student() -- Student 틀로 실제 학생 객체(인스턴스) 생성
const studentA = new Student('임소라', 3);

console.log(studentA.introduce()); // '저는 3학년 임소라입니다.'
```

### 실습 2: 클래스 상속과 super 호출

**문제:** 위에서 만든 `Student` 클래스를 상속받는 `ClubMember` 클래스를 만드세요. 신규 속성으로 `clubName`을 추가하고, `introduce` 메서드를 오버라이딩하여 부모의 자기소개 뒤에 "동아리는 x입니다"를 덧붙이세요.

```python
%%writefile ex_club.js

// 부모 클래스 -- 학생 기본 정보
class Student {
  constructor(name, grade) {
    this.name  = name;
    this.grade = grade;
  }
  introduce() {
    return `저는 ${this.grade}학년 ${this.name}입니다.`;
  }
}

// 자식 클래스 -- Student를 상속받고 동아리 정보 추가
class ClubMember extends Student {

  constructor(name, grade, clubName) {
    super(name, grade);      // 부모 Student의 constructor 먼저 실행
    this.clubName = clubName; // 동아리 이름 추가
  }

  // 부모의 introduce 덮어쓰기 (Overriding)
  introduce() {
    // super.introduce() -- 부모 자기소개 가져와서 뒤에 동아리 붙임
    return `${super.introduce()} 동아리는 ${this.clubName}입니다.`;
  }
}

const member = new ClubMember('임소라', 3, '영화감상부');

console.log(member.introduce()); // '저는 3학년 임소라입니다. 동아리는 영화감상부입니다.'
```

```python
!node ex_club.js
```

### 실습 3: 정적 메서드 구현

**문제:** `DateConverter` 클래스를 만들고, 연도(YYYY), 월(MM), 일(DD)을 인자로 받아 "YYYY-MM-DD" 형식을 반환하는 정적 메서드 `formatDate`를 구현하세요.

```python
%%writefile ex_date.js

// DateConverter 클래스 -- 날짜 변환 유틸
// 인스턴스 없이 바로 호출하는 도구 모음
class DateConverter {

  // static -- new 없이 클래스에서 직접 호출 가능
  static formatDate(year, month, day) {

    // padStart(2, '0') -- 한 자리 숫자면 앞에 0 붙여줌 (3 -> '03')
    const mm = String(month).padStart(2, '0');
    const dd = String(day).padStart(2, '0');

    return `${year}-${mm}-${dd}`;
  }
}

// 인스턴스 생성 없이 바로 호출
console.log(DateConverter.formatDate(2026, 3, 22));  // '2026-03-22'
console.log(DateConverter.formatDate(2026, 1, 5));   // '2026-01-05'
```

```python
!node ex_date.js
```

### 실습 4: 상속 체인에서의 메서드 접근

**문제:** `Parent` 클래스에 `greet` 메서드가 있고, `Child` 클래스가 이를 상속받습니다. `Child` 인스턴스에서 부모의 `greet`을 실행하면서 추가 로직을 더하는 코드를 작성하고, 왜 `super`를 써야 하는지 설명하세요.

```python
%%writefile ex_super.js

class Parent {
  greet() { return '좋은 아침입니다.'; }
}

class Child extends Parent {

  greet() {
    // super.greet() -- 부모의 greet 결과를 가져옴
    // this.greet() 로 하면 자기 자신을 계속 호출해서 무한루프 발생
    return `${super.greet()} 오늘도 화이팅!`;
  }
}

const child = new Child();
console.log(child.greet()); // '좋은 아침입니다. 오늘도 화이팅!'
```

### 실습 5: 정적 메서드와 인스턴스 메서드의 혼용

**문제:** `Account` 클래스를 만드세요. 각 인스턴스는 `owner`와 `balance`를 가집니다. 또한 클래스 전체에서 공통으로 사용할 `bankName`을 정적 프로퍼티나 정적 메서드로 구현하여 모든 인스턴스가 "한성은행"임을 알 수 있게 하세요.

```python
%%writefile ex_account.js

class Account {

  // static 프로퍼티 -- 인스턴스 없이 클래스에서 직접 접근
  // 모든 계좌가 공유하는 은행 이름
  static bankName = '한성은행';

  constructor(owner, balance) {
    this.owner   = owner;   // 계좌 주인
    this.balance = balance; // 잔액
  }

  // 인스턴스 메서드 -- 각 계좌의 정보 출력
  getInfo() {
    // Account.bankName -- static이라 클래스명으로 접근
    return `[${Account.bankName}] ${this.owner} 님의 잔액: ${this.balance.toLocaleString()}원`;
  }

  // static 메서드 -- 은행 이름만 필요할 때
  static getBankName() {
    return Account.bankName;
  }
}

const sorah  = new Account('임소라', 100000);
const minsoo = new Account('김민수', 50000);

console.log(sorah.getInfo());   // '[한성은행] 임소라 님의 잔액: 100,000원'
console.log(minsoo.getInfo());  // '[한성은행] 김민수 님의 잔액: 50,000원'

// 인스턴스 없이 바로 호출
console.log(Account.getBankName()); // '한성은행'
```

# Execution Context의 생성과 활성화, Call Stack과 Lexical Environment

## 1. 배경

자바스크립트 엔진이 소스코드를 실행하기 위해 필요한 환경 정보를 모아놓은 객체를 **실행 컨텍스트**(Execution Context)라고 합니다. 자바스크립트는 단순한 스크립트 언어를 넘어 복잡한 애플리케이션을 구동하기 위해, 식별자(변수, 함수명)의 유효범위와 순서를 엄격하게 관리해야 합니다.

- **실행의 단위:** 자바스크립트는 코드를 읽기 전 '평가' 과정을 거쳐 실행에 필요한 준비를 마칩니다.
- **추적의 필요성:** 함수가 중첩 호출될 때 현재 어디를 실행 중인지, 이전 위치는 어디인지 기억해야 합니다.
- **자원 관리:** 어떤 변수가 현재 유효하고 어떤 변수가 폐기되어야 하는지를 결정하는 기준이 됩니다.

## 2. 설명

### 2.1. Execution Context의 생성과 활성화

자바스크립트 엔진은 코드를 실행할 때 **평가**(Creation Phase)와 **실행**(Execution Phase)의 두 단계를 거칩니다.

- **생성(평가) 단계:** 실행 컨텍스트를 생성하고 변수, 함수 선언문을 파악하여 레코드에 등록합니다. (호이스팅이 발생하는 시점)
- **활성화(실행) 단계:** 선언문 외의 코드를 순차적으로 실행하며 변수에 값을 할당하고 함수를 호출합니다.

### 2.2. Call Stack

실행 컨텍스트가 쌓이는 데이터 구조입니다. 자바스크립트는 단일 스레드(Single Thread) 방식으로 동작하므로, 한 번에 하나의 컨텍스트만 실행할 수 있습니다.

- **Push:** 함수가 호출되면 새로운 실행 컨텍스트가 생성되어 스택의 맨 위에 쌓입니다.
- **Pop:** 함수의 실행이 종료되면 해당 컨텍스트는 스택에서 제거되고, 제어권은 바로 아래 컨텍스트로 돌아갑니다.

### 2.3. Lexical Environment

실행 컨텍스트의 핵심 구성 요소로, 식별자와 그에 일치하는 값을 관리하는 자료구조입니다.

- **Environment Record (환경 레코드):** 현재 스코프 내의 변수, 함수 선언 등을 저장합니다.
- **Outer Lexical Environment Reference (외부 렉시컬 환경 참조):** 상위 스코프의 렉시컬 환경을 가리킵니다. 이를 통해 **스코프 체인**이 형성됩니다.

## 3. 예제

### 3.1. 콜 스택의 흐름 제어

```python
%%writefile ex4.js

function first() {
    console.log("첫 번째 함수 시작");
    second();
    console.log("첫 번째 함수 종료");
}

function second() {
    console.log("두 번째 함수 실행 (박지민의 요청)");
}

first();
/*
출력 순서:
1. 첫 번째 함수 시작 (Stack: Global -> first)
2. 두 번째 함수 실행 (Stack: Global -> first -> second)
3. 첫 번째 함수 종료 (Stack: Global -> first)
*/
```

```python
!node ex4.js
```

3.3. 외부 참조를 통한 스코프 탐색

```python
%%writefile ex6.js

const globalValue = "전역";

function outer() {
    const outerValue = "외부";

    function inner() {
        // inner의 환경 레코드에 없으므로 outer의 참조를 따라가서 찾음
        console.log(outerValue); // "외부"
        // outer에도 없으므로 전역 참조를 따라감
        console.log(globalValue); // "전역"
    }
    inner();
}

outer();
```

### 실습 1: 콜 스택 추론하기

**문제:** 아래 코드가 실행될 때 콜 스택의 변화 과정을 순서대로 기술하세요. (Global 컨텍스트 포함)

- **모범 답안:**
    1. `Global` 컨텍스트가 생성되어 스택에 쌓임.
    2. `startTask()` 호출 시 `startTask` 컨텍스트가 생성되어 `Global` 위에 쌓임.
    3. 내부에서 `completeTask()` 호출 시 `completeTask` 컨텍스트가 `startTask` 위에 쌓임.
    4. `completeTask` 실행 완료 후 스택에서 제거(`Pop`).
    5. `startTask` 실행 완료 후 스택에서 제거(`Pop`).
    6. 프로그램 종료 시 `Global` 컨텍스트 제거.


### 실습 2: 환경 레코드의 식별자 결정

**문제:** 실행 컨텍스트의 **평가 단계**에서 아래 코드의 `inner` 함수 내부 환경 레코드에 등록되는 식별자들을 모두 나열하세요.

```python
%%writefile ex_cal.js

function calculate(base) {
    const bonus = 10;
    function inner(num) {
        const result = base + bonus + num;
        return result;
    }
    return inner(5);
}
calculate(100);
```

- **모범 답안:** `num` (매개변수), `result` (변수)
- **상세 설명:** `inner` 컨텍스트의 환경 레코드는 현재 자신의 스코프에서 선언된 것들만 직접 관리합니다. 이영희 개발자가 궁금해할 `base`나 `bonus`는 `inner` 레코드에는 없지만, **외부 렉시컬 환경 참조**를 통해 상위 컨텍스트에서 찾아오게 됩니다.
