## 1. 브라우저 히스토리 (Stack)

- **LIFO (Last-In, First-Out) 구조**
    - **후입선출**: 가장 최근에 방문한 URL이 가장 먼저 뒤로가기 대상이 됩니다.
    - 실제 브라우저는 **backStack / forwardStack 두 개의 스택 구조**로 이동 흐름을 관리합니다.
- **현실 시나리오 (실제 URL 흐름)**
    - [https://www.google.com](https://www.google.com/) → [https://www.naver.com](https://www.naver.com/) → [https://www.youtube.com](https://www.youtube.com/)
    - 뒤로가기 → naver.com
    - 뒤로가기 → google.com
    - 앞으로가기 → naver.com
- **핵심 동작**
    - 방문 → backStack.push(url), forwardStack 초기화
    - 뒤로가기 → backStack.pop() → forwardStack.push()
    - 앞으로가기 → forwardStack.pop() → backStack.push()

````python
%%writefile ex1.js

// 브라우저 히스토리 예제 1: URL 방문 기록

const backStack = [];
const forwardStack = [];

// 페이지 이동
function visit(url) {
    backStack.push(url);
    forwardStack.length = 0; // 앞으로가기 초기화
}

// 실행
visit('https://www.google.com');
visit('https://www.naver.com');
visit('https://www.youtube.com');

console.log('현재 페이지:', backStack[backStack.length - 1]);
// 현재 페이지: https://www.youtube.com
````

````python
!node ex1.js
````

````python
%%writefile ex2.js

const backStack = ['https://www.google.com', 'https://www.naver.com', 'https://www.youtube.com'];
const forwardStack = [];

function goBack() {
    if (backStack.length <= 1) return;

    const current = backStack.pop();
    forwardStack.push(current);

    return backStack[backStack.length - 1];
}

console.log('뒤로가기:', goBack());
// -> https://www.naver.com

// 실행 후 상태
// backStack: ['https://www.google.com', 'https://www.naver.com']
// forwardStack: ['https://www.youtube.com']
````

````python
%%writefile ex4.js

let backStack = [];
let forwardStack = [];

function visit(url) {
    backStack.push(url);
    forwardStack = []; // 새 페이지 방문 시 forwardStack 초기화
}

function goBack() {
    if (backStack.length <= 1) return;
    const current = backStack.pop();
    forwardStack.push(current);
    return backStack[backStack.length - 1];
}

function goForward() {
    if (forwardStack.length === 0) return;
    const url = forwardStack.pop();
    backStack.push(url);
    return url;
}

visit('https://www.google.com');
visit('https://news.naver.com');
visit('https://shopping.naver.com');

goBack();    // -> news.naver.com
goBack();    // -> google.com
goForward(); // -> news.naver.com

visit('https://www.github.com'); // forwardStack 초기화

console.log('backStack:', backStack);
console.log('forwardStack:', forwardStack);

// backStack: ['https://www.google.com', 'https://news.naver.com', 'https://www.github.com']
// forwardStack: []
````

### 문제 1: 함수 구현

다음 조건을 만족하는 `goBack()` 함수를 직접 구현하시오.

- 현재 페이지는 항상 `backStack`의 top
- 뒤로가기는 최소 1개 페이지는 유지해야 함
- 뒤로가기 시 현재 페이지는 `forwardStack`으로 이동

````python
%%writefile test1.js

const backStack = ['https://www.google.com', 'https://www.naver.com', 'https://www.youtube.com'];
const forwardStack = [];

function goBack() {
    // backStack에 1개 이하면 더 이상 뒤로 못 감 -> 종료
    if (backStack.length <= 1) return null;

    // 현재 페이지(top)를 backStack에서 꺼냄
    const current = backStack.pop();

    // 꺼낸 페이지를 forwardStack에 넣음 (앞으로가기 대비)
    forwardStack.push(current);

    // 이제 backStack의 새로운 top이 현재 페이지
    return backStack[backStack.length - 1];
}

console.log(goBack()); // -> https://www.naver.com
console.log(goBack()); // -> https://www.google.com
console.log(goBack()); // -> null (더 이상 뒤로 못 감)
````

### 문제 4: 전체 상태 추적

다음 실행 결과를 정확히 계산하시오.

```jsx
visit('https://www.google.com');

visit('https://www.youtube.com');

goBack();        // 1
visit('https://www.github.com'); // 2

```

- 각 단계에서 반환되는 URL을 순서대로 작성하시오
- 최종 backStack / forwardStack 상태를 작성하시오

````python
%%writefile test4.js

let backStack = [];
let forwardStack = [];

function visit(url) {
    backStack.push(url);
    forwardStack = []; // 새 페이지 방문 시 앞으로가기 기록 초기화
}

function goBack() {
    if (backStack.length <= 1) return null; // 1개 이하면 뒤로 못 감
    const current = backStack.pop();        // 현재 페이지 꺼냄
    forwardStack.push(current);             // 앞으로가기 스택에 넣음
    return backStack[backStack.length - 1]; // 새로운 현재 페이지 반환
}

function goForward() {
    if (forwardStack.length === 0) return null; // 앞으로가기 없으면 종료
    const page = forwardStack.pop();            // 앞으로가기 스택에서 꺼냄
    backStack.push(page);                       // 현재 페이지로 이동
    return page;
}

visit('https://www.google.com');
visit('https://www.naver.com');
visit('https://www.youtube.com');

console.log('1 goBack()   :', goBack());                          // -> naver
visit('https://www.github.com');                                  // forwardStack 초기화
console.log('2 visit(github) 후 forwardStack:', forwardStack);    // -> []
console.log('3 goBack()   :', goBack());                          // -> naver
console.log('4 goForward():', goForward());                       // -> github
console.log('5 goForward():', goForward());                       // -> null

console.log('\n최종 backStack   :', backStack);    // -> [google, naver, github]
console.log('최종 forwardStack:', forwardStack);  // -> []
````

### 문제 5: 설계 문제

다음 요구사항을 만족하도록 구조를 설계하시오.

- 최대 히스토리 저장 개수: 3개
- 3개 초과 시 가장 오래된 URL 자동 삭제
- Stack 구조 유지

````python
%%writefile test5.js

let backStack = [];
let forwardStack = [];
const MAX_HISTORY = 3; // 최대 히스토리 저장 개수

function visit(url) {
    backStack.push(url);

    // 최대 개수 초과 시 가장 오래된 URL 삭제 (앞에서 제거 -> Queue 방식)
    if (backStack.length > MAX_HISTORY) {
        backStack.shift(); // index 0 (가장 오래된 것) 제거
    }

    forwardStack = []; // 새 페이지 방문 시 forwardStack 초기화
}

function goBack() {
    if (backStack.length <= 1) return null;
    const current = backStack.pop();
    forwardStack.push(current);
    return backStack[backStack.length - 1];
}

function goForward() {
    if (forwardStack.length === 0) return null;
    const page = forwardStack.pop();
    backStack.push(page);
    return page;
}

// 테스트: 4개 방문 -> 가장 오래된 google 자동 삭제
visit('https://www.google.com');
visit('https://www.naver.com');
visit('https://www.youtube.com');
visit('https://www.github.com'); // google 삭제됨

console.log('backStack:', backStack);
// -> ['https://www.naver.com', 'https://www.youtube.com', 'https://www.github.com']

console.log('goBack():', goBack());
// -> https://www.youtube.com

console.log('backStack:', backStack);
// -> ['https://www.naver.com', 'https://www.youtube.com']


/* 핵심 포인트:

push() -> 뒤에 추가 (Stack)
shift() -> 앞에서 제거 (Queue)
=> 둘 다 쓰면 Deque (양방향 큐) 형태 */
````

````python
!node test5.js
````

## 정리

- 단순 push/pop 이해 수준은 실무에서는 부족
- 핵심은 **상태 전이(State Transition) 정확성**
- 특히 중요한 포인트
    - forwardStack 초기화 타이밍
    - boundary 조건 (빈 스택 처리)
    - 히스토리 제한 정책

## 2. 작업 대기열 시스템 (Queue)

- **FIFO (First-In, First-Out) 구조**
    - **선입선출**: 가장 먼저 들어온 작업이 가장 먼저 처리됩니다.
    - 실제 시스템에서는 요청 순서 보장을 위해 Queue 구조를 사용합니다.
- **현실 시나리오 (실제 서비스 흐름)**
    - 사용자 요청 발생 → 서버에 작업 등록
    - 작업1 → 작업2 → 작업3 순서로 처리
    - 예: 파일 업로드 → 이메일 발송 → 데이터 처리
- **핵심 동작**
    - 작업 등록 → queue.push(task)
    - 작업 처리 → queue.shift()
    - 현재 처리 대상 → queue[0]

````python
%%writefile eex1.js

// 작업 대기열 예제 1: 작업 등록

const queue = [];

function addTask(task) {
    queue.push(task); // 대기열 뒤에 추가 (FIFO -> 나중에 들어온 건 나중에 처리)
}

addTask('파일 업로드 요청');
addTask('이메일 발송 요청');
addTask('데이터 분석 요청');

console.log('현재 대기 작업:', queue);
// -> ['파일 업로드 요청', '이메일 발송 요청', '데이터 분석 요청']
````

````python
!node eex1.js
````

````python
%%writefile eex2.js

// 작업 대기열 예제 2: 작업 처리

const queue = ['파일 업로드 요청', '이메일 발송 요청', '데이터 분석 요청']; // eex1 실행 후 상태

function processTask() {
    if (queue.length === 0) return null; // 대기 작업 없으면 종료
    const task = queue.shift();          // 가장 먼저 들어온 작업 꺼냄 (FIFO)
    return task;
}

console.log('처리 작업:', processTask()); // -> 파일 업로드 요청
console.log('남은 대기열:', queue);       // -> ['이메일 발송 요청', '데이터 분석 요청']
````

````python
!node eex2.js
````

````python
%%writefile eex3.js

// 작업 대기열 예제 3: 현실 흐름

let queue = [];

function addTask(task) {
    queue.push(task); // 대기열 뒤에 추가
}

function processTask() {
    if (queue.length === 0) return null;
    return queue.shift(); // 앞에서 꺼냄 (FIFO)
}

addTask('이미지 업로드');
addTask('DB 저장');
addTask('알림 전송');

console.log('처리:', processTask()); // -> 이미지 업로드
console.log('처리:', processTask()); // -> DB 저장

console.log('남은 작업:', queue); // -> ['알림 전송']
````

````python
!node eex3.js
````

````python
%%writefile eex4.js

// 작업 대기열 예제 4: 서버 요청 처리 흐름

let queue = [];

function addTask(task) {
    queue.push(task); // 대기열 뒤에 추가
}

function processTask() {
    if (queue.length === 0) return null;
    return queue.shift(); // 앞에서 꺼냄 (FIFO)
}

addTask('회원가입 요청');
addTask('로그인 요청');
addTask('상품 조회 요청');

// 대기열 비워질 때까지 순차 처리
while (queue.length > 0) {
    console.log('처리:', processTask());
}

// -> 회원가입 요청
// -> 로그인 요청
// -> 상품 조회 요청
````

````python
!node eex4.js
````

````python
%%writefile eex5.js

// 작업 대기열 예제 5: 상태 확인

const queue = ['회원가입 요청', '로그인 요청', '상품 조회 요청']; // eex4 실행 전 상태

function printState() {
    console.log('현재 처리 대상:', queue[0]); // 가장 먼저 처리될 작업
    console.log('전체 대기열:', queue);
}

printState();
// -> 현재 처리 대상: 회원가입 요청
// -> 전체 대기열: ['회원가입 요청', '로그인 요청', '상품 조회 요청']
````

### 문제 5: 설계 문제

다음 요구사항을 만족하도록 구조를 설계하시오.

- shift() 사용 금지
- O(1) 성능 유지
- 인덱스 기반 처리

````python
%%writefile ttest5.js

// shift()는 앞 요소 제거 후 나머지를 전부 앞으로 당김 -> O(n)
// 인덱스 포인터를 앞으로 옮기는 방식으로 O(1) 유지

let queue = [];
let head = 0; // 현재 처리할 작업의 인덱스 포인터

function addTask(task) {
    queue.push(task); // 뒤에 추가 -> O(1)
}

function processTask() {
    if (head >= queue.length) return undefined; // 처리할 작업 없음
    const task = queue[head]; // 포인터 위치의 작업 가져옴
    head++;                   // 포인터를 한 칸 앞으로 -> 실제로 제거 안 하고 건너뜀
    return task;
}

// 테스트
addTask('요청1');
addTask('요청2');
addTask('요청3');

console.log(processTask()); // -> 요청1
console.log(processTask()); // -> 요청2
console.log(processTask()); // -> 요청3
console.log(processTask()); // -> undefined

/* ── shift() vs 포인터 방식 비교 ────────────
 *
 *  shift()    : 앞 요소 삭제 + 나머지 전부 앞으로 이동 -> O(n)
 *  포인터 방식: head 숫자만 1 증가 -> 실제 배열 안 건드림 -> O(1)
 *
 * ───────────────────────────── */
````

## 3. 주소록 검색 (Trie)

- **Prefix Tree 구조**
    - **접두사 기반 탐색**: 문자열을 문자 단위로 분해하여 트리 형태로 저장하고 검색합니다.
    - 실제 주소록 검색에서는 이름의 앞부분만 입력해도 빠르게 후보를 찾기 위해 Trie 구조를 사용합니다.
- **현실 시나리오 (실제 서비스 흐름)**
    - 사용자가 주소록 검색창에 `김` 입력
    - 시스템은 `김민수`, `김영희`, `김하늘`처럼 같은 접두사를 가진 이름을 빠르게 탐색
    - 예: 연락처 검색, 자동완성, 사내 인명 검색
- **핵심 동작**
    - 연락처 저장 → 문자 단위로 노드 생성
    - 접두사 검색 → prefix를 따라 이동
    - 추천 목록 출력 → 해당 prefix 아래의 모든 이름 수집

````python
%%writefile eeex1.js

// 주소록 검색 예제 1: 이름 저장

class TrieNode {
    constructor() {
        this.children = {}; // 다음 글자로 이어지는 자식 노드들
        this.isEnd = false; // 이 노드에서 끝나는 이름이 있는지 표시
    }
}

class Trie {
    constructor() {
        this.root = new TrieNode(); // 모든 이름의 시작점
    }

    insert(name) {
        let node = this.root;
        for (const char of name) {
            // 해당 글자 노드 없으면 새로 생성
            if (!node.children[char]) {
                node.children[char] = new TrieNode();
            }
            node = node.children[char]; // 다음 글자 노드로 이동
        }
        node.isEnd = true; // 마지막 글자 노드에 이름 끝 표시
    }
}

const contactTrie = new Trie();

contactTrie.insert('김민수');
contactTrie.insert('김영희');
contactTrie.insert('김하늘');

console.log('주소록 이름 저장 완료');
````

````python
%%writefile eeex2.js

// 주소록 검색 예제 2: prefix 존재 여부 확인

class TrieNode {
    constructor() {
        this.children = {};
        this.isEnd = false;
    }
}

class Trie {
    constructor() {
        this.root = new TrieNode();
    }

    insert(name) {
        let node = this.root;
        for (const char of name) {
            if (!node.children[char]) {
                node.children[char] = new TrieNode();
            }
            node = node.children[char];
        }
        node.isEnd = true;
    }
}

// prototype으로 메서드 추가 -> 기존 클래스 수정 없이 기능 확장
Trie.prototype.startsWith = function(prefix) {
    let node = this.root;
    for (const char of prefix) {
        // 경로 중간에 노드 없으면 해당 prefix 없음
        if (!node.children[char]) return false;
        node = node.children[char];
    }
    return true; // 끝까지 경로 있으면 prefix 존재
};

const contactTrie = new Trie();
contactTrie.insert('김민수');
contactTrie.insert('김영희');
contactTrie.insert('김하늘');

console.log(contactTrie.startsWith('김')); // -> true
console.log(contactTrie.startsWith('박')); // -> false
````

````python
%%writefile eeex3.js

// 주소록 검색 예제 3: 자동완성 후보 수집

class TrieNode {
    constructor() {
        this.children = {};
        this.isEnd = false;
    }
}

class Trie {
    constructor() {
        this.root = new TrieNode();
    }

    insert(name) {
        let node = this.root;
        for (const char of name) {
            if (!node.children[char]) {
                node.children[char] = new TrieNode();
            }
            node = node.children[char];
        }
        node.isEnd = true;
    }
}

// prefix 끝 노드 찾기 -> 없으면 null 반환
Trie.prototype.searchPrefixNode = function(prefix) {
    let node = this.root;
    for (const char of prefix) {
        if (!node.children[char]) return null; // 경로 없으면 null
        node = node.children[char];
    }
    return node; // prefix 마지막 글자 노드 반환
};

// 특정 노드 아래의 모든 이름을 재귀로 수집
Trie.prototype.collectNames = function(node, prefix, result) {
    if (node.isEnd) {
        result.push(prefix); // isEnd=true -> 완성된 이름 결과에 추가
    }
    for (const char in node.children) {
        // 자식 노드들을 재귀 탐색하며 이름 계속 완성
        this.collectNames(node.children[char], prefix + char, result);
    }
};

// prefix 입력 -> 해당 노드 찾기 -> 하위 이름 전부 수집
Trie.prototype.autoComplete = function(prefix) {
    const startNode = this.searchPrefixNode(prefix); // prefix 끝 노드
    const result = [];
    if (!startNode) return result; // prefix 없으면 빈 배열 반환
    this.collectNames(startNode, prefix, result); // 하위 이름 수집 시작
    return result;
};

const contactTrie = new Trie();
contactTrie.insert('김민수');
contactTrie.insert('김영희');
contactTrie.insert('김하늘');

console.log(contactTrie.autoComplete('김'));
// -> ['김민수', '김영희', '김하늘']
````

## 5. 캐시 시스템 (LRU)

- **LRU (Least Recently Used) 구조**
    - **가장 오래 사용되지 않은 데이터 제거**: 제한된 메모리 안에서 최근에 사용된 데이터는 유지하고, 오랫동안 사용되지 않은 데이터는 제거합니다.
    - 실제 시스템에서는 조회 속도를 높이기 위해 자주 사용하는 데이터를 캐시에 저장하고, 공간이 부족하면 가장 오래 안 쓴 항목부터 삭제합니다.
- **현실 시나리오 (실제 서비스 흐름)**
    - 사용자가 상품 상세 페이지를 조회
    - 시스템은 DB를 직접 조회하지 않고 먼저 캐시를 확인
    - 캐시에 있으면 즉시 반환
    - 캐시에 없으면 DB 조회 후 캐시에 저장
    - 저장 공간이 가득 찼다면 가장 오래 사용하지 않은 데이터를 제거
    - 예: 상품 상세 조회, 사용자 프로필 조회, 게시글 조회
- **핵심 동작**
    - 데이터 조회 → 캐시에 있으면 최근 사용으로 갱신
    - 데이터 저장 → 캐시에 추가
    - 용량 초과 → 가장 오래 사용되지 않은 항목 제거
    - 최근 사용 순서 유지 → Map의 삽입 순서 활용

````python
%%writefile ex11.js

// 캐시 시스템 예제 1: LRU 캐시 클래스

class LRUCache {
    constructor(limit) {
        this.limit = limit;
        this.cache = new Map();
    }

    get(key) {
        if (!this.cache.has(key)) return null;

        const value = this.cache.get(key);

        this.cache.delete(key);
        this.cache.set(key, value);

        return value;
    }

    put(key, value) {
        if (this.cache.has(key)) {
            this.cache.delete(key);
        }

        this.cache.set(key, value);

        if (this.cache.size > this.limit) {
            const oldestKey = this.cache.keys().next().value;
            this.cache.delete(oldestKey);
        }
    }
}

const productCache = new LRUCache(3);

console.log('LRU 캐시 생성 완료');
````

- **자료구조**: LRU Cache
- **핵심 로직**
    - 최근 사용 데이터는 뒤로 이동
    - 오래 사용하지 않은 데이터는 앞에서 제거
- **시간 복잡도**
    - Map 기반 조회: O(1)
    - 삽입: O(1)
    - 삭제: O(1)

    - 사용자 프로필 캐시
