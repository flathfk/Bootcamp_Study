# PART I 클라우드 컴퓨팅

## 1 클라우드 컴퓨팅의 본질과 패러다임 시프트

### 1.1 클라우드 컴퓨팅의 기술적 정의

클라우드 컴퓨팅은 컴퓨팅 자원을 전기나 수도와 같은 '공공재'의 성격으로 전환한 기술적 혁명이다. 이는 네트워크를 통해 구성 가능한 컴퓨팅 자원(예: 네트워크, 서버, 스토리지, 애플리케이션, 서비스)의 공유 풀에 대해 어디서나 편리하게, 주문형으로 접속할 수 있게 하는 모델이다.

- **NIST(미국 국립표준기술연구소) 정의 5대 특징:**
    1. **On-demand Self-service:** 인간의 개입 없이 사용자가 직접 자원을 할당받음.
    2. **Broad Network Access:** 다양한 플랫폼(모바일, 태블릿, PC 등)을 통해 표준 메커니즘으로 접속.
    3. **Resource Pooling:** 다중 임차(Multi-tenancy) 모델을 통해 물리적 자원을 여러 사용자에게 동적 할당.
    4. **Rapid Elasticity:** 수요에 따라 자원을 즉각적으로 확장하거나 축소 가능.
    5. **Measured Service:** 사용한 만큼 비용을 지불하는 투명한 측정 시스템.

### 1.2 온프레미스(On-Premise) 환경의 한계 분석

전통적인 IT 인프라 운영 방식인 온프레미스는 '소유'에 기반한다. 이는 다음과 같은 심각한 비효율성을 초래한다.

- **프로비저닝의 딜레마:** 최대 트래픽(Peak Time)을 기준으로 장비를 구매하면 평상시 자원 낭비가 발생하고, 평균 트래픽을 기준으로 구매하면 피크 타임에 서비스가 중단된다.
- **기술 부채의 누적:** 하드웨어 노후화에 따른 유지보수 비용 증가와 신기술 도입의 지연.
- **물리적 제약:** 데이터센터 상면(Space), 전력, 냉각 설비 등 물리적 확장의 한계.

### 1.3 경제적 가치: CAPEX에서 OPEX로의 전환

클라우드 도입의 가장 큰 비즈니스 동인은 재무적 유연성이다.

- **CAPEX (Capital Expenditure):** 대규모 초기 자본 투자. 장비 구매와 동시에 감가상각이 시작되며 자산의 유동성이 낮음.
- **OPEX (Operating Expenditure):** 운영 비용 처리. 사용한 만큼 비용을 지불하며, 비즈니스 상황에 따라 비용 지출을 즉각적으로 조절 가능(Pay-as-you-go).

## 2 클라우드 핵심 기술: 가상화와 추상화

### 2.1 가상화(Virtualization)의 동작 원리

가상화는 물리적 하드웨어와 운영체제 사이의 강한 결합을 소프트웨어 계층을 통해 분리하는 기술이다. 이를 통해 하나의 물리적 서버(Host) 위에 여러 개의 논리적 서버(Guest VM)를 구동할 수 있다.

- **추상화(Abstraction):** 하드웨어 자원을 논리적 단위로 변환.
- **격리(Isolation):** 하나의 가상 머신에서 발생한 오류나 보안 침해사고가 다른 가상 머신으로 전이되지 않도록 보호.
- **캡슐화(Encapsulation):** 가상 머신의 모든 상태를 파일 형태로 저장하여 이동 및 복제가 용이함.

### 2.2 하이퍼바이저(Hypervisor) 심층 비교

하이퍼바이저는 가상 머신을 생성하고 관리하며, 자원을 분배하는 '교통 관제소' 역할을 한다.

1. **Type 1 (Bare-metal):**
    - **특징:** 호스트 OS 없이 하드웨어 위에서 직접 실행.
    - **성능:** 오버헤드가 적고 안정성이 매우 높음.
    - **사용 사례:** AWS Nitro, VMWare ESXi, Xen, KVM (대규모 클라우드 데이터센터의 표준).
2. **Type 2 (Hosted):**
    - **특징:** 기존 OS(Windows, Linux 등) 위에서 애플리케이션으로 실행.
    - **성능:** 호스트 OS를 거쳐야 하므로 I/O 성능 저하 발생.
    - **사용 사례:** VMWare Workstation, Oracle VirtualBox (개인 개발 및 테스트용).

### 2.3 컨테이너(Container) 기술의 부상

가상 머신(VM)이 OS 전체를 가상화한다면, 컨테이너는 OS 커널을 공유하면서 애플리케이션 실행 환경만 격리한다.

- **장점:** VM보다 훨씬 가볍고 부팅 속도가 빠르며, 인프라 환경에 관계없이 동일한 동작(Portability)을 보장한다.
- **도커(Docker)와 쿠버네티스(Kubernetes):** 현대 클라우드 네이티브 아키텍처의 핵심으로, 수만 개의 컨테이너를 효율적으로 배포하고 관리(Orchestration)하는 체계를 제공한다.

## 3 글로벌 인프라 설계와 고가용성(High Availability)

클라우드의 진정한 위력은 전 세계에 분산된 물리적 거점을 소프트웨어로 제어하는 데서 온다. 엔지니어는 지리적 특성을 고려하여 서비스의 지연 시간(Latency)과 안정성을 설계해야 한다.

### 3.1 리전(Region) 선정 전략

리전은 클라우드 서비스가 제공되는 지리적 영역(예: 서울 리전, 버지니아 북부 리전)이다. 리전 선정 시 고려해야 할 4대 요소는 다음과 같다.

- **데이터 주권 및 컴플라이언스:** 국가별 법령에 따라 특정 데이터(예: 금융, 개인정보)는 반드시 해당 국가 영토 내에 저장되어야 한다.
- **지연 시간(Latency):** 서비스 주 사용자와 리전 사이의 물리적 거리가 가까울수록 응답 속도가 빠르다.
- **서비스 가용성:** 모든 리전이 동일한 서비스를 제공하지 않는다. 신규 기능은 특정 주요 리전에 먼저 배포되는 경향이 있다.
- **비용:** 리전마다 전력료, 인건비, 세금이 다르므로 동일한 자원이라도 리전별 이용료가 5~20%가량 차이 날 수 있다.

### 3.2 가용 영역(Availability Zone, AZ)과 결함 격리

하나의 리전은 최소 3개 이상의 AZ로 구성된다.

- **AZ의 물리적 독립성:** 각 AZ는 서로 수십 km 떨어져 있으며, 전력망(Substation), 냉각 시설, 네트워크 회선이 완전히 독립적이다.
- **데이터 동기화:** AZ 간에는 전용 광섬유로 연결되어 매우 낮은 지연 시간(1ms 미만)을 보장한다. 이를 통해 여러 AZ에 걸친 실시간 데이터 복제가 가능하다.

### 3.3 고가용성(High Availability) 및 재해 복구(DR)

- **Active-Active 구성:** 두 개 이상의 AZ에서 서비스를 동시에 구동하여 부하를 분산하고 장애 시 즉각 대응한다.
- **재해 복구(Disaster Recovery) 전략:**
    - **Pilot Light:** 최소한의 핵심 자원(주로 DB)만 타 리전에 상시 구동.
    - **Warm Standby:** 서비스가 가능한 최소 규모의 인프라를 타 리전에 미리 구축.
    - **Multi-Site:** 두 리전 이상에서 100% 동일한 규모로 서비스 운영 (가장 고비용, 높은 안정성).

## 4 클라우드 서비스 모델(XaaS) 및 책임 공유 모델

현대 클라우드는 단순 인프라(IaaS)를 넘어 플랫폼(PaaS)과 소프트웨어(SaaS)를 통합 제공하며, 최근에는 관리형 서비스(Managed Services)의 비중이 높아지고 있다.

### 4.1 서비스 모델별 기술 스택 제어권

- **IaaS (Infrastructure as a Service):** 하드웨어의 추상화. 엔지니어는 커널 파라미터 튜닝, OS 보안 강화(Hardening) 등 로우 레벨 제어가 가능하다.
- **PaaS (Platform as a Service):** 런타임의 추상화. 웹 서버(Nginx, Apache)나 데이터베이스 엔진의 직접적인 관리를 클라우드 제공사(CSP)에 위임한다. 개발자는 '코드'와 '데이터'에만 집중한다.
- **SaaS (Software as a Service):** 비즈니스 로직의 서비스화. 사용자는 애플리케이션의 내부 구조를 알 필요 없이 인터페이스만 활용한다.

### 4.2 서버리스(Serverless, FaaS) 아키텍처

서버리스는 "서버가 없다"는 뜻이 아니라, 사용자가 "서버를 관리하지 않는다"는 의미다.

- **Event-Driven:** 특정 이벤트(파일 업로드, API 호출, DB 변경 등) 발생 시에만 코드가 실행된다.
- **Zero Idle Cost:** 코드가 실행되지 않을 때는 비용이 0원이다.
- **Micro-Scaling:** 요청 1개당 1개의 컨테이너/프로세스가 대응하여 극단적인 확장성을 제공한다.

### 4.3 책임 공유 모델(Shared Responsibility Model)의 핵심

클라우드 보안 사고의 90% 이상은 사용자의 설정 오류에서 발생한다.

- **CSP의 책임 (Security 'of' the Cloud):** 데이터센터 물리 보안, 하드웨어 유지보수, 하이퍼바이저 패치.
- **사용자의 책임 (Security 'in' the Cloud):** 게스트 OS 패치, 방화벽 규칙 설정, IAM 권한 관리, 데이터 암호화.

## 5 가상 네트워크 아키텍처 (VPC & Networking)

클라우드 네트워크는 물리적 스위치와 케이블을 소프트웨어 정의 네트워크(SDN)로 구현한 것이다.

### 5.1 VPC(Virtual Private Cloud) 설계

VPC는 CSP의 거대 네트워크 망 안에 사용자만을 위해 할당된 논리적인 격리 공간이다.

- **IP 주소 설계 (CIDR):** 보통 RFC 1918 사설 IP 대역(10.0.0.0/16 등)을 사용하여 충분한 IP 주소를 확보한다.
- **서브네팅 전략:** 보안 경계에 따라 네트워크를 쪼갠다.
    - **Public Subnet:** 외부와 직접 통신이 필요한 자원(Bastion Host, Load Balancer).
    - **Private Subnet:** 외부 노출이 차단된 자원(Application Server, Database).

### 5.2 게이트웨이 및 연결성 엔진

- **Internet Gateway (IGW):** VPC와 인터넷 간의 양방향 통신 통로.
- **NAT Gateway:** 프라이빗 서브넷의 서버가 외부로 업데이트 등을 위해 '나가는' 통신만 허용하고, 외부에서 '들어오는' 접속은 차단한다.
- **VPC Peering:** 서로 다른 VPC 간의 내부 통신을 가능하게 한다. 암호화된 전용망을 사용하므로 보안성이 높다.

### 5.3 로드밸런싱(Load Balancing) 메커니즘

로드밸런서는 단일 장애점(SPOF)을 제거하고 트래픽을 분산하는 핵심 장치다.

- **L4 로드밸런서 (Network LB):** IP와 포트 번호를 기반으로 초고속 패킷 포워딩을 수행한다. 게임 서버나 대용량 스트리밍에 적합하다.
- **L7 로드밸런서 (Application LB):** HTTP 헤더, 쿠키, URL 경로를 분석하여 스마트한 라우팅을 수행한다. (예: `/api` 경로는 A 서버군으로, `/static` 경로는 B 서버군으로 전송)

## 6 클라우드 스토리지 및 데이터베이스 전략

데이터는 기업의 가장 중요한 자산이며, 클라우드 환경에서는 데이터의 성격(정형, 비정형, 빈도)에 따라 최적의 저장소를 선택하는 것이 성능과 비용의 핵심이다.

### 6.1 클라우드 스토리지 트리니티 (Block, Object, File)

1. **블록 스토리지 (Block Storage):**
    - **기술적 원리:** 데이터를 고정된 크기의 '블록'으로 나누어 저장하며, 가상 서버(VM)에 직접 마운트하여 사용한다.
    - **특징:** 지연 시간(Latency)이 매우 낮고 데이터 처리량(Throughput)이 높다. OS 부팅 드라이브, 데이터베이스 데이터 파일 저장에 필수적이다.
    - **주요 용어:** IOPS (초당 입출력 횟수) 설정이 성능의 척도가 된다.
2. **객체 스토리지 (Object Storage):**
    - **기술적 원리:** 데이터를 '객체' 단위로 관리하며, 메타데이터와 고유 식별자(ID)를 부여한다. 계층 없는 플랫 구조다.
    - **특징:** 무제한 확장성을 가지며 HTTP/HTTPS API를 통해 어디서든 접근 가능하다. 이미지, 영상, 백업 데이터, 로그 저장에 최적화되어 있다.
    - **데이터 내구성:** 보통 99.999999999% (11 nines)의 내구성을 설계하여 데이터 유실 가능성을 사실상 제로화한다.
3. **파일 스토리지 (File Storage):**
    - **기술적 원리:** 표준 NFS/SMB 프로토콜을 사용하는 공유 파일 시스템이다.
    - **특징:** 여러 대의 가상 서버가 동시에 동일한 저장소에 접근하여 파일을 읽고 쓸 수 있다. 콘텐츠 관리 시스템(CMS)이나 공용 홈 디렉토리에 사용된다.

### 6.2 데이터베이스(DB) 현대화: RDBMS vs NoSQL

클라우드 제공사는 설치형 DB(EC2 installed) 대신 **관리형 DB 서비스(Managed DB)** 사용을 권장한다. 패치, 백업, 고가용성 구성을 CSP가 대신 수행하기 때문이다.

- **관계형 데이터베이스 (RDBMS):**
    - **강점:** ACID 트랜잭션 보장, 데이터 무결성.
    - **활용:** 금융 결제, 사용자 정보 관리, 복잡한 쿼리가 필요한 서비스.
- **NoSQL 데이터베이스:**
    - **강점:** 수평적 확장(Scale-out) 용이, 유연한 스키마.
    - **활용:** 실시간 게임 순위, IoT 센서 데이터 수집, 대규모 세션 관리.
    - **유형:** Key-Value(Redis), Document(MongoDB), Wide-column(Cassandra), Graph(Neo4j).

## 7 클라우드 보안 및 거버넌스 (IAM & Compliance)

클라우드 환경에서는 물리적 장벽이 없으므로, **자격 증명**(Identity)이 새로운 보안 경계(Perimeter)가 된다.

### 7.1 IAM(Identity and Access Management) 설계의 정석

IAM은 "누가(Principal) 어떤 자원(Resource)에 대해 어떤 행위(Action)를 할 수 있는가"를 정의한다.

- **최소 권한의 원칙 (Least Privilege):** 업무 수행에 필요한 최소한의 권한만 부여한다. 예를 들어, 개발자에게 DB 삭제 권한을 주지 않는 식이다.
- **ABAC vs RBAC:** * **RBAC (Role-based):** '관리자', '개발자' 등 역할에 따라 권한 부여.
    - **ABAC (Attribute-based):** 부서, 프로젝트명, 시간대 등 속성값에 따라 동적으로 권한 제어.
- **MFA (Multi-Factor Authentication):** 단순 비밀번호 외에 OTP, 생체 인식 등을 추가하여 계정 탈취를 원천 차단한다.

### 7.2 데이터 암호화 및 키 관리

- **At-Rest 암호화:** 저장된 데이터(디스크, S3 객체)를 암호화한다. 설령 물리 디스크가 탈취되어도 내용을 볼 수 없다.
- **In-Transit 암호화:** 네트워크 구간 이동 시 TLS/SSL 프로토콜로 데이터를 보호한다.
- **KMS (Key Management Service):** 암호화 키를 생성, 보관, 회전(Rotation)시키는 중앙 집중식 관리 서비스다. 하드웨어 보안 모듈(HSM)을 사용하여 키 자체의 유출을 물리적으로 방지한다.

## 8 데브옵스(DevOps)와 클라우드 자동화

클라우드의 장점을 극대화하려면 인프라 관리 방식이 '수동'에서 '코드'로 변해야 한다.

### 8.1 코드형 인프라 (IaC, Infrastructure as Code)

인프라를 프로그래밍 코드로 정의하여 버전 관리 시스템(Git)에 저장하고 실행한다.

- **선언적 방식:** "서버 3대, DB 1개를 만들어줘"라고 정의하면 툴(Terraform 등)이 현재 상태를 분석하여 목표 상태로 만든다.
- **장점:** 인프라 구축의 재현성 보장, 휴먼 에러 제거, 인프라 변경 이력 추적 가능.

### 8.2 CI/CD 파이프라인과 배포 전략

클라우드 환경은 자원 교체가 쉬우므로 중단 없는 배포(Zero Downtime Deployment)가 기본이다.

- **Blue-Green 배포:** 기존 환경(Blue)과 동일한 신규 환경(Green)을 통째로 만들어 트래픽을 한꺼번에 전환한다. 장애 시 즉각 롤백이 가능하다.
- **Canary 배포:** 신규 버전의 서버를 1대만 투입하여 일부 사용자(5%)에게만 노출한 뒤, 문제가 없으면 점진적으로 확대한다.

### 8.3 모니터링 및 옵저버빌리티 (Observability)

단순히 "서버가 떠 있는가"를 넘어 시스템 내부를 깊이 있게 관찰해야 한다.

- **Metric:** CPU 사용량, 메모리 점유율 등 수치 데이터.
- **Log:** 애플리케이션 실행 기록 및 에러 메시지.
- **Trace:** 사용자 요청이 여러 마이크로서비스를 거칠 때의 흐름 추적.

## 9 클라우드 비용 최적화 (FinOps)

클라우드 도입 후 기업이 직면하는 가장 큰 과제는 '통제되지 않는 비용'이다. FinOps는 재무(Finance)와 운영(DevOps)의 합성어로, 클라우드 소비 모델에 맞게 비용을 관리하고 비즈니스 가치를 극대화하는 문화적 프랙티스다.

### 9.1 클라우드 과금 모델의 전략적 활용

사용자는 워크로드의 특성에 따라 최적의 구매 옵션을 조합해야 한다.

- **On-Demand (온디맨드):** 약정 없이 초 단위로 지불. 확정되지 않은 신규 서비스나 테스트 환경에 적합하다.
- **Reserved Instances (RI, 예약 인스턴스):** 1년 또는 3년 약정을 통해 최대 72%까지 할인받는다. 24시간 상시 구동되는 DB나 웹 서버에 필수적이다.
- **Savings Plans:** 특정 인스턴스 타입이 아닌 '시간당 지출 금액'을 약정하여 유연하게 할인을 적용받는 현대적 모델이다.
- **Spot Instances (스팟 인스턴스):** 클라우드사의 유휴 자원을 경매 방식으로 낙찰받아 최대 90% 저렴하게 이용한다. 단, 자원이 회수될 수 있으므로 중단되어도 무관한 배치 작업이나 가상화 학습(ML)에 사용한다.

### 9.2 비용 누수 차단 및 거버넌스 수립

- **Right-sizing (최적화):** 실제 사용률이 10% 미만인 고사양 서버를 저사양으로 하향 조정한다.
- **자원 태깅(Tagging):** 모든 자원에 '부서', '프로젝트', '소유자' 태그를 부여하여 비용의 출처를 명확히 시각화한다.
- **Unused Resources 삭제:** 연결되지 않은 볼륨(EBS), 사용하지 않는 탄력적 IP, 방치된 스냅샷 등을 정기적으로 청소하여 고정 지출을 줄인다.

## 10 미래 기술과 클라우드의 확장

클라우드는 단순한 서버 대여를 넘어 인류의 기술적 한계를 극복하는 거대한 플랫폼으로 진화하고 있다.

### 10.1 인공지능(AI)과 클라우드의 결합

현재의 생성형 AI 혁명은 클라우드 인프라가 없었다면 불가능했을 것이다.

- **GPU 팜(Farm):** 수만 개의 GPU를 고속 네트워크(InfiniBand)로 연결하여 거대언어모델(LLM)을 학습시키는 환경을 제공한다.
- **Managed AI:** 모델 학습(SageMaker, Vertex AI)부터 배포까지의 전 과정을 자동화하여, 기업이 모델 아키텍처에만 집중할 수 있게 한다.

### 10.2 에지 컴퓨팅(Edge Computing)

데이터가 생성되는 물리적 장소와 가까운 곳에서 처리를 수행하는 기술이다.

- **초저지연 서비스:** 자율주행, 원격 수술, 실시간 스마트 팩토리 등 1ms의 지연 시간도 허용되지 않는 분야에서 중앙 클라우드와 에지 노드가 협업한다.

### 10.3 양자 컴퓨팅(Quantum Computing) 클라우드

아직 초기 단계이나, 일반 기업이 수조 원대의 양자 컴퓨터를 소유할 수는 없다. 클라우드 제공사는 양자 컴퓨터에 대한 접근 권한을 API로 제공하여 차세대 암호 해독 및 신약 개발 연구를 지원하고 있다.
