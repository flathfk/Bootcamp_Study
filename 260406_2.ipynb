```python
%%writefile /content/bootcamp/src/Welcome.jsx

/**
 * 15.1 기본 구조 설명
 */

// Welcome이라는 이름의 함수형 컴포넌트를 정의합니다.
// React 컴포넌트 이름은 반드시 대문자로 시작해야 합니다.
function Welcome() {
  // return 키워드 뒤에 화면에 그려질 UI(JSX)를 작성합니다.
  return (
    // HTML처럼 보이지만 JavaScript의 확장 문법인 JSX입니다.
    <h1>안녕하세요</h1>
  );
}

// 이 파일을 다른 파일(예: App.js)에서 불러와서 사용할 수 있도록 내보냅니다.
export default Welcome;
```

```python
%%writefile /content/bootcamp/src/App.jsx
import React from 'react';

// 1. 외부 파일에 정의된 Welcome 컴포넌트를 가져옵니다.
// './Welcome'은 현재 파일과 같은 위치에 있는 Welcome.js 파일을 의미합니다.
import Welcome from './Welcome';

// 2. App이라는 이름의 메인 컴포넌트를 정의합니다.
// 이 컴포넌트는 우리 애플리케이션의 전체적인 틀(Root) 역할을 합니다.
function App() {
  // 3. 컴포넌트가 화면에 나타낼 내용을 return 문 안에 작성합니다.
  return (
    // 4. HTML의 div 태그를 사용하여 구조를 잡습니다.
    <div>
      {/* 5. 위에서 import한 Welcome 컴포넌트를 호출합니다.
         이렇게 작성하면 Welcome 함수가 실행되어 그 안의 JSX 내용이 이 자리에 삽입됩니다.
      */}
      <Welcome />
    </div>
  );
}

// 6. 작성한 App 컴포넌트를 다른 곳(보통 index.js)에서 사용할 수 있도록 내보냅니다.

export default App;
```
