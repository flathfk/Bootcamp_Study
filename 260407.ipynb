```python
%%writefile /content/bootcamp/src/App.jsx

// useState: 컴포넌트 안에서 변하는 데이터(상태)를 관리하는 Hook
import { useState } from 'react';

function App() {
  // [count]: 현재 숫자값 (화면에 표시됨)
  // [setCount]: count를 바꾸는 함수 (직접 count = 1 이렇게 못 바꾸고 얘를 써야 함)
  // useState(0): 초기값 0으로 시작
  const [count, setCount] = useState(0);

  function increase() {
    // setCount 호출 -> count 값 변경 -> 화면 자동 재렌더링
    // count를 직접 수정하면 화면이 안 바뀜, 반드시 setCount 사용
    setCount(count + 1);
  }

  return (
    <div>
      {/* count 값이 바뀔 때마다 이 부분만 다시 그려짐 */}
      <h1>{count}</h1>
      {/* onClick: 버튼 클릭 시 increase 함수 실행 */}
      <button onClick={increase}>증가</button>
    </div>
  );
}

export default App;
```

```python
%%writefile /content/bootcamp/src/App.jsx

import { useState, useEffect } from 'react';

function App() {
  // [1] 초기값 0으로 count state 선언
  const [count, setCount] = useState(0);

  // [5] count가 바뀔 때마다 useEffect 실행
  // 의존성 배열 [count] -> count 변경 시 자동 실행
  useEffect(() => {
    // [6] 현재 count 값 콘솔 출력
    console.log('count:', count);
  }, [count]);

  return (
    <div>
      {/* [4] count state 바뀌면 여기 숫자 자동 갱신 */}
      <h1>{count}</h1>

      {/* [2] 버튼 클릭 -> [3] setCount 실행 -> count + 1 */}
      <button onClick={() => setCount(count + 1)}>증가</button>
    </div>
  );
}

export default App;
```

```python
%%writefile /content/bootcamp/src/App.jsx

import { useState, useEffect } from 'react';

function App() {
  // count: 버튼 클릭으로 증가하는 원래 값
  const [count, setCount] = useState(0);
  // double: count의 2배 값 — useEffect가 자동으로 계산해서 저장
  const [double, setDouble] = useState(0);

  // count가 바뀔 때마다 실행 -> double을 count * 2로 갱신
  // 하나의 state 변화가 다른 state 계산으로 이어지는 패턴
  useEffect(() => {
    setDouble(count * 2);
  }, [count]); // count가 의존성 -> count 바뀔 때만 실행

  return (
    <div>
      <p>원래 값: {count}</p>   {/* count state 표시 */}
      <p>2배 값: {double}</p>   {/* useEffect가 계산한 double 표시 */}
      {/* 클릭 -> setCount -> count 증가 -> useEffect 실행 -> double 갱신 */}
      <button onClick={() => setCount(count + 1)}>증가</button>
    </div>
  );
}

export default App;
```

```python
%%writefile /content/bootcamp/src/App.jsx

import { useState } from 'react';

function App() {
  const [name, setName] = useState('');

  function handleSubmit(e) {
    e.preventDefault();
    console.log('제출값:', name);
  }

  return (
    <form onSubmit={handleSubmit}>
      <input type="text" onChange={e => setName(e.target.value)} />
      <button type="submit">제출</button>
    </form>
  );
}

export default App;
```
